```python
%%sql
CREATE DATABASE baitap_db
```

```python
%%sql
USE baitap_db;

CREATE TABLE course (
    id INT PRIMARY KEY,
    course_name VARCHAR(100) NOT NULL
);

CREATE TABLE student (
    student_id INT PRIMARY KEY,
    name VARCHAR(100) NOT NULL,
    class VARCHAR(50),
    course_id INT,
    score FLOAT,
    FOREIGN KEY (course_id) REFERENCES course(id)
);
```

```python
%%sql
SELECT *
FROM student s
JOIN course c ON s.course_id = c.id;
```

```python
%%sql
SELECT s.name, c.course_name, s.score
FROM student s
INNER JOIN course c
ON s.course_id = c.id;
```

```python
%%sql
SELECT s.name, c.course_name, s.score
FROM student s
LEFT JOIN course c
ON s.course_id = c.id;
```

```python
%%sql
SELECT s.name, c.course_name, s.score
FROM student s
RIGHT JOIN course c
ON s.course_id = c.id;
```

```python
%%sql
SELECT s.*, c.course_name 
FROM student s 
FULL OUTER JOIN course c ON s.course_id = c.id;
```

```python
%%sql
SELECT s.name, c.course_name, s.score
FROM student s
LEFT JOIN course c
ON s.course_id = c.id

UNION

SELECT s.name, c.course_name, s.score
FROM student s
RIGHT JOIN course c
ON s.course_id = c.id;
```

```python
%%sql
UPDATE student
SET course_id = 12
WHERE course_id IS NULL;
```

```python
%%sql
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course);
```

```python
%%sql
SELECT 
    class,
    COUNT(*) AS tong_sinh_vien,
    ROUND(AVG(score), 2) AS diem_trung_binh
FROM student
GROUP BY class;
```

```python
%%sql
SELECT 
    c.course_name,
    COUNT(s.student_id) AS tong_sinh_vien,
    ROUND(AVG(s.score), 2) AS diem_trung_binh
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name;
```

```python
%%sql
SELECT 
    c.course_name,
    ROUND(AVG(s.score), 2) AS diem_trung_binh,
    CASE 
        WHEN AVG(s.score) >= 9.0 THEN 'Xuat sac'
        WHEN AVG(s.score) >= 6.0 THEN 'Tot'
        ELSE 'Kem'
    END AS xep_loai
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name;
```

```python
%%sql
SELECT 
    student_id,
    name,
    class,
    score,
    RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student;
```

```python
%%sql
SELECT *
FROM (
    SELECT 
        student_id,
        name,
        score,
        RANK() OVER (ORDER BY score DESC) AS r
    FROM student
) t
WHERE r <= 3;
```

```python
%%sql
SELECT *
FROM (
    SELECT 
        student_id,
        name,
        score,
        RANK() OVER (ORDER BY score ASC) AS r
    FROM student
) t
WHERE r <= 3;
```

```python
%%sql
SELECT 
    student_id,
    name,
    class,
    score,
    RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_in_class
FROM student;
```

```python
%%sql
SELECT *
FROM (
    SELECT 
        student_id,
        name,
        class,
        score,
        RANK() OVER (PARTITION BY class ORDER BY score DESC) AS r
    FROM student
) t
WHERE r <= 3;
```

```python
%%sql
ALTER TABLE student
ADD COLUMN graduation_date DATETIME;
```

```python
%%sql
UPDATE student s
JOIN (
    SELECT 
        student_id,
        RANK() OVER (ORDER BY score DESC) AS r
    FROM student
) t ON s.student_id = t.student_id
SET s.graduation_date = NOW() + INTERVAL t.r DAY;
```

```python
%%sql
SELECT student_id, name, score, graduation_date
FROM student
ORDER BY score DESC;
```
