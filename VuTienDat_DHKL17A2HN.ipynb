```python
import sqlite3
import pandas as pd
```

**Câu 1**

```python
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE student (
    student_id INT,
    name TEXT,
    class TEXT,
    course_id INT,
    score REAL
)
""")

cursor.execute("""
CREATE TABLE course (
    id INT,
    course_name TEXT
)
""")
```

Thêm dữ liệu

```python
cursor.executemany("""
INSERT INTO student VALUES (?, ?, ?, ?, ?)
""", [
(1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
(2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
(3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
(4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
(5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
(6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
(7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
(8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
(9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
(10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
])

cursor.executemany("""
INSERT INTO course VALUES (?, ?)
""", [
(12, 'Giai tich'),
(34, 'Thong ke'),
(26, 'Tin hoc')
])

conn.commit()
```

INNER JOIN

```python
pd.read_sql_query("""
SELECT *
FROM student s
INNER JOIN course c
ON s.course_id = c.id
""", conn)
```

LEFT JOIN

```python
pd.read_sql_query("""
SELECT *
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
""", conn)
```

RIGHT JOIN

```python
pd.read_sql_query("""
SELECT *
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
""", conn)
```

FULL OUTER JOIN

```python
pd.read_sql_query("""
SELECT *
FROM student s
LEFT JOIN course c
ON s.course_id = c.id

UNION

SELECT *
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
""", conn)
```

**Kết luận**

- Các phép JOIN giúp kết nối dữ liệu giữa hai bảng student và course.
- INNER JOIN chỉ giữ các bản ghi khớp giữa hai bảng.
- LEFT JOIN giữ toàn bộ sinh viên kể cả khi chưa có môn học.
- RIGHT JOIN và FULL OUTER JOIN (mô phỏng trong SQLite) giúp quan sát đầy đủ dữ liệu từ cả hai bảng.
- Tích Descartes tạo ra tất cả các tổ hợp dữ liệu, thường dùng cho mục đích kiểm thử hoặc phân tích.

# Câu 2 : Cập nhật và làm sạch dữ liệu

```python
cursor.execute("""
UPDATE student
SET course_id = 26
WHERE course_id IS NULL
""")
conn.commit()
```

```python
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")
conn.commit()
```

**a.Tổng sinh viên và điểm trung bình theo lớp**

```python
pd.read_sql_query("""
SELECT class,
       COUNT(*) AS total_students,
       ROUND(AVG(score),2) AS avg_score
FROM student
GROUP BY class
""", conn)
```

**b.Tổng sinh viên và điểm trung bình theo môn**

```python
pd.read_sql_query("""
SELECT c.course_name,
       COUNT(*) AS total_students,
       ROUND(AVG(s.score),2) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)
```

**c.Phân loại theo điểm**

```python
pd.read_sql_query("""
SELECT c.course_name,
       ROUND(AVG(s.score),2) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)
```

- Dữ liệu đã được làm sạch bằng cách cập nhật các giá trị course_id còn thiếu bằng các giá trị hợp lệ từ bảng course và loại bỏ các bản ghi không thỏa mãn ràng buộc khóa ngoại.

- Sau khi chuẩn hóa dữ liệu, việc thống kê cho thấy:
  + Có sự khác biệt về số lượng sinh viên và điểm trung bình giữa các lớp.
  + Mỗi môn học có mức điểm trung bình riêng phản ánh chất lượng học tập.

- Dựa trên điểm trung bình, các môn học được phân loại thành:
  + ≥ 9.0: Xuất sắc
  + 6.0 – 8.9: Tốt
  + < 6.0: Kém

- Việc xử lý và phân tích đảm bảo tính toàn vẹn dữ liệu và hỗ trợ đánh giá kết quả học tập một cách rõ ràng, nhất quán.

# Câu 3

**a.Xếp hạng sinh viên qua điểm số**


```python
pd.read_sql_query("""
SELECT *,
       RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student
""", conn)
```

**b.Xếp hạng sinh viên qua điểm số theo lớp học**

```python
pd.read_sql_query("""
SELECT *,
       RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_class
FROM student
""", conn)
```

**c.Xếp hạng sinh viên qua điểm số theo mã môn học**

```python
pd.read_sql_query("""
SELECT s.*, c.course_name,
       RANK() OVER (PARTITION BY s.course_id ORDER BY score DESC) AS rank_course
FROM student s
JOIN course c ON s.course_id = c.id
""", conn)
```

**Top 3 sinh viên hạng cao nhất**

```python
pd.read_sql_query("""
SELECT *
FROM student
ORDER BY score DESC
LIMIT 3
""", conn)
```

**Top 3 sinh viên hạng thấp nhất**

```python
pd.read_sql_query("""
SELECT *
FROM student
ORDER BY score ASC
LIMIT 3
""", conn)
```

**Kết luận**

- Sử dụng hàm RANK() cho phép xếp hạng sinh viên theo điểm một cách linh hoạt.
- Có thể xếp hạng theo toàn bộ dữ liệu, theo lớp hoặc theo từng môn học.
- Việc lấy top 3 cao nhất và thấp nhất giúp nhanh chóng xác định sinh viên nổi bật và sinh viên cần cải thiện.
- Kết quả xếp hạng hỗ trợ đánh giá và so sánh hiệu quả học tập một cách rõ ràng.

# Câu 4

```python
cursor.execute("""
ALTER TABLE student ADD COLUMN graduation_date TEXT
""")
```

```python
cursor.execute("""
UPDATE student
SET graduation_date = DATE('now', '+' || (
    SELECT rank_score
    FROM (
        SELECT student_id,
               RANK() OVER (ORDER BY score DESC) AS rank_score
        FROM student
    ) r
    WHERE r.student_id = student.student_id
) || ' days')
""")
conn.commit()
```

```python
pd.read_sql_query("SELECT * FROM student", conn)
```

**Kết luận**

- Việc bổ sung cột graduation_date giúp mô phỏng thời gian tốt nghiệp của sinh viên.
- Thời gian tốt nghiệp được tính dựa trên thời điểm hiện tại và thứ hạng học tập.
- Sinh viên có thứ hạng cao sẽ có thời gian tốt nghiệp sớm hơn.
- Cách tiếp cận này thể hiện khả năng áp dụng dữ liệu để xây dựng các mô hình tính toán thực tế.
