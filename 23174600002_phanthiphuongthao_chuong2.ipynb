```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Tạo bảng student
cursor.execute("""
CREATE TABLE student (
    student_id INT,
    name TEXT,
    class TEXT,
    course_id INT,
    score FLOAT
)
""")

# Insert dữ liệu
students = [
    (1,"Nguyen Minh Hoang","May Tinh",12,6.7),
    (2,"Tran Thi Lan","Kinh Te",34,9.2),
    (3,"Pham Van Nam","Toan Tin",None,7.9),
    (4,"Le Thanh Huyen","Toan Tin",20,7.2),
    (5,"Vu Quoc Anh","May Tinh",24,8.0),
    (6,"Dang Thuy Linh","May Tinh",24,5.5),
    (7,"Bui Tien Dung","Kinh Te",34,9.2),
    (8,"Ho Ngoc Mai","Toan Tin",20,8.8),
    (9,"Duong Huu Phuc","Kinh Te",None,7.2),
    (10,"Cao Thi Hanh","May Tinh",None,7.0)
]

cursor.executemany("INSERT INTO student VALUES (?,?,?,?,?)", students)

# Tạo bảng course
cursor.execute("""
CREATE TABLE course (
    id INT,
    course_name TEXT
)
""")

courses = [
    (12,"Giai tich"),
    (34,"Thong ke"),
    (26,"Tin hoc")
]

cursor.executemany("INSERT INTO course VALUES (?,?)", courses)

conn.commit()
```

```python
# Yêu cầu 1: kết nối hai bảng 
#  tích Decartes
pd.read_sql("""
            SELECT *
            FROM student, course
            """, conn)
```

```python
# Sử dụng JOIN: INNER JOIN, LEFT JOIN, RIGHT JOIN, FULL OUTER JOIN.
# INNER JOIN
pd.read_sql("""
            SELECT *
            FROM student
            INNER JOIN course ON student.course_id = course.id
            """, conn)
```

```python
# LEFT JOIN
pd.read_sql("""
            SELECT *
            FROM student
            LEFT JOIN course ON student.course_id = course.id
            """, conn)
```

```python
# RIGHT JOIN
pd.read_sql("""
            SELECT *
            FROM student
            RIGHT JOIN course ON student.course_id = course.id
            """, conn)
```

```python
#FULL OUTER JOIN.
pd.read_sql("""
            SELECT *
            FROM student
            FULL OUTER JOIN course ON student.course_id = course.id
            """, conn)
```

```python
#Yêu cầu 2:
# cập nhật những giá trị course_id còn thiếu trong bảng student
cursor.execute("""
            UPDATE student
            SET course_id = 26
            WHERE course_id IS NULL
            """)
```

```python
#  loại bỏ những bản ghi tham gia những môn học không tồn tại bảng course
cursor.execute("""
            DELETE FROM student
            WHERE course_id NOT IN (SELECT id FROM course)
            """)

```

```python
# Câu a: Tổng số sinh viên, điểm trung bình của từng lớp
pd.read_sql("""
            SELECT 
                class,
                COUNT(*) AS total_students, 
                AVG(score) AS avg_score 
            FROM student
            GROUP BY class
            """, conn)

```

```python
# b. Tổng số sinh viên, điểm trung bình của từng môn học.
pd.read_sql("""
            SELECT 
                c.course_name,
                COUNT(*) AS total_students,
                AVG(s.score) AS avg_score
            FROM student s
            INNER JOIN course c ON s.course_id = c.id
            GROUP BY c.course_name
            """, conn)
```

```python
# c. Phân loại thi đua theo số điểm
pd.read_sql("""
            SELECT 
                c.course_name,
                AVG(s.score) AS avg_score,
                CASE 
                    WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
                    WHEN AVG(s.score) BETWEEN 6 AND 8.9 THEN 'Tot'
                    ELSE 'Kem'
                END AS classification
            FROM student s
            INNER JOIN course c ON s.course_id = c.id
            GROUP BY c.course_name
            """, conn)
```

```python
#Câu 3
pd.read_sql("""
            SELECT 
                student_id,
                name,
                class,
                score,
                ROW_NUMBER() OVER (ORDER BY score DESC) AS rank_all
            FROM student
            """, conn)
```

```python
# Top 3 sinh viên có điểm cao nhất trong mỗi môn học.
pd.read_sql("""
            SELECT * FROM (
                SELECT 
                    student_id,
                    name,
                    class,
                    score,
                    ROW_NUMBER() OVER (ORDER BY score DESC) AS rank_num
                FROM student
                ) AS ranked_students    
            WHERE rank_num <= 3
            """, conn)
```

```python
# Top 3 sinh viên có điểm thấp nhất trong mỗi môn học.
pd.read_sql("""
            SELECT * FROM (
                SELECT 
                    student_id,
                    name,
                    class,
                    score,
                    ROW_NUMBER() OVER (ORDER BY score ASC) AS rank_num
                FROM student
                ) AS ranked_students
            WHERE rank_num <= 3
            """, conn)
```

```python
#b Lớp học
pd.read_sql("""
            SELECT 
                student_id,
                name,
                class,
                score,
                ROW_NUMBER() OVER (
                    PARTITION BY class 
                    ORDER BY score DESC
                ) AS rank_in_class
            FROM student
            """, conn)
```

```python
# Top 3 sinh viên có điểm cao nhất trong mỗi lớp.
pd.read_sql("""
            SELECT *
            FROM (
                SELECT 
                student_id,
                name,
                class,
                score,
                RANK() OVER (
                    PARTITION BY class 
                    ORDER BY score DESC
                ) AS rank_high
                FROM student
            ) AS ranked_students
            WHERE rank_high = 1
            """, conn)
```

```python
#Top 3 sinh viên có điểm thấp nhất trong mỗi lớp.
pd.read_sql("""
            SELECT *
            FROM (
                SELECT 
                    student_id,
                    name,
                    class,
                    score,
                    RANK() OVER (
                        PARTITION BY class 
                        ORDER BY score ASC
                    ) AS rank_low
                FROM student
            ) AS ranked_students
            WHERE rank_low = 1
            """, conn)
```
