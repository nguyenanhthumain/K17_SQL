# Load bảng data

```python
import sqlite3
import pandas as pd

# Tạo database
conn = sqlite3.connect(':memory:')

# Tạo bảng + insert data
conn.executescript("""
CREATE TABLE student (
    student_id INT,
    name TEXT,
    class TEXT,
    course_id INT,
    score FLOAT
);

CREATE TABLE course (
    id INT,
    course_name TEXT
);

INSERT INTO student VALUES
(1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
(2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
(3, 'Pham Van Nam', 'Toan Tin', NULL, 7.9),
(4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
(5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
(6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
(7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
(8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
(9, 'Duong Huu Phuc', 'Kinh Te', NULL, 7.2),
(10, 'Cao Thi Hanh', 'May Tinh', NULL, 7.0);

INSERT INTO course VALUES
(12, 'Giai tich'),
(34, 'Thong ke'),
(26, 'Tin hoc');
""")
```

## Câu 1

### 1.Tích Descartes (CROSS JOIN)

```python
query = """
SELECT *
FROM student
CROSS JOIN course
"""
pd.read_sql_query(query, conn)
```

### 2.INNER JOIN

```python
query = """
SELECT s.*, c.course_name
FROM student s
INNER JOIN course c
ON s.course_id = c.id
"""
pd.read_sql_query(query, conn)
```

### 3.LEFT JOIN

```python
query = """
SELECT s.*, c.course_name
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
"""
pd.read_sql_query(query, conn)
```

### 4.RIGHT JOIN (SQLite không có → đảo bảng)

```python
query = """
SELECT s.*, c.course_name
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""
pd.read_sql_query(query, conn)
```

### 5.FULL OUTER JOIN (dùng UNION)

```python
query = """
SELECT s.*, c.course_name
FROM student s
LEFT JOIN course c
ON s.course_id = c.id

UNION

SELECT s.*, c.course_name
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""
pd.read_sql_query(query, conn)
```

## 📝 Nhận xét từng phép JOIN

### 🔹 1. CROSS JOIN
- Kết quả tạo ra tất cả các tổ hợp giữa bảng `student` và `course`.
- Số lượng bản ghi rất lớn (student × course).
- Không có điều kiện nối nên dữ liệu bị lặp nhiều.
- Ít dùng trong thực tế, chủ yếu dùng khi cần sinh tổ hợp dữ liệu.

---

### 🔹 2. INNER JOIN
- Chỉ giữ lại các bản ghi có liên kết hợp lệ giữa hai bảng.
- Các sinh viên không có `course_id` hoặc không khớp sẽ bị loại bỏ.
- Kết quả gọn, chính xác và có ý nghĩa nhất.
- Đây là phép JOIN được sử dụng phổ biến nhất.

---

### 🔹 3. LEFT JOIN
- Giữ toàn bộ dữ liệu từ bảng `student`.
- Những sinh viên không có môn học tương ứng sẽ có giá trị `NULL` ở cột `course_name`.
- Giúp phát hiện các dữ liệu bị thiếu liên kết.
- Phù hợp khi cần giữ đầy đủ danh sách sinh viên.

---

### 🔹 4. RIGHT JOIN (giả lập)
- Do SQLite không hỗ trợ RIGHT JOIN nên thực hiện bằng cách đảo bảng và dùng LEFT JOIN.
- Giữ toàn bộ dữ liệu từ bảng `course`.
- Các môn học không có sinh viên vẫn xuất hiện trong kết quả.
- Cách làm đúng và phù hợp với SQLite.

---

### 🔹 5. FULL OUTER JOIN (dùng UNION)
- Kết hợp toàn bộ dữ liệu từ cả hai bảng.
- Bao gồm:
  - Sinh viên không có môn học
  - Môn học không có sinh viên
- Sử dụng UNION của hai LEFT JOIN để mô phỏng.
- Cho cái nhìn đầy đủ và tổng quát nhất về dữ liệu.

#### 🔹Nhận xét Bài 2.1: Làm sạch dữ liệu

### 2a. Tổng số sinh viên & điểm TB theo từng lớp

```python
query = """
SELECT class,
       COUNT(*) AS total_students,
       ROUND(AVG(score), 2) AS avg_score
FROM student
GROUP BY class
"""
pd.read_sql_query(query, conn)
```

#### 🔹Nhận xét Bài 2.2a: Thống kê theo lớp

- Cho biết số lượng sinh viên từng lớp.
- Tính điểm trung bình của từng lớp.
- 👉 Lớp có điểm cao → học tốt, lớp điểm thấp → cần cải thiện.

### 2b.Tổng số sinh viên & điểm TB theo từng môn

```python
query = """
SELECT c.course_name,
       COUNT(s.student_id) AS total_students,
       ROUND(AVG(s.score), 2) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
pd.read_sql_query(query, conn)
```

#### 🔹Nhận xét Bài 2.2b: Thống kê theo môn học

- Cho biết số sinh viên mỗi môn.
- Tính điểm trung bình từng môn.
- 👉 Giúp đánh giá độ khó và mức độ quan tâm của môn học.

### 2c.Phân loại theo điểm trung bình môn

```python
query = """
SELECT c.course_name,
       ROUND(AVG(s.score), 2) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9.0 THEN 'Xuat sac'
           WHEN AVG(s.score) BETWEEN 6.0 AND 8.9 THEN 'Tot'
           ELSE 'Kem'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
pd.read_sql_query(query, conn)
```

#### 🔹Nhận xét Bài 2.2c: Phân loại theo điểm

- ≥ 9.0 → Xuất sắc
- 6.0 – 8.9 → Tốt
- < 6.0 → Kém

- 👉 Giúp đánh giá tổng quan chất lượng học tập.

#### 🔹 Kết luận Bài 2

- Dữ liệu sau khi làm sạch có độ tin cậy cao hơn.
- Các thống kê giúp hiểu rõ:
  - Chất lượng học tập
  - Sự phân bố sinh viên
- 👉 Đây là bước chuyển từ dữ liệu thô → thông tin có giá trị.

### 3a.Xếp hạng theo điểm toàn bộ sinh viên

```python
query = """
SELECT student_id, name, class, course_id, score,
       RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student
"""
pd.read_sql_query(query, conn)
```

#### ==> Top 3 sinh viên đạt điểm cao nhất

```python
query = """
SELECT *
FROM (
    SELECT student_id, name, class, course_id, score,
           RANK() OVER (ORDER BY score DESC) AS rnk
    FROM student
)
WHERE rnk <= 3
"""
pd.read_sql_query(query, conn)
```

#### ==> Top 3 sinh viên đạt điểm thấp nhất

```python
query = """
SELECT *
FROM (
    SELECT student_id, name, class, course_id, score,
           RANK() OVER (ORDER BY score ASC) AS rnk
    FROM student
)
WHERE rnk <= 3
"""
pd.read_sql_query(query, conn)
```
