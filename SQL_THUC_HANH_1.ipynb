TẠO BẢNG COURSE

```python
course = pd.DataFrame({
    "id": [12,34,26],
    "course_name": ["Giai tich","Thong ke","Tin hoc"]
})

course
```

BÀI 1

TÍCH DECARTES

```python
cross_join = student.merge(course, how="cross")
cross_join
```

NHẬN XÉT:

Phép nối này tạo ra tất cả các tổ hợp có thể giữa hai bảng, nghĩa là mỗi sinh viên sẽ được ghép với tất cả các môn học. Điều này làm cho số lượng bản ghi tăng rất lớn (bằng tích số dòng của hai bảng).

INNER JOIN

```python
inner_join = pd.merge(
    student, course,
    left_on="course_id",
    right_on="id",
    how="inner"
)
inner_join
```

NHẬN XÉT:

Chỉ giữ lại các bản ghi có sự khớp giữa course_id và id. Điều này đảm bảo rằng mọi dữ liệu sau khi nối đều hợp lệ.

→ Tuy nhiên, các sinh viên chưa có môn học hoặc có course_id không hợp lệ sẽ bị loại bỏ, dẫn đến mất dữ liệu.

LEFT JOIN

```python
left_join = pd.merge(
    student, course,
    left_on="course_id",
    right_on="id",
    how="left"
)
left_join
```

NHẬN XÉT:

Giữ toàn bộ dữ liệu từ bảng student, kể cả khi không có thông tin môn học tương ứng.

→ Đây là phép nối phù hợp khi mục tiêu chính là phân tích sinh viên, vì không làm mất bất kỳ sinh viên nào.

RIGHT JOIN

```python
right_join = pd.merge(
    student, course,
    left_on="course_id",
    right_on="id",
    how="right"
)
right_join
```

NHẬN XÉT:

Giữ toàn bộ dữ liệu từ bảng course, kể cả khi không có sinh viên đăng ký.

→ Giúp phát hiện các môn học chưa có sinh viên tham gia

FULL OUTER JOIN

```python
outer_join = pd.merge(
    student, course,
    left_on="course_id",
    right_on="id",
    how="outer"
)
outer_join
```

NHẬN XÉT:

Kết hợp toàn bộ dữ liệu của cả hai bảng, bao gồm cả các bản ghi không khớp.

→ Đây là cách tổng quát nhất nhưng tạo ra nhiều giá trị NULL

TẠO DATABASE VÀ BẢNG

```python
import sqlite3
import pandas as pd

# Tạo kết nối
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Tạo bảng student
cursor.execute("""
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

# Tạo bảng course
cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")
```

INSERT DỮ LIỆU

```python
# Insert student
cursor.executemany("""
INSERT INTO student VALUES (?, ?, ?, ?, ?)
""", [
(1,"Nguyen Minh Hoang","May Tinh",12,6.7),
(2,"Tran Thi Lan","Kinh Te",34,9.2),
(3,"Pham Van Nam","Toan Tin",None,7.9),
(4,"Le Thanh Huyen","Toan Tin",20,7.2),
(5,"Vu Quoc Anh","May Tinh",24,8.0),
(6,"Dang Thuy Linh","May Tinh",24,5.5),
(7,"Bui Tien Dung","Kinh Te",34,9.2),
(8,"Ho Ngoc Mai","Toan Tin",20,8.8),
(9,"Duong Huu Phuc","Kinh Te",None,7.2),
(10,"Cao Thi Hanh","May Tinh",None,7.0)
])

# Insert course
cursor.executemany("""
INSERT INTO course VALUES (?, ?)
""", [
(12,"Giai tich"),
(34,"Thong ke"),
(26,"Tin hoc")
])

conn.commit()
```

CẬP NHẬT COURSE_ID CÒN THIẾU

```python
import random

# Lấy danh sách course_id hợp lệ
course_ids = [12, 34, 26]

# Cập nhật từng dòng NULL
cursor.execute("SELECT student_id FROM student WHERE course_id IS NULL")
rows = cursor.fetchall()

for row in rows:
    random_id = random.choice(course_ids)
    cursor.execute("""
        UPDATE student
        SET course_id = ?
        WHERE student_id = ?
    """, (random_id, row[0]))

conn.commit()
```

XÓA CÁC BẢN GHI CÓ COURSE KO TỒN TẠI

```python
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")

conn.commit()
```

TỔNG SỐ SV VÀ ĐIỂM TB THEO LỚP

```python
query_a = """
SELECT class,
       COUNT(*) AS tong_sinh_vien,
       ROUND(AVG(score), 2) AS diem_trung_binh
FROM student
GROUP BY class
"""
pd.read_sql_query(query_a, conn)
```

NHẬN XÉT:

Kết quả cho thấy số lượng sinh viên phân bố không đồng đều giữa các lớp, gồm 6 sinh viên trong đó lớp Kinh Tế nhiều nhất với 3 sinh viên, lớp Máy Tính có 2 và ít nhất là lớp Toán Tin với 1 sinh viên

Điểm trung bình của từng lớp phản ánh chất lượng học tập:

Lớp có điểm trung bình cao nhất là lớp Kinh Tế với 8,53 điểm

Lớp có điểm trung bình thấp nhất là lớp Máy Tính với 6,85 điểm

TỔNG SỐ SV VÀ ĐIỂM TB THEO MÔN HỌC

```python
query_b = """
SELECT c.course_name,
       COUNT(*) AS tong_sinh_vien,
       ROUND(AVG(s.score), 2) AS diem_trung_binh
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
pd.read_sql_query(query_b, conn)
```

NHẬN XÉT:

Môn học có điểm trung bình cao nhất là môn Thống kê

Môn học có điểm trung bình thấp nhất là môn Tin học

PHÂN LOẠI THI ĐUA THEO MÔN HỌC

```python
query_c = """
SELECT c.course_name,
       ROUND(AVG(s.score), 2) AS diem_trung_binh,
       CASE
           WHEN AVG(s.score) >= 9.0 THEN 'Xuat sac'
           WHEN AVG(s.score) BETWEEN 6.0 AND 8.9 THEN 'Tot'
           ELSE 'Kem'
       END AS xep_loai
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
pd.read_sql_query(query_c, conn)
```

NHẬN XÉT:

Môn Giải tích có điểm trung bình là 7,3 và xếp loại Tốt

Môn Thống kê có điểm trung bình là 8,53 và xếp loại Tốt

Môn Tin học có điểm trung bình là 7 và xếp loại Tốt

XẾP HẠNG THEO TOÀN BỘ ĐIỂM SV

```python
query_rank_all = """
SELECT *,
       RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student
"""
pd.read_sql_query(query_rank_all, conn)
```

NHẬN XÉT:

Có tất cả 7 sinh viên, trong đó:

Sinh viên Trần Thị Lan và Bùi Tiến Dũng đều có điểm trung bình là 9,2 và xếp thứ nhất

Sinh viên Nguyễn Minh Hoàng có điểm trung bình thấp nhất là 6,7 và xếp cuối cùng



TOP 3 CAO NHẤT

```python
query_top3_high = """
SELECT *
FROM (
    SELECT *,
           RANK() OVER (ORDER BY score DESC) AS rank_score
    FROM student
)
WHERE rank_score <= 3
"""
pd.read_sql_query(query_top3_high, conn)
```

NHẬN XÉT:

Top 3 sinh viên có điểm cao nhất gồm Trần Thị Lan, Bùi Tiến Dũng và Phạm Văn Nam

TOP 3 THẤP NHẤT

```python
query_top3_low = """
SELECT *
FROM (
    SELECT *,
           RANK() OVER (ORDER BY score ASC) AS rank_score
    FROM student
)
WHERE rank_score <= 3
"""
pd.read_sql_query(query_top3_low, conn)
```

NHẬN XÉT:

Top 3 sinh viên coa điểm thấp nhất là Nguyễn Minh Hoàng, Cao Thị Hạnh, Dương Hữu Phúc

THÊM CỘT GRADUATION_DATE

```python
cursor.execute("""
ALTER TABLE student
ADD COLUMN graduation_date DATETIME
""")
conn.commit()
```

CẬP NHẬT GRADUATION_DATE THEO RANK

```python
cursor.execute("""
UPDATE student
SET graduation_date = DATETIME(
    'now',
    '+' || (
        SELECT rank_score
        FROM (
            SELECT student_id,
                   RANK() OVER (ORDER BY score DESC) AS rank_score
            FROM student
        ) r
        WHERE r.student_id = student.student_id
    ) || ' days'
)
""")
conn.commit()
```

XEM KẾT QUẢ

```python
df = pd.read_sql_query("""
SELECT student_id, name, score, graduation_date
FROM student
ORDER BY score DESC
""", conn)

df
```
