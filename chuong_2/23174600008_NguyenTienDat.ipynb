```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

def run_query(query, description=""):
    if description:
        print(f"{description}")
    try:
        display(pd.read_sql_query(query, conn))
    except Exception as e:
        print("Lỗi hoặc không có dữ liệu trả về:", e)
    print("\n")

# 1. Tạo bảng 
cursor.execute('''
    CREATE TABLE course (
        id INTEGER PRIMARY KEY,
        course_name TEXT
    )
''')

# 2. Tạo bảng student
cursor.execute('''
    CREATE TABLE student (
        student_id INTEGER PRIMARY KEY,
        name TEXT,
        class TEXT,
        course_id INTEGER,
        score REAL
    )
''')

# 3. Chèn dữ liệu vào bảng course 
course_data = [
    (12, 'Giai tich'),
    (34, 'Thong ke'),
    (26, 'Tin hoc')
]
cursor.executemany('INSERT INTO course VALUES (?, ?)', course_data)

# 4. Chèn dữ liệu vào bảng student 
student_data = [
    (1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
    (2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
    (3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
    (4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
    (5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
    (6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
    (7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
    (8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
    (9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
    (10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
]
cursor.executemany('INSERT INTO student VALUES (?, ?, ?, ?, ?)', student_data)
conn.commit()

run_query("SELECT * FROM student", "Dữ liệu bảng Student")
run_query("SELECT * FROM course", "Dữ liệu bảng Course")
```

```python
# Tích Decartes 
run_query('''
SELECT * FROM student CROSS JOIN course;
''', "Tích Decartes")

# INNER JOIN 
run_query('''
SELECT * FROM student 
INNER JOIN course ON student.course_id = course.id;
''', "INNER JOIN")

# LEFT JOIN 
run_query('''
SELECT * FROM student 
LEFT JOIN course ON student.course_id = course.id;
''', "LEFT JOIN")

# RIGHT JOIN 
run_query('''
SELECT * FROM student 
RIGHT JOIN course ON student.course_id = course.id;
''', "RIGHT JOIN")

# FULL OUTER JOIN 
run_query('''
SELECT * FROM student 
FULL OUTER JOIN course ON student.course_id = course.id;
''', "FULL OUTER JOIN")
```

```python
# Cập nhật course_id còn thiếu bằng một giá trị có trong bảng course 
cursor.execute('''
UPDATE student 
SET course_id = 26 
WHERE course_id IS NULL;
''')

# Loại bỏ các bản ghi có course_id không tồn tại 
cursor.execute('''
DELETE FROM student 
WHERE course_id NOT IN (SELECT id FROM course);
''')
conn.commit()

run_query("SELECT * FROM student", "Bảng student sau khi Cập nhật và Xóa")

# a. Tổng số sinh viên, điểm trung bình của từng lớp 
run_query('''
SELECT class, COUNT(student_id) as tong_sv, AVG(score) as diem_tb 
FROM student 
GROUP BY class;
''', "2a. Thống kê theo lớp")

# b. Tổng số sinh viên, điểm trung bình của từng môn học 
run_query('''
SELECT c.course_name, COUNT(s.student_id) as tong_sv, AVG(s.score) as diem_tb 
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name;
''', "2b. Thống kê theo môn học")

# c. Phân loại thi đua theo điểm số môn học 
run_query('''
SELECT name, score, c.course_name,
    CASE 
        WHEN score >= 9.0 THEN 'Xuất sắc'
        WHEN score >= 6.0 AND score <= 8.9 THEN 'Tốt'
        WHEN score < 6.0 THEN 'Kém'
    END as xep_loai
FROM student s
JOIN course c ON s.course_id = c.id;
''', "2c. Phân loại thi đua (Xuất sắc, Tốt, Kém)")
```

```python
# a. Xếp hạng tổng thể theo Điểm số 
run_query('''
WITH Ranked AS (
    SELECT student_id, name, score, 
           RANK() OVER (ORDER BY score DESC) as rank_cao_nhat,
           RANK() OVER (ORDER BY score ASC) as rank_thap_nhat
    FROM student
)
SELECT * FROM Ranked WHERE rank_cao_nhat <= 3 OR rank_thap_nhat <= 3
ORDER BY score DESC;
''', "3a. Top 3 cao nhất và thấp nhất theo toàn bộ điểm số")

# b. Xếp hạng theo lớp học  
run_query('''
WITH RankedClass AS (
    SELECT student_id, name, class, score,
           RANK() OVER (PARTITION BY class ORDER BY score DESC) as rank_cao_nhat,
           RANK() OVER (PARTITION BY class ORDER BY score ASC) as rank_thap_nhat
    FROM student
)
SELECT * FROM RankedClass WHERE rank_cao_nhat <= 3 OR rank_thap_nhat <= 3
ORDER BY class, score DESC;
''', "3b. Top 3 cao nhất/thấp nhất trong từng Lớp")

# c. Xếp hạng theo môn học  
run_query('''
WITH RankedCourse AS (
    SELECT student_id, name, course_id, score,
           RANK() OVER (PARTITION BY course_id ORDER BY score DESC) as rank_cao_nhat,
           RANK() OVER (PARTITION BY course_id ORDER BY score ASC) as rank_thap_nhat
    FROM student
)
SELECT * FROM RankedCourse WHERE rank_cao_nhat <= 3 OR rank_thap_nhat <= 3
ORDER BY course_id, score DESC;
''', "3c. Top 3 cao nhất/thấp nhất trong từng Môn học")
```

```python
# Bổ sung trường graduation_date với kiểu DATETIME 
cursor.execute('''
ALTER TABLE student ADD COLUMN graduation_date DATETIME;
''')

# Cập nhật thời gian
cursor.execute('''
WITH RankData AS (
    SELECT student_id, RANK() OVER (ORDER BY score DESC) as rnk 
    FROM student
)
UPDATE student
SET graduation_date = datetime('now', '+' || (SELECT rnk FROM RankData WHERE RankData.student_id = student.student_id) || ' days');
''')
conn.commit()

run_query("SELECT student_id, name, score, graduation_date FROM student ORDER BY score DESC;", "4. Cập nhật ngày tốt nghiệp theo thứ hạng")

conn.close()
```
