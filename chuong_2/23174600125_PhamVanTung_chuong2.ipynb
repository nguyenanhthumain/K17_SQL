1. Hãy kết nối hai bảng trên theo những cách sau: 
- Sử dụng tích Decartes. 
- Sử dụng JOIN: INNER JOIN, LEFT JOIN, RIGHT JOIN, FULL OUTER JOIN.

```python
# Tích decartes
decartes = """
SELECT *
FROM student, course
"""
df_decartes = pd.read_sql_query(decartes, conn)
print(df_decartes)
```

```python
# inner join
query_inner = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM student s
INNER JOIN course c
    ON s.course_id = c.id
"""
df_inner = pd.read_sql_query(query_inner, conn)
print(df_inner)
```

```python
# left join
query_left = """
SELECT s.student_id, s.name, s.class, s.course_id,s.score, c.course_name
FROM student s
LEFT JOIN course c
    ON s.course_id = c.id
"""
df_left = pd.read_sql_query(query_left, conn)
print(df_left)
```

```python
# right join
query_right = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM course c
LEFT JOIN student s
    ON s.course_id = c.id
"""
df_right = pd.read_sql_query(query_right, conn)
print(df_right)
```

```python
# full join
query_full = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.id AS course_table_id, c.course_name
FROM student s
LEFT JOIN course c
    ON s.course_id = c.id

UNION

SELECT  s.student_id, s.name, s.class, s.course_id, s.score, c.id AS course_table_id, c.course_name
FROM course c
LEFT JOIN student s
    ON s.course_id = c.id
"""
df_full = pd.read_sql_query(query_full, conn)
print(df_full)
```

2. Hãy cập nhật những giá trị course_id còn thiếu trong bảng student bằng câu lệnh SQL, 
trong đó các giá trị được điền là những giá trị nằm trong bảng course  và loại bỏ những 
bản ghi tham gia những môn học không tồn tại bảng course. Sau đó hãy cho biết: 

a. Tổng số sinh viên, điểm trung bình của từng lớp. 

b. Tổng số sinh viên, điểm trung bình của từng môn học. 

c. Phân loại thi đua theo số điểm của từng môn học biết: 
- Điểm TB ≥  9.0: Xuất sắc. 
- 6.0 ≤ Điểm TB ≤ 8.9: Tốt. 
- Điểm TB < 6.0: Kém.

```python
# 1. Cập nhật course_id còn thiếu
cur.execute("""
UPDATE student
SET course_id = (SELECT MIN(id) FROM course)
WHERE course_id IS NULL
""")

# 2. Xóa những bản ghi có môn học không tồn tại
cur.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")
conn.commit()
```

```python
# a. Tổng số sinh viên, điểm trung bình từng lớp
query_a = """
SELECT  class,
    COUNT(*) AS tong_so_sinh_vien,
    ROUND(AVG(score), 2) AS diem_trung_binh
FROM student
GROUP BY class
"""
print("Tổng số sinh viên và điểm trung bình theo từng lớp: ")
print(pd.read_sql_query(query_a, conn))
```

```python
# b. Tổng số sinh viên, điểm trung bình của từng môn học
query_b = """
SELECT c.course_name,
    COUNT(s.student_id) AS tong_so_sinh_vien,
    ROUND(AVG(s.score), 2) AS diem_trung_binh
FROM course c
LEFT JOIN student s
    ON s.course_id = c.id
GROUP BY c.id, c.course_name
"""
print("\n Tổng số sinh viên, điểm trung bình của từng môn học: ")
print(pd.read_sql_query(query_b, conn))
```

```python
# c. Phân loại thi đua
query_c = """
SELECT c.course_name,
    COUNT(s.student_id) AS tong_so_sinh_vien,
    ROUND(AVG(s.score), 2) AS diem_trung_binh,
    CASE
        WHEN AVG(s.score) >= 9.0 THEN 'Xuất sắc'
        WHEN AVG(s.score) >= 6.0 AND AVG(s.score) <= 8.9 THEN 'Tốt'
        WHEN AVG(s.score) < 6.0 THEN 'Kém'
        ELSE 'Chưa có dữ liệu'
    END AS xep_loai
FROM course c
LEFT JOIN student s
    ON s.course_id = c.id
GROUP BY c.id, c.course_name
"""
print("\nPhân loại thi đua: ")
print(pd.read_sql_query(query_c, conn))
```

3. Hãy xếp hạng sinh viên thông qua: 

a. Điểm số. 

b. Điểm số theo lớp học. 

c. Điểm số theo mã môn học. 

và cho biết top 3 sinh viện đạt thứ hạng cao nhất, top 3 sinh viên đạt thứ hạng thấp nhất theo 
từng trường hợp trên. 

```python
# Xếp hạng điểm số
rank_diem_so = """
SELECT *
FROM (
    SELECT
        student_id,
        name,
        class,
        course_id,
        score,
        ROW_NUMBER() OVER (ORDER BY score DESC) AS rank_high
    FROM student
)
"""

df_diem_so = pd.read_sql_query(rank_diem_so, conn)
print(df_diem_so)
```

```python
query_a_top3_high = """
SELECT *
FROM (
    SELECT
        student_id,
        name,
        class,
        course_id,
        score,
        ROW_NUMBER() OVER (ORDER BY score DESC) AS rank_high
    FROM student
)
WHERE rank_high <= 3
"""

df_a_top3_high = pd.read_sql_query(query_a_top3_high, conn)
print(df_a_top3_high)
```

```python
# Xếp hạng điểm số theo lớp học
lop_hoc =(""" 
SELECT
    student_id,
    name,
    class,
    course_id,
    score,
    DENSE_RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_in_class_high
FROM student;
""")
df_lop_hoc = pd.read_sql_query(lop_hoc, conn)
print(df_lop_hoc)
```

```python
query_b_high = """
SELECT *
FROM (
    SELECT
        student_id, name, class, course_id, score,
        ROW_NUMBER() OVER (PARTITION BY class ORDER BY score DESC) AS ranking
    FROM student
)
WHERE ranking <= 3
ORDER BY class, ranking
"""
print("\nTop 3 cao nhất theo lớp:")
print(pd.read_sql_query(query_b_high, conn))
```

```python
# Xếp hạng điểm số theo mã môn học
ma_mon_hoc = """
SELECT *
FROM (
    SELECT
        student_id, name, class, course_id, score,
        DENSE_RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS ranking
    FROM student
)
WHERE ranking <= 3
ORDER BY course_id, ranking
"""
df_ma_mon_hoc = pd.read_sql_query(ma_mon_hoc, conn)
print(df_ma_mon_hoc)
```

```python
query_c_top3_high = """
SELECT *
FROM (
    SELECT
        student_id,
        name,
        class,
        course_id,
        score,
        RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_in_course_high
    FROM student
)
WHERE rank_in_course_high <= 3
ORDER BY course_id, rank_in_course_high, score DESC
"""

df_c_top3_high = pd.read_sql_query(query_c_top3_high, conn)
print(df_c_top3_high)
```

4. Hãy bổ sung thêm một trường graduation_date có kiểu dữ liệu là DATETIME vào bảng 
student để xác định thời gian tốt nghiệp của từng bạn, trong đó thời gian tốt nghiệp được 
xác định bởi thời gian hiện tại cộng với số hạng tương ứng của bạn đó tính theo điểm số.

```python
cur.execute("ALTER TABLE student ADD COLUMN graduation_date DATETIME")

cur.execute("""
UPDATE student
SET graduation_date = datetime('now', '+' || CAST(score AS INTEGER) || ' days')
""")

conn.commit()

print("\n--- STUDENT SAU KHI THÊM GRADUATION DATE ---")
print(pd.read_sql("SELECT * FROM student", conn))
```
