# nạp dữ liệu 

```python
import sqlite3
import pandas as pd

# 1. Tạo kết nối tới cơ sở dữ liệu (tạm thời trong RAM)
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# 2. Tạo bảng và nạp dữ liệu (Sử dụng dữ liệu từ ảnh của bạn)
cursor.executescript('''
CREATE TABLE course (
    id INT PRIMARY KEY,
    course_name TEXT
);

CREATE TABLE student (
    student_id INT PRIMARY KEY,
    name TEXT,
    class TEXT,
    course_id INT,
    score REAL
);

INSERT INTO course VALUES (12, 'Giai tich'), (34, 'Thong ke'), (26, 'Tin hoc');

INSERT INTO student VALUES 
(1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
(2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
(3, 'Pham Van Nam', 'Toan Tin', NULL, 7.9),
(4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
(5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
(6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
(7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
(8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
(9, 'Duong Huu Phuc', 'Kinh Te', NULL, 7.2),
(10, 'Cao Thi Hanh', 'May Tinh', NULL, 7.0);
''')

# 3. Hàm hỗ trợ chạy SQL và in kết quả ra màn hình
def run_sql(query):
    df = pd.read_sql_query(query, conn)
    print("\n--- Kết quả truy vấn ---")
    print(df)
    print("-" * 25)
# Xem nội dung bảng 1: Sinh viên
print("\n--- DỮ LIỆU BẢNG STUDENT (BẢNG 1) ---")
run_sql("SELECT * FROM student")

# Xem nội dung bảng 2: Học phần
print("\n--- DỮ LIỆU BẢNG COURSE (BẢNG 2) ---")
run_sql("SELECT * FROM course")

```

# bài 1

# 1a •	Sử dụng tích Descartes (Cartesian Product).

```python
# Đúng nè: Truyền câu lệnh vào hàm run_sql
query_descartes = '''
SELECT s.*, c.course_name 
FROM student s
CROSS JOIN course c
'''

run_sql(query_descartes)
```

# Ý 2a: INNER JOIN

```python
query_inner = '''
SELECT s.student_id, s.name, s.course_id, c.course_name, s.score
FROM student s
INNER JOIN course c ON s.course_id = c.id
'''
print("--- KẾT QUẢ INNER JOIN ---")
run_sql(query_inner)
```

# Ý 2b: LEFT JOIN

```python
query_left = '''
SELECT s.student_id, s.name, s.course_id, c.course_name, s.score
FROM student s
LEFT JOIN course c ON s.course_id = c.id
'''
print("--- KẾT QUẢ LEFT JOIN ---")
run_sql(query_left)
```

# Ý 2c: RIGHT JOIN

```python
query_right = '''
SELECT s.student_id, s.name, c.id AS course_id, c.course_name, s.score
FROM course c
LEFT JOIN student s ON s.course_id = c.id
'''
print("--- KẾT QUẢ RIGHT JOIN ---")
run_sql(query_right)
```

# ý 2d:  FULL OUTER JOIN

```python
# SQLite không hỗ trợ trực tiếp, ta dùng phép UNION giữa LEFT và RIGHT JOIN
query_full = '''
SELECT s.name, c.course_name FROM student s LEFT JOIN course c ON s.course_id = c.id
UNION
SELECT s.name, c.course_name FROM course c LEFT JOIN student s ON c.id = s.course_id
'''
print("\nKết quả FULL OUTER JOIN:")
display(run_sql(query_full))
```

# ý 2a và 2b

```python
# a. Tổng số sinh viên, điểm trung bình của từng lớp
query_2a = '''
SELECT 
    class AS [Lớp], 
    COUNT(*) AS [Tổng số SV], 
    ROUND(AVG(score), 2) AS [Điểm TB Lớp]
FROM student
GROUP BY class
'''
print("Thống kê theo Lớp:")
display(run_sql(query_2a))

# b. Tổng số sinh viên, điểm trung bình của từng môn học
query_2b = '''
SELECT 
    c.course_name AS [Tên Môn], 
    COUNT(s.student_id) AS [Tổng số SV], 
    ROUND(AVG(s.score), 2) AS [Điểm TB Môn]
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
'''
print("\nThống kê theo Môn học:")
display(run_sql(query_2b))
```

# ý 2c

```python
# c. Phân loại thi đua theo từng môn học
query_2c = '''
SELECT 
    c.course_name AS [Tên Môn], 
    ROUND(AVG(s.score), 2) AS [Điểm TB],
    CASE 
        WHEN AVG(s.score) >= 9.0 THEN 'Xuất sắc'
        WHEN AVG(s.score) >= 6.0 THEN 'Tốt'
        ELSE 'Kém'
    END AS [Phân loại]
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
'''
print("Kết quả phân loại thi đua:")
display(run_sql(query_2c))
```

# ý 3a

```python
# 3a. Xếp hạng tổng thể và lấy Top 3 / Bottom 3
query_3a = '''
WITH RankedStudents AS (
    SELECT 
        name, class, score,
        RANK() OVER (ORDER BY score DESC) AS rank_top,
        RANK() OVER (ORDER BY score ASC) AS rank_bottom
    FROM student
)
-- Lấy Top 3 cao nhất
SELECT name, class, score, 'Top 3 Cao Nhất' AS [Phân loại], rank_top AS [Thứ hạng]
FROM RankedStudents WHERE rank_top <= 3

UNION ALL

-- Lấy Top 3 thấp nhất
SELECT name, class, score, 'Top 3 Thấp Nhất' AS [Phân loại], rank_bottom AS [Thứ hạng]
FROM RankedStudents WHERE rank_bottom <= 3

ORDER BY [Phân loại] ASC, [Thứ hạng] ASC;
'''
print("--- XẾP HẠNG TỔNG THỂ (Ý 3a) ---")
display(run_sql(query_3a))

# KẾT LUẬN 3a: 
# Lan và Dũng cùng đạt 9.2 điểm nên đồng hạng 1 (Top). Nam hạng 3.
# Ngược lại ở nhóm Bottom, Hoàng thấp nhất (6.7) ở hạng 1, tiếp đến là Hạnh và Phúc.
```

# ý 3b

```python
# 3b. Xếp hạng trong nội bộ từng Lớp (PARTITION BY class)
query_3b = '''
WITH RankedByClass AS (
    SELECT 
        name, class, score,
        RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_top,
        RANK() OVER (PARTITION BY class ORDER BY score ASC) AS rank_bottom
    FROM student
)
SELECT name, class, score, 'Top 3 Cao Nhất' AS [Phân loại], rank_top AS [Thứ hạng]
FROM RankedByClass WHERE rank_top <= 3

UNION ALL

SELECT name, class, score, 'Top 3 Thấp Nhất' AS [Phân loại], rank_bottom AS [Thứ hạng]
FROM RankedByClass WHERE rank_bottom <= 3

ORDER BY class, [Phân loại] ASC, [Thứ hạng] ASC;
'''
print("--- XẾP HẠNG THEO LỚP (Ý 3b) ---")
display(run_sql(query_3b))

# KẾT LUẬN 3b: 
# Bảng xếp hạng giờ đây được chia ranh giới theo 'Kinh Te', 'May Tinh', 'Toan Tin'.
# Tại mỗi lớp, sinh viên được lấy ra top 3 người giỏi nhất và top 3 người thấp nhất nội bộ lớp đó.
```

# ý 3c

```python
# 3c. Xếp hạng trong nội bộ từng Môn học (PARTITION BY course_id)
query_3c = '''
WITH RankedByCourse AS (
    SELECT 
        s.name, c.course_name, s.score,
        RANK() OVER (PARTITION BY s.course_id ORDER BY s.score DESC) AS rank_top,
        RANK() OVER (PARTITION BY s.course_id ORDER BY s.score ASC) AS rank_bottom
    FROM student s
    JOIN course c ON s.course_id = c.id
)
SELECT name, course_name AS [Tên Môn], score, 'Top 3 Cao Nhất' AS [Phân loại], rank_top AS [Thứ hạng]
FROM RankedByCourse WHERE rank_top <= 3

UNION ALL

SELECT name, course_name AS [Tên Môn], score, 'Top 3 Thấp Nhất' AS [Phân loại], rank_bottom AS [Thứ hạng]
FROM RankedByCourse WHERE rank_bottom <= 3

ORDER BY [Tên Môn], [Phân loại] ASC, [Thứ hạng] ASC;
'''
print("--- XẾP HẠNG THEO MÔN HỌC (Ý 3c) ---")
display(run_sql(query_3c))

# KẾT LUẬN 3c: 
# Dữ liệu được nhóm theo Môn học (Thống kê, Giải tích). 
# Từng môn học sẽ tự tìm ra Top 3 điểm cao nhất và thấp nhất của riêng môn đó.
```
