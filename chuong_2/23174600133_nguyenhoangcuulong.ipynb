```
import sqlite3
import pandas as pd

# 1. Khởi tạo kết nối SQLite (tạo DB trong bộ nhớ tạm để thực hành)
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# 2. Tạo bảng 'course'
cursor.execute('''
CREATE TABLE course (
    id INTEGER PRIMARY KEY,
    course_name TEXT
)
''')

# 3. Tạo bảng 'student'
cursor.execute('''
CREATE TABLE student (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL,
    FOREIGN KEY (course_id) REFERENCES course (id)
)
''')

# 4. Chèn dữ liệu vào bảng 'course'
courses = [
    (12, 'Giai tich'),
    (34, 'Thong ke'),
    (26, 'Tin hoc')
]
cursor.executemany('INSERT INTO course VALUES (?, ?)', courses)

# 5. Chèn dữ liệu vào bảng 'student'
# Lưu ý: Một số học sinh không có course_id (để None/NULL)
students = [
    (1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
    (2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
    (3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
    (4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2), # ID 20 không có trong bảng course
    (5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),   # ID 24 không có trong bảng course
    (6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
    (7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
    (8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
    (9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
    (10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
]
cursor.executemany('INSERT INTO student VALUES (?, ?, ?, ?, ?)', students)

print(students)
print(courses)
```

```
query_descartes = """
SELECT
    s.student_id,
    s.name,
    c.course_name
FROM student s
CROSS JOIN course c
"""

df_descartes = pd.read_sql_query(query_descartes, conn)
print(df_descartes)
```

```
inner_join = """
SELECT
    s.student_id,
    s.name,
    c.course_name
FROM student s
inner JOIN course c ON s.course_id =  c.id
"""

df_inner_join = pd.read_sql_query(inner_join, conn)
print(df_inner_join)
```

```
left_join = """
SELECT
    s.student_id,
    s.name,
    c.course_name
FROM student s
left join course c ON s.course_id =  c.id
"""

df_left_join = pd.read_sql_query(left_join, conn)
print(df_left_join)
```

```
right_join = """
SELECT
    s.student_id,
    s.name,
    c.course_name
FROM course c
LEFT JOIN student s ON c.id = s.course_id
"""

df_right_join = pd.read_sql_query(right_join, conn)
print(df_right_join)
```

```
full_outer_join = """
SELECT s.name, c.course_name FROM student s LEFT JOIN course c ON s.course_id = c.id
    UNION
    SELECT s.name, c.course_name FROM course c LEFT JOIN student s ON c.id = s.course_id
"""

df_full_outer_join = pd.read_sql_query(full_outer_join, conn)
print(df_full_outer_join)
```

```
# 1. Xử lý dữ liệu
cursor.execute("UPDATE student SET course_id = 12 WHERE course_id IS NULL")
cursor.execute("DELETE FROM student WHERE course_id NOT IN (SELECT id FROM course)")
```

```
query_a ="""SELECT
    class AS Lớp,
    COUNT(student_id) AS Tổng_Số_SV,
    ROUND(AVG(score), 2) AS Điểm_TB_Lớp
FROM student
GROUP BY class"""
print(pd.read_sql(query_a, conn))
```

```
query_b ="""SELECT
    c.course_name AS Môn_Học,
    COUNT(s.student_id) AS Tổng_Số_SV,
    ROUND(AVG(s.score), 2) AS Điểm_TB_Môn
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name;"""
print(pd.read_sql(query_b, conn))
```

```

query_c = """
SELECT c.course_name, ROUND(AVG(s.score),2) as Diem_TB,
CASE
    WHEN AVG(s.score) >= 9.0 THEN 'Xuat sac'
    WHEN AVG(s.score) >= 6.0 THEN 'Tot'
    ELSE 'Kem'
END as Phan_Loai
FROM student s JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
print(pd.read_sql(query_c, conn))
```

#câu 3

```
query_3a ="""SELECT
    student_id,
    name,
    class,
    course_id,
    score,
    RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student
"""
print(pd.read_sql(query_3a, conn))
```

```
query_3b ="""SELECT
    class,
    name,
    score,
    DENSE_RANK() OVER (PARTITION BY class ORDER BY score DESC) as 'Hạng trong lớp'
FROM student"""
print(pd.read_sql(query_3b, conn))
```

```
query_3c ="""SELECT
    course_id,
    name,
    score,
    DENSE_RANK() OVER (PARTITION BY course_id ORDER BY score DESC) as 'Hạng theo môn'
FROM student"""
print(pd.read_sql(query_3c, conn))
```

```
cursor.execute('''ALTER TABLE student
ADD COLUMN graduation_date DATETIME''')
```

```
cursor.execute("""WITH ranked AS (
    SELECT
        student_id,
        RANK() OVER (ORDER BY score DESC) AS rnk
    FROM student
)
UPDATE student
SET graduation_date = datetime('now', '+' || (
    SELECT rnk
    FROM ranked
    WHERE ranked.student_id = student.student_id
) || ' days');""")
```

```
query_4 ="""SELECT student_id, name, score, graduation_date
FROM student
ORDER BY score DESC"""
print(pd.read_sql(query_4, conn))
```
