TẠO 2 BẢNG ĐÃ CHO

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")

students = [
    (1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
    (2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
    (3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
    (4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
    (5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
    (6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
    (7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
    (8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
    (9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
    (10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
]

courses = [
    (12, 'Giai tich'),
    (34, 'Thong ke'),
    (26, 'Tin hoc')
]

cursor.executemany("INSERT INTO student VALUES (?, ?, ?, ?, ?)", students)
cursor.executemany("INSERT INTO course VALUES (?, ?)", courses)
conn.commit()
display(pd.read_sql_query("SELECT * FROM student", conn))
display(pd.read_sql_query("SELECT * FROM course", conn))
print("Đã tạo xong 2 bảng")
```

Sử dụng tích Dercartes

```python
query = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score,
       c.id AS course_table_id, c.course_name
FROM student s
CROSS JOIN course c
"""

df_cross = pd.read_sql_query(query, conn)
display(df_cross)

print("Tổng số dòng:", len(df_cross))
```

SỬ DỤNG JOIN:

```python
# INNER JOIN
query = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM student s
INNER JOIN course c
ON s.course_id = c.id
"""

df_inner = pd.read_sql_query(query, conn)
display(df_inner)
print("Tổng số dòng:", len(df_inner))
```

```python
# LEFT JOIN
query = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
"""

df_left = pd.read_sql_query(query, conn)
display(df_left)

print("Tổng số dòng:", len(df_left))
```

```python
# RIGHT JOIN
query = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""

df_right = pd.read_sql_query(query, conn)
display(df_right)

print("Tổng số dòng:", len(df_right))
```

```python
# FULL OUTER JOIN
query = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM student s
LEFT JOIN course c
ON s.course_id = c.id

UNION

SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""

df_full = pd.read_sql_query(query, conn)
display(df_full)
print("Tổng số dòng:", len(df_full))
```

2.a TỔNG SỐ SINH VIÊN, ĐIỂM TRUNG BÌNH CỦA TỪNG LỚP

```python
query = """
SELECT class,
       COUNT(*) AS tong_sinh_vien,
       ROUND(AVG(score), 2) AS diem_trung_binh
FROM student
GROUP BY class
"""

df_class = pd.read_sql_query(query, conn)
display(df_class)
print(f"Tổng số sinh viên và điểm trung bình theo lớp: {len(df_class)}")
```

2.b TỔNG SỐ SINH VIÊN, ĐIỂM TRUNG BÌNH TỪNG MÔN HỌC

```python
query = """
SELECT c.course_name,
       COUNT(s.student_id) AS tong_sinh_vien,
       ROUND(AVG(s.score), 2) AS diem_trung_binh
FROM student s
JOIN course c
ON s.course_id = c.id
GROUP BY c.course_name
"""

df_course = pd.read_sql_query(query, conn)
display(df_course)
print(f"Tổng số sinh viên và điểm trung bình theo môn học: {len(df_course)}")
```

2.c PHÂN LOẠI THI ĐUA THEO SỐ ĐIỂM CỦA TỪNG MÔN HỌC 

```python
query = """
SELECT c.course_name AS mon_hoc,
       ROUND(AVG(s.score), 2) AS diem_tb,
       CASE
           WHEN AVG(s.score) >= 9.0 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6.0 AND AVG(s.score) <= 8.9 THEN 'Tot'
           ELSE 'Kem'
       END AS xep_loai
FROM student s
JOIN course c
ON s.course_id = c.id
GROUP BY c.course_name
"""

df_thidua = pd.read_sql_query(query, conn)
display(df_thidua)
print(f"Tổng số môn học và xếp loại thi đua: {len(df_thidua)}")
```

3. XẾP HẠNG SINH VIÊN

```python
#Bảng xếp hạng toàn bộ sinh viên
query = """
SELECT student_id,
       name,
       class,
       course_id,
       score,
       RANK() OVER (ORDER BY score DESC) AS xep_hang
FROM student
"""

df_rank_score = pd.read_sql_query(query, conn)
display(df_rank_score)
```

```python
# Top 3 sinh viên có thứ hạng cao nhất theo điểm số
query = """
SELECT *
FROM (
    SELECT student_id,
           name,
           class,
           course_id,
           score,
           RANK() OVER (ORDER BY score DESC) AS xep_hang
    FROM student
)
WHERE xep_hang <= 3
"""

top3_high_score = pd.read_sql_query(query, conn)
display(top3_high_score)
```

```python
# Top 3 sinh viên có thứ hạng thấp nhất theo điểm số
query = """
SELECT *
FROM (
    SELECT student_id,
           name,
           class,
           course_id,
           score,
           RANK() OVER (ORDER BY score ASC) AS xep_hang_thap
    FROM student
)
WHERE xep_hang_thap <= 3
"""

top3_low_score = pd.read_sql_query(query, conn)
display(top3_low_score)
```

3.b ĐIỂM SỐ THEO LỚP HỌC

```python
# Bảng xếp hạng theo lớp
query = """
SELECT student_id,
       name,
       class,
       course_id,
       score,
       RANK() OVER (PARTITION BY class ORDER BY score DESC) AS xep_hang_trong_lop
FROM student
ORDER BY class, xep_hang_trong_lop
"""

df_rank_class = pd.read_sql_query(query, conn)
display(df_rank_class)
```

```python
# Top 3 sinh viên có thứ hạng cao nhất theo từng lớp
query = """
SELECT *
FROM (
    SELECT student_id,
           name,
           class,
           course_id,
           score,
           RANK() OVER (PARTITION BY class ORDER BY score DESC) AS xep_hang
    FROM student
)
WHERE xep_hang <= 3
ORDER BY class, xep_hang
"""

top3_high_class = pd.read_sql_query(query, conn)
display(top3_high_class)
```

```python
# Top 3 sinh viên có thứ hạng thấp nhất theo từng lớp
query = """
SELECT *
FROM (
    SELECT student_id,
           name,
           class,
           course_id,
           score,
           RANK() OVER (PARTITION BY class ORDER BY score ASC) AS xep_hang
    FROM student
)
WHERE xep_hang <= 3
ORDER BY class, xep_hang
"""

top3_low_class = pd.read_sql_query(query, conn)
display(top3_low_class)
```

4. BỔ SUNG TRƯỜNG graduation_date

```python
from datetime import datetime, timedelta
# Thêm graduation_date
cursor.execute("ALTER TABLE student ADD COLUMN graduation_date DATETIME")
current_time = datetime.now()
cursor.execute("SELECT student_id, score FROM student")
for student_id, score in cursor.fetchall():
    grad_date = current_time + timedelta(days=int(score * 30))
    cursor.execute("UPDATE student SET graduation_date = ? WHERE student_id = ?", (grad_date, student_id))
conn.commit()

cursor.execute("SELECT student_id, name, graduation_date FROM student")
df_graduation = pd.DataFrame(cursor.fetchall(), columns=["student_id", "name", "graduation_date"])
print(df_graduation)
```
