```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")
```

```python
students = [
(1,'Nguyen Minh Hoang','May Tinh',12,6.7),
(2,'Tran Thi Lan','Kinh Te',34,9.2),
(3,'Pham Van Nam','Toan Tin',None,7.9),
(4,'Le Thanh Huyen','Toan Tin',20,7.2),
(5,'Vu Quoc Anh','May Tinh',24,8.0),
(6,'Dang Thuy Linh','May Tinh',24,5.5),
(7,'Bui Tien Dung','Kinh Te',34,9.2),
(8,'Ho Ngoc Mai','Toan Tin',20,8.8),
(9,'Duong Huu Phuc','Kinh Te',None,7.2),
(10,'Cao Thi Hanh','May Tinh',None,7.0)
]

cursor.executemany("INSERT INTO student VALUES (?,?,?,?,?)", students)

courses = [
(12,'Giai tich'),
(34,'Thong ke'),
(26,'Tin hoc')
]

cursor.executemany("INSERT INTO course VALUES (?,?)", courses)

conn.commit()
```

# câu 1:

```python
query = """
SELECT *
FROM student
CROSS JOIN course
"""
df = pd.read_sql(query, conn)
df
```

```python
# inter join
query = """
SELECT s.*, c.course_name
FROM student s
INNER JOIN course c
ON s.course_id = c.id
"""
df = pd.read_sql(query, conn)
df
```

```python
# left join
query = """
SELECT s.*, c.course_name
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
"""
df = pd.read_sql(query, conn)
df
```

```python
# right join
query = """
SELECT s.*, c.course_name
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""
df = pd.read_sql(query, conn)
df
```

```python
# full outer join
query = """
SELECT s.*, c.course_name
FROM student s
LEFT JOIN course c ON s.course_id = c.id

UNION

SELECT s.*, c.course_name
FROM course c
LEFT JOIN student s ON s.course_id = c.id
"""
df = pd.read_sql(query, conn)
df
```

```python
# A
query = """
SELECT class,
       COUNT(*) AS total_students,
       ROUND(AVG(score),2) AS avg_score
FROM student
GROUP BY class
"""
df = pd.read_sql(query, conn)
df
```

```python
# B
query = """
SELECT c.course_name,
       COUNT(*) AS total_students,
       ROUND(AVG(s.score),2) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
df = pd.read_sql(query, conn)
df
```

```python
# C
query = """
SELECT c.course_name,
       ROUND(AVG(s.score),2) AS avg_score,
       CASE
           WHEN AVG(s.score) > 9 THEN 'Xuat sac'
           WHEN AVG(s.score) > 6 AND AVG(s.score) < 8.9 THEN 'Tot'
           ELSE 'Kem'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
df = pd.read_sql(query, conn)
df
```

```python
# A
query = """
SELECT *,
       RANK() OVER (ORDER BY score DESC) AS rank_high,
       RANK() OVER (ORDER BY score ASC) AS rank_low
FROM student
"""
df = pd.read_sql(query, conn)
df
```

```python
# top 3 cao nhất
query = """
SELECT *
FROM (
    SELECT *,
           RANK() OVER (ORDER BY score DESC) AS r
    FROM student
)
WHERE r <= 3
"""
pd.read_sql(query, conn)
```

```python
cursor.execute("""
ALTER TABLE student
ADD COLUMN graduation_date DATETIME
""")

conn.commit()
```

```python
query = """
WITH ranked AS (
    SELECT student_id,
           RANK() OVER (ORDER BY score DESC) AS r
    FROM student
)
UPDATE student
SET graduation_date = DATETIME('now', '+' || (
    SELECT r FROM ranked WHERE ranked.student_id = student.student_id
) || ' days')
"""
cursor.execute(query)
conn.commit()
```
