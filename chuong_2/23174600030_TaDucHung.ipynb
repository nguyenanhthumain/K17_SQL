# 1. Tạo dữ liệu

```python
import pandas as pd
import sqlite3
from datetime import datetime, timedelta

# Tạo bảng student
student_data = [
    (1, "Nguyen Minh Hoang", "May Tinh", 12, 6.7),
    (2, "Tran Thi Lan", "Kinh Te", 34, 9.2),
    (3, "Pham Van Nam", "Toan Tin", None, 7.9),
    (4, "Le Thanh Huyen", "Toan Tin", 20, 7.2),
    (5, "Vu Quoc Anh", "May Tinh", 24, 8.0),
    (6, "Dang Thuy Linh", "May Tinh", 24, 5.5),
    (7, "Bui Tien Dung", "Kinh Te", 34, 9.2),
    (8, "Ho Ngoc Mai", "Toan Tin", 20, 8.8),
    (9, "Duong Huu Phuc", "Kinh Te", None, 7.2),
    (10, "Cao Thi Hanh", "May Tinh", None, 7.0)
]

student_df = pd.DataFrame(student_data, 
    columns=["student_id", "name", "class", "course_id", "score"]
)

# Tạo bảng course
course_data = [
    (12, "Giai tich"),
    (34, "Thong ke"),
    (26, "Tin hoc")
]

course_df = pd.DataFrame(course_data, columns=["id", "course_name"])
```

# 2. Kết nối SQLite

```python
conn = sqlite3.connect(":memory:")

student_df.to_sql("student", conn, index=False, if_exists="replace")
course_df.to_sql("course", conn, index=False, if_exists="replace")
```

# 3. JOIN & Decartes

```python
query = "SELECT * FROM student CROSS JOIN course"
pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM student s
INNER JOIN course c ON s.course_id = c.id
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM student s
LEFT JOIN course c ON s.course_id = c.id
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM course c
LEFT JOIN student s ON s.course_id = c.id
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM student s
LEFT JOIN course c ON s.course_id = c.id
UNION
SELECT * FROM course c
LEFT JOIN student s ON s.course_id = c.id
"""
pd.read_sql(query, conn)
```

# 4. Xử lý dữ liệu thiếu

```python
query = """
UPDATE student
SET course_id = 12
WHERE course_id IS NULL
"""
conn.execute(query)
conn.commit()
```

```python
query = """
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
"""
conn.execute(query)
conn.commit()
```

# 5. Thống kê

```python
query = """
SELECT class, COUNT(*) AS total_students, AVG(score) AS avg_score
FROM student
GROUP BY class
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT c.course_name, COUNT(*) AS total_students, AVG(score) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT c.course_name,
       AVG(score) as avg_score,
       CASE 
           WHEN AVG(score) >= 9 THEN 'Xuat sac'
           WHEN AVG(score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END as rank
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
pd.read_sql(query, conn)
```

# 6. Xếp hạng

```python
query = """
SELECT *, RANK() OVER (ORDER BY score DESC) as rank
FROM student
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT *, RANK() OVER (PARTITION BY class ORDER BY score DESC) as rank
FROM student
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT *, RANK() OVER (PARTITION BY course_id ORDER BY score DESC) as rank
FROM student
"""
pd.read_sql(query, conn)
```

# 7. Top 3

```python
query = """
SELECT * FROM (
    SELECT *, RANK() OVER (ORDER BY score DESC) as rank
    FROM student
)
WHERE rank <= 3
"""
pd.read_sql(query, conn)
```

```python
query = """
SELECT * FROM (
    SELECT *, RANK() OVER (ORDER BY score ASC) as rank
    FROM student
)
WHERE rank <= 3
"""
pd.read_sql(query, conn)
```

# 8. Thêm graduation_date

```python
query = "ALTER TABLE student ADD COLUMN graduation_date TEXT"
conn.execute(query)

# Cập nhật
students = pd.read_sql("SELECT * FROM student", conn)

current_time = datetime.now()

students["graduation_date"] = students["score"].apply(
    lambda x: current_time + timedelta(days=int(x))
)

students.to_sql("student", conn, index=False, if_exists="replace")

pd.read_sql("SELECT * FROM student", conn)
```
