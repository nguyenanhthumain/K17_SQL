# 0. Khởi tạo bảng

## Tạo bảng courses

```python
import sqlite3

# Kết nối tới database (tạo file mới nếu chưa có)
conn = sqlite3.connect("school.db")
cursor = conn.cursor()

# Tạo bảng course
cursor.execute("""
CREATE TABLE IF NOT EXISTS course (
    id INTEGER PRIMARY KEY,
    course_name TEXT NOT NULL
)
""")

# Thêm dữ liệu cho course
courses = [
    (12, "Giai tich"),
    (34, "Thong ke"),
    (26, "Tin hoc")
]
cursor.executemany("INSERT OR IGNORE INTO course VALUES (?, ?)", courses)

# Truy vấn dữ liệu từ bảng course
cursor.execute("SELECT * FROM course")
rows = cursor.fetchall()
print("\nBảng course:")
for row in rows:
    print(row)

conn.commit()
conn.close()
```

## Tạo bảng student

```python
import sqlite3

# Mở kết nối
conn = sqlite3.connect("school.db")
cursor = conn.cursor()

# Xóa bảng student nếu đã tồn tại để tránh dữ liệu trùng lặp
cursor.execute("DROP TABLE IF EXISTS student")

# Tạo bảng student
cursor.execute("""
CREATE TABLE student (
    student_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    class TEXT,
    course_id INTEGER,
    score REAL,
    FOREIGN KEY(course_id) REFERENCES course(id)
)
""")

# Thêm dữ liệu cho student
students = [
    ("Nguyen Minh Hoang", "May Tinh", 12, 6.7),
    ("Tran Thi Lan", "Kinh Te", 34, 9.2),
    ("Pham Van Nam", "Toan Tin", None, 7.9),
    ("Le Thanh Huyen", "Toan Tin", 20, 7.2),
    ("Vu Quoc Anh", "May Tinh", 24, 8.0),
    ("Dang Thuy Linh", "May Tinh", 24, 5.5),
    ("Bui Tien Dung", "Kinh Te", 34, 9.2),
    ("Ho Ngoc Mai", "Toan Tin", 20, 8.8),
    ("Duong Huu Phuc", "Kinh Te", None, 7.2),
    ("Cao Thi Hanh", "May Tinh", None, 7.0)
]
cursor.executemany("INSERT INTO student (name, class, course_id, score) VALUES (?, ?, ?, ?)", students)

# Truy vấn dữ liệu từ bảng student
cursor.execute("SELECT * FROM student")
rows = cursor.fetchall()
print("Bảng student:")
for row in rows:
    print(row)

conn.commit()
conn.close()
```

## Sử dụng join

```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

# INNER JOIN
print("\nINNER JOIN:")
cursor.execute("""
SELECT s.name, s.class, s.score, c.course_name
FROM student s
INNER JOIN course c ON s.course_id = c.id
""")
for row in cursor.fetchall():
    print(row)

# LEFT JOIN
print("\nLEFT JOIN:")
cursor.execute("""
SELECT s.name, s.class, s.score, c.course_name
FROM student s
LEFT JOIN course c ON s.course_id = c.id
""")
for row in cursor.fetchall():
    print(row)

# RIGHT JOIN (SQLite không hỗ trợ trực tiếp, mô phỏng bằng LEFT JOIN đảo ngược)
print("\nRIGHT JOIN (mô phỏng):")
cursor.execute("""
SELECT s.name, s.class, s.score, c.course_name
FROM course c
LEFT JOIN student s ON s.course_id = c.id
""")
for row in cursor.fetchall():
    print(row)

# FULL OUTER JOIN (SQLite không hỗ trợ trực tiếp, mô phỏng bằng UNION)
print("\nFULL OUTER JOIN (mô phỏng):")
cursor.execute("""
SELECT s.name, s.class, s.score, c.course_name
FROM student s
LEFT JOIN course c ON s.course_id = c.id
UNION
SELECT s.name, s.class, s.score, c.course_name
FROM course c
LEFT JOIN student s ON s.course_id = c.id
""")
for row in cursor.fetchall():
    print(row)

conn.close()
```

# 2. Cập nhật và loại bỏ giá trị


```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

# 1. Xóa các bản ghi có course_id không tồn tại trong bảng course
cursor.execute("""
DELETE FROM student
WHERE course_id IS NOT NULL
  AND course_id NOT IN (SELECT id FROM course)
""")

# 2. Cập nhật course_id còn thiếu theo class
cursor.execute("""
UPDATE student
SET course_id = 26
WHERE course_id IS NULL AND class = 'May Tinh'
""")

cursor.execute("""
UPDATE student
SET course_id = 34
WHERE course_id IS NULL AND class = 'Kinh Te'
""")

cursor.execute("""
UPDATE student
SET course_id = 12
WHERE course_id IS NULL AND class = 'Toan Tin'
""")

conn.commit()

# Kiểm tra lại dữ liệu sau khi cập nhật
cursor.execute("SELECT * FROM student")
rows = cursor.fetchall()
print("Dữ liệu student sau khi cập nhật:")
for row in rows:
    print(row)

conn.close()
```

## a) Tổng số sinh viên, điểm trung bình của từng lớp

```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

print("Tổng số sinh viên và điểm trung bình theo từng lớp:")
cursor.execute("""
SELECT class, COUNT(*) AS total_students, AVG(score) AS avg_score
FROM student
GROUP BY class
""")
for row in cursor.fetchall():
    print(row)

conn.close()
```

## b) Tổng số sinh viên, điểm trung bình của từng môn học

```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

print("Tổng số sinh viên và điểm trung bình theo từng môn học:")
cursor.execute("""
SELECT c.course_name, COUNT(s.student_id) AS total_students, AVG(s.score) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""")
for row in cursor.fetchall():
    print(row)

conn.close()
```

## c) Phân loại thi đua theo số điểm của từng môn học

```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

print("Phân loại thi đua theo điểm trung bình từng môn học:")
cursor.execute("""
SELECT c.course_name,
       AVG(s.score) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9.0 THEN 'Xuất sắc'
           WHEN AVG(s.score) > 6.0 AND AVG(s.score) < 8.9 THEN 'Tốt'
           ELSE 'Kém'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""")
for row in cursor.fetchall():
    print(row)

conn.close()
```

# 3. Xếp hạng sinh viên thông qua các tiêu chí và đưa ra top 3 sinh viên có điểm số cao và top 3 sinh viên có điểm số thấp theo từng tiêu chí

## a) Điểm số

```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

print("Xếp hạng theo điểm số toàn bộ sinh viên:")

# Top 3 cao nhất
cursor.execute("""
SELECT name, class, score
FROM student
ORDER BY score DESC
LIMIT 3
""")
print("Top 3 cao nhất:")
for row in cursor.fetchall():
    print(row)

# Top 3 thấp nhất
cursor.execute("""
SELECT name, class, score
FROM student
ORDER BY score ASC
LIMIT 3
""")
print("\nTop 3 thấp nhất:")
for row in cursor.fetchall():
    print(row)

conn.close()
```

## b) Điểm số theo lớp học

```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

print("Xếp hạng theo điểm số trong từng lớp:")

# Top 3 cao nhất mỗi lớp
cursor.execute("""
SELECT class, name, score
FROM (
    SELECT class, name, score,
           ROW_NUMBER() OVER (PARTITION BY class ORDER BY score DESC) AS rank_high
    FROM student
)
WHERE rank_high <= 3
""")
print("Top 3 cao nhất theo lớp:")
for row in cursor.fetchall():
    print(row)

# Top 3 thấp nhất mỗi lớp
cursor.execute("""
SELECT class, name, score
FROM (
    SELECT class, name, score,
           ROW_NUMBER() OVER (PARTITION BY class ORDER BY score ASC) AS rank_low
    FROM student
)
WHERE rank_low <= 3
""")
print("\nTop 3 thấp nhất theo lớp:")
for row in cursor.fetchall():
    print(row)

conn.close()
```

## c) Điểm số theo mã môn học

```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

print("Xếp hạng theo điểm số trong từng môn học:")

# Top 3 cao nhất mỗi môn
cursor.execute("""
SELECT c.course_name, s.name, s.score
FROM (
    SELECT course_id, name, score,
           ROW_NUMBER() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_high
    FROM student
) s
JOIN course c ON s.course_id = c.id
WHERE rank_high <= 3
""")
print("Top 3 cao nhất theo môn:")
for row in cursor.fetchall():
    print(row)

# Top 3 thấp nhất mỗi môn
cursor.execute("""
SELECT c.course_name, s.name, s.score
FROM (
    SELECT course_id, name, score,
           ROW_NUMBER() OVER (PARTITION BY course_id ORDER BY score ASC) AS rank_low
    FROM student
) s
JOIN course c ON s.course_id = c.id
WHERE rank_low <= 3
""")
print("\nTop 3 thấp nhất theo môn:")
for row in cursor.fetchall():
    print(row)

conn.close()
```

# 4. Tạo thêm cột và tính thời gian tốt nghiệp

## Tạo thêm cột

```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

# Thêm cột graduation_date nếu chưa có
cursor.execute("ALTER TABLE student ADD COLUMN graduation_date DATETIME")

conn.commit()
conn.close()
```

## Tính thời gian tốt nghiệp theo hạng

```python
import sqlite3

conn = sqlite3.connect("school.db")
cursor = conn.cursor()

# Tính thứ hạng theo điểm số (cao nhất rank = 1) và cập nhật graduation_date
cursor.execute("""
WITH ranked AS (
    SELECT student_id,
           ROW_NUMBER() OVER (ORDER BY score DESC) AS rank_order
    FROM student
)
UPDATE student
SET graduation_date = datetime('now', '+' || ranked.rank_order || ' days')
FROM ranked
WHERE student.student_id = ranked.student_id
""")

conn.commit()

# Kiểm tra kết quả
cursor.execute("SELECT student_id, name, score, graduation_date FROM student ORDER BY score DESC")
rows = cursor.fetchall()
print("Danh sách sinh viên với ngày tốt nghiệp (điểm cao tốt nghiệp sớm):")
for row in rows:
    print(row)

conn.close()
```
