# 1. Kết nối 2 bảng theo từng cách:

```python
# Sử dụng tích Descartes
cursor.execute("""
SELECT students.name, students.class, students.course_id, course.course_name
FROM students, course
""")
print("Sử dụng tích Descartes")
for row in cursor.fetchall():
    print(row)
```

Kết nối 2 bảng bằng tích Descartes sẽ kết hợp tất cả sinh viên với tất cả môn học.

```python
# Sử dụng INNER JOIN
cursor.execute("""
SELECT students.name, students.class, course.course_name, students.score
FROM students
INNER JOIN course ON students.course_id = course.id
""")
print("Sử dụng INNER JOIN")
for row in cursor.fetchall():
    print(row)
```

Kết nối 2 bảng bằng INNER JOIN sẽ chỉ hiện sinh viên có môn học hợp lệ.

```python
# Sử dụng LEFT JOIN
cursor.execute("""
SELECT students.name, students.class, course.course_name, students.score
FROM students
LEFT JOIN course ON students.course_id = course.id
""")
print("Sử dụng LEFT JOIN")
for row in cursor.fetchall():
    print(row)
```

Kết nối 2 bảng bằng LEFT JOIN sẽ hiện tất cả sinh viên, kể cả khi chưa có môn học.

```python
# Sử dụng RIGHT JOIN
cursor.execute("""
SELECT students.name, students.class, course.course_name, students.score
FROM course
LEFT JOIN students ON students.course_id = course.id
""")
print("Sử dụng RIGHT JOIN")
for row in cursor.fetchall():
    print(row)
```

Kết nối 2 bảng bằng RIGHT JOIN sẽ hiện tất cả môn học, kể cả khi chưa có sinh viên.

```python
# Sử dụng FULL OUTER JOIN
cursor.execute("""
SELECT students.name, students.class, course.course_name, students.score
FROM students
LEFT JOIN course ON students.course_id = course.id
UNION
SELECT students.name, students.class, course.course_name, students.score
FROM course
LEFT JOIN students ON students.course_id = course.id
""")
print("Sử dụng FULL OUTER JOIN")
for row in cursor.fetchall():
    print(row)
```

Kết nối 2 bảng bằng FULL OUTER JOIN sẽ kết hợp cả 2 cách LEFT JOIN và RIGHT JOIN

a. Tổng số sinh viên, điểm trung bình của từng lớp

```python
cursor.execute(
"""
SELECT class,
  COUNT(*) AS tong_sv,
  ROUND(AVG(score),2) AS diem_tb
FROM students
GROUP BY class
""")
for row in cursor.fetchall():
    print(row)
```

Thấy được số lượng sinh viên của từng lớp và xếp hạng điểm trung bình của 3 lớp Kinh tế, Máy tính và Toán tin theo số lượng sinh viên của từng lớp

b. Tổng số sinh viên, điểm trung bình của từng môn học

```python
cursor.execute(
"""
SELECT c.course_name,
  COUNT(*) AS tong_sv,
  ROUND(AVG(s.score),2) AS diem_tb
FROM students s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
)
for row in cursor.fetchall():
    print(row)
```

Cho thấy được điểm trung bình của từng môn học , biết được môn nào điểm thấp, môn nào điểm cao

c. Phân loại thi đua theo số điểm của từng môn học biết

```python
cursor.execute(
    """
SELECT c.course_name,
       ROUND(AVG(s.score),2) AS diem_tb,
       CASE
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END AS xep_loai
FROM students s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
)
for row in cursor.fetchall():
    print(row)
```

Thấy được xếp hạng 3 môn đều ở xếp loại tốt bằng cách sử dụng WHEN-THEN

# 3. Xếp hạng sinh viên

a. Qua điểm số

```python
cursor.execute(
    """
SELECT *,
RANK() OVER (ORDER BY score DESC) AS rank
FROM students
    """
)
for row in cursor.fetchall():
    print(row)

# 3 sv cao điểm nhất
print("\nXếp hạng sinh viên cao điểm nhất: ")
cursor.execute(
    """
SELECT * FROM (
SELECT *,
RANK() OVER (ORDER BY score DESC) AS rank
FROM students)
WHERE rank <= 3
    """
)
for row in cursor.fetchall():
    print(row)

# 3 sv thấp điểm nhất
print("\nXếp hạng sinh viên thấp điểm nhất: ")
cursor.execute(
    """
SELECT * FROM (
SELECT *,
RANK() OVER (ORDER BY score ASC) AS rank
FROM students)
WHERE rank <= 3
    """
)
for row in cursor.fetchall():
    print(row)
```

Xếp hạng sinh viên cao điểm nhất và sinh viên tháp điểm nhát theo thứ tự 1 2 3 không phân biệt theo từng môn học

b. Qua điểm số theo lớp học

```python
cursor.execute(
    """
SELECT student_id, name, class, score,
       RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_in_class
FROM students
ORDER BY class, rank_in_class ASC
    """
)
for row in cursor.fetchall():
    print(row)

# 3 sv cao điểm nhất
print("\nXếp hạng sinh viên cao điểm nhất theo từng lớp học: ")
cursor.execute(
    """
SELECT student_id, name, class, score, rank_in_class FROM (
    SELECT student_id, name, class, score,
           RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_in_class
    FROM students
)
WHERE rank_in_class <= 3
ORDER BY class, rank_in_class ASC
    """
)
for row in cursor.fetchall():
    print(row)

# 3 sv thấp điểm nhất
print("\nXếp hạng sinh viên thấp điểm nhất theo từng lớp học: ")
cursor.execute(
    """
SELECT student_id, name, class, score, rank_in_class FROM (
    SELECT student_id, name, class, score,
           RANK() OVER (PARTITION BY class ORDER BY score ASC) AS rank_in_class
    FROM students
)
WHERE rank_in_class <= 3
ORDER BY class, rank_in_class ASC
    """
)
for row in cursor.fetchall():
    print(row)
```

Xếp hạng sinh viên cao điểm nhất và sinh viên tháp điểm nhát theo thứ tự 1 2 3 không phân biệt theo từng lớp học kinh tế, máy tính, toán tin từ đó so sánh với những lớp khác

c. Qua điểm số theo mã môn học.

```python
cursor.execute(
    """
SELECT student_id, name, class, course_id, score,
       RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_in_course
FROM students
ORDER BY course_id, rank_in_course ASC
    """
)
for row in cursor.fetchall():
    print(row)

# 3 sv cao điểm nhất
print("\nXếp hạng sinh viên cao điểm nhất theo từng môn học: ")
cursor.execute(
    """
SELECT student_id, name, class, course_id, score, rank_in_course FROM (
    SELECT student_id, name, class, course_id, score,
           RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_in_course
    FROM students
)
WHERE rank_in_course <= 3
ORDER BY course_id, rank_in_course ASC
    """
)
for row in cursor.fetchall():
    print(row)

# 3 sv thấp điểm nhất
print("\nXếp hạng sinh viên thấp điểm nhất theo từng môn học: ")
cursor.execute(
    """
SELECT student_id, name, class, course_id, score, rank_in_course FROM (
    SELECT student_id, name, class, course_id, score,
           RANK() OVER (PARTITION BY course_id ORDER BY score ASC) AS rank_in_course
    FROM students
)
WHERE rank_in_course <= 3
ORDER BY course_id, rank_in_course ASC
    """
)
for row in cursor.fetchall():
    print(row)
```

Xếp hạng sinh viên cao điểm nhất và sinh viên tháp điểm nhát theo thứ tự 1 2 3 không phân biệt theo từng môn học, từ đó đưa ra những đánh giá và nhận xét cho từng sinh viên

# 4. Bổ sung thêm một trường graduation_date có kiểu dữ liệu là DATETIME vào bảng student để xác định thời gian tốt nghiệp của từng bạn, trong đó thời gian tốt nghiệp được xác định bởi thời gian hiện tại cộng với số hạng tương ứng của bạn đó tính theo điểm số.

```python
from datetime import datetime

# cursor.execute("""
# ALTER TABLE students
# ADD COLUMN graduation_date DATE
# """)
# conn.commit()

# xếp hạng sinh viên
cursor.execute(
  """
SELECT *,
ROW_NUMBER() OVER (ORDER BY score DESC) AS rank_score
FROM students
  """
)

# Cập nhật thời gian theo số hạng
cursor.execute("""
UPDATE students
SET graduation_date = date(
    'now',
    '+' || (
        SELECT rank_score
        FROM (
            SELECT student_id,
                   ROW_NUMBER() OVER (ORDER BY score DESC) AS rank_score
            FROM students
        ) AS ranking
        WHERE ranking.student_id = students.student_id
    ) || ' months'
)
""")
conn.commit()


# in ra kq
cursor.execute(
    """
SELECT
    student_id,
    name,
    class,
    course_id,
    score,
    ROW_NUMBER() OVER (ORDER BY score DESC) AS rank_score,
    strftime('%d/%m/%Y', graduation_date) AS graduation_date
FROM students
ORDER BY score DESC
    """
)
for row in cursor.fetchall():
    print(row)
```

Từ bảng, thấy được thời gian tốt nghiệp của từng sinh viên qua điểm số và xếp hạng, điểm cao và xếp hạng cao sẽ được ra trường sớm hơn những người còn lại
