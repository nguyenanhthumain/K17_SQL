```python
# import thư viện
import sqlite3
import pandas as pd
from datetime import datetime, timedelta
```

```python
# Tạo database và bảng
conn = sqlite3.connect("student.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")
```

```python
query = "SELECT * FROM student CROSS JOIN course"
pd.read_sql(query, conn)
```

```python
# INNER JOIN
query = """
SELECT *
FROM student s
INNER JOIN course c
ON s.course_id = c.id
"""
pd.read_sql(query, conn)
```

```python
# LEFT JOIN
query = """
SELECT *
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
"""
pd.read_sql(query, conn)
```

```python
# RIGHT JOIN
query = """
SELECT *
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""
pd.read_sql(query, conn)
```

```python
# OUTER JOIN
query = """
SELECT *
FROM student s
LEFT JOIN course c ON s.course_id = c.id
UNION
SELECT *
FROM course c
LEFT JOIN student s ON s.course_id = c.id
"""
pd.read_sql(query, conn)
```

```python
# Update course_id bị thiếu
cursor.execute("""
UPDATE student
SET course_id = 26
WHERE course_id IS NULL
""")
conn.commit()

pd.read_sql("SELECT * FROM student", conn)
```

```python
# Xoá môn khong tồn tại
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")
conn.commit()
```

```python
# Tổng sinh viên và điểm trung bình theo lớp
query = """
SELECT class,
       COUNT(student_id) AS so_sinh_vien,
       AVG(score) AS diem_trung_binh
FROM student
GROUP BY class
"""
pd.read_sql(query, conn)
```

```python
# Tổng sinh viên và điểm trung bình theo môn
query = """
SELECT c.course_name,
       COUNT(s.student_id) AS so_sinh_vien,
       AVG(s.score) AS diem_trung_binh
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
pd.read_sql(query, conn)
```

```python
# Phân loại thi đua
query = """
SELECT c.course_name,
       AVG(s.score) AS diem_tb,
       CASE
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) BETWEEN 6 AND 8.9 THEN 'Tot'
           ELSE 'Kem'
       END AS xep_loai
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
pd.read_sql(query, conn)
```

```python
# Xếp hạng theo điểm
query = """
SELECT *,
       RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student
"""
pd.read_sql(query, conn)
```

```python
# Xếp hạng theo lớp
query = """
SELECT *,
       RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_class
FROM student
"""
pd.read_sql(query, conn)
```

```python
# Xếp hạng theo môn
query = """
SELECT *,
       RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_course
FROM student
"""
pd.read_sql(query, conn)
```

```python
# Top 3 cao nhất
query = """
SELECT *
FROM (
    SELECT *, RANK() OVER (ORDER BY score DESC) AS r
    FROM student
)
WHERE r <= 3
"""
pd.read_sql(query, conn)
```

```python
# Top 3 thấp nhất
query = """
SELECT *
FROM (
    SELECT *, RANK() OVER (ORDER BY score ASC) AS r
    FROM student
)
WHERE r <= 3
"""
pd.read_sql(query, conn)
```

```python
# Thêm gradution_dâte
cursor.execute("ALTER TABLE student ADD graduation_date TEXT")
conn.commit()
```

```python
# Cập nhật ngày tốt nghiệp theo hạng
query = """
SELECT student_id,
       RANK() OVER (ORDER BY score DESC) AS r
FROM student
"""
rank_df = pd.read_sql(query, conn)

for index, row in rank_df.iterrows():
    grad_date = datetime.now() + timedelta(days=int(row['r']))
    cursor.execute("""
    UPDATE student
    SET graduation_date = ?
    WHERE student_id = ?
    """, (grad_date.strftime("%Y-%m-%d"), int(row['student_id'])))

conn.commit()

pd.read_sql("SELECT * FROM student", conn)
```
