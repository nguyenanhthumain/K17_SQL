1. Hãy kết nối hai bảng trên theo những cách sau:

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
```

```python
cursor.execute('''
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
''')

cursor.execute('''
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
''')
```

```python
students = [
(1,'Nguyen Minh Hoang','May Tinh',12,6.7),
(2,'Tran Thi Lan','Kinh Te',34,9.2),
(3,'Pham Van Nam','Toan Tin',None,7.9),
(4,'Le Thanh Huyen','Toan Tin',20,7.2),
(5,'Vu Quoc Anh','May Tinh',24,8.0),
(6,'Dang Thuy Linh','May Tinh',24,5.5),
(7,'Bui Tien Dung','Kinh Te',34,9.2),
(8,'Ho Ngoc Mai','Toan Tin',20,8.8),
(9,'Duong Huu Phuc','Kinh Te',None,7.2),
(10,'Cao Thi Hanh','May Tinh',None,7.0)
]

courses = [
(12,'Giai tich'),
(34,'Thong ke'),
(26,'Tin hoc')
]

cursor.executemany("INSERT INTO student VALUES (?,?,?,?,?)", students)
cursor.executemany("INSERT INTO course VALUES (?,?)", courses)

conn.commit()
```

Tích Descartes (CROSS JOIN)

```python
df_cross = pd.read_sql_query("""
SELECT *
FROM student
CROSS JOIN course
""", conn)

df_cross
```

INNER JOIN

```python
df_inner = pd.read_sql_query("""
SELECT *
FROM student s
INNER JOIN course c
ON s.course_id = c.id
""", conn)

df_inner
```

LEFT JOIN

```python
df_left = pd.read_sql_query("""
SELECT *
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
""", conn)

df_left
```

RIGHT JOIN

```python
df_left = pd.read_sql_query("""
SELECT *
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
""", conn)

df_left
```

FULL OUTER JOIN

```python
df_full = pd.read_sql_query("""
SELECT *
FROM student s
LEFT JOIN course c ON s.course_id = c.id

UNION

SELECT *
FROM course c
LEFT JOIN student s ON s.course_id = c.id
""", conn)

df_full
```

2. XỬ LÝ DỮ LIỆU & THỐNG KÊ

```python
# Xóa course_id không tồn tại
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")

# Cập nhật NULL -> gán mặc định = 12
cursor.execute("""
UPDATE student
SET course_id = 12
WHERE course_id IS NULL
""")

conn.commit()

pd.read_sql_query("SELECT * FROM student", conn)
```

a. Tổng số sinh viên & điểm TB theo lớp

```python
df_class = pd.read_sql_query("""
SELECT class,
       COUNT(*) AS total_students,
       ROUND(AVG(score),2) AS avg_score
FROM student
GROUP BY class
""", conn)

df_class
```

b. Tổng số sinh viên & điểm TB theo môn

```python
df_course = pd.read_sql_query("""
SELECT c.course_name,
       COUNT(*) AS total_students,
       ROUND(AVG(s.score),2) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)

df_course
```

c. Phân loại thi đua

```python
df_rank = pd.read_sql_query("""
SELECT c.course_name,
       ROUND(AVG(s.score),2) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) BETWEEN 6 AND 8.9 THEN 'Tot'
           ELSE 'Kem'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)

df_rank
```

3. XẾP HẠNG SINH VIÊN

a. Xếp hạng theo điểm (toàn bộ)

```python
df_rank_all = pd.read_sql_query("""
SELECT *,
       RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student
""", conn)

df_rank_all
```

```python
top3_high = pd.read_sql_query("""
SELECT * FROM (
    SELECT *,
           RANK() OVER (ORDER BY score DESC) AS r
    FROM student
)
WHERE r <= 3
""", conn)

top3_high
```

```python
top3_low = pd.read_sql_query("""
SELECT * FROM (
    SELECT *,
           RANK() OVER (ORDER BY score ASC) AS r
    FROM student
)
WHERE r <= 3
""", conn)

top3_low
```

b. Xếp hạng theo lớp

```python
df_rank_class = pd.read_sql_query("""
SELECT *,
       RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_in_class
FROM student
""", conn)

df_rank_class
```

```python
top3_class_high = pd.read_sql_query("""
SELECT * FROM (
    SELECT *,
           RANK() OVER (PARTITION BY class ORDER BY score DESC) AS r
    FROM student
)
WHERE r <= 3
""", conn)

top3_class_high
```

```python
top3_class_low = pd.read_sql_query("""
SELECT * FROM (
    SELECT *,
           RANK() OVER (PARTITION BY class ORDER BY score ASC) AS r
    FROM student
)
WHERE r <= 3
""", conn)

top3_class_low
```

c. Xếp hạng theo môn học

```python
df_rank_course = pd.read_sql_query("""
SELECT *,
       RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_in_course
FROM student
""", conn)

df_rank_course
```

```python
top3_course_high = pd.read_sql_query("""
SELECT * FROM (
    SELECT *,
           RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS r
    FROM student
)
WHERE r <= 3
""", conn)

top3_course_high
```

```python
top3_course_low = pd.read_sql_query("""
SELECT * FROM (
    SELECT *,
           RANK() OVER (PARTITION BY course_id ORDER BY score ASC) AS r
    FROM student
)
WHERE r <= 3
""", conn)

top3_course_low
```

4. THÊM TRƯỜNG graduation_date

Thêm cột mới

```python
cursor.execute("""
ALTER TABLE student
ADD COLUMN graduation_date DATETIME
""")

conn.commit()
```

Cập nhật graduation_date

```python
cursor.execute("""
UPDATE student
SET graduation_date = datetime('now', '+' || (
    SELECT r FROM (
        SELECT student_id,
               RANK() OVER (ORDER BY score DESC) AS r
        FROM student
    ) t
    WHERE t.student_id = student.student_id
) || ' days')
""")

conn.commit()
```

Xem kết quả

```python
df_grad = pd.read_sql_query("""
SELECT student_id, name, score, graduation_date
FROM student
ORDER BY score DESC
""", conn)

df_grad
```
