kết nối 2 bảng sử dụng JOIN: INNER JOIN, LEFT JOIN, RIGHT JOIN, FULL OUTER JOIN

* INNER JOIN

```python
print("INNER JOIN:")
cursor.execute('''
SELECT * FROM student INNER JOIN course ON student.course_id = course.id;
''')
df_inner = pd.DataFrame(cursor.fetchall(), columns=['student_id', 'name', 'class', 'course_id', 'score', 'id', 'course_name'])
print(df_inner)
print("\n")
```

Chỉ những bản ghi có mã môn học trùng khớp giữa hai bảng mới được hiển thị.
Trong dữ liệu hiện tại chỉ có 3 sinh viên có course_id trùng với id trong bảng course, nên kết quả trả về 3 dòng.

* LEFT JOIN

```python
print("LEFT JOIN:")
cursor.execute('''
SELECT * FROM student LEFT JOIN course ON student.course_id = course.id;
''')
df_left = pd.DataFrame(cursor.fetchall(), columns=['student_id', 'name', 'class', 'course_id', 'score', 'id', 'course_name'])
print(df_left)
print("\n")
```

LEFT JOIN giữ toàn bộ dữ liệu của bảng bên trái (student).
Chỉ những bản ghi có mã môn học phù hợp mới hiển thị thông tin môn học từ bảng course.
Các bản ghi không khớp vẫn được giữ lại nhưng giá trị môn học sẽ là NULL/None.

* RIGHT JOIN

```python
print("RIGHT JOIN:")
cursor.execute('''
SELECT * FROM course LEFT JOIN student ON course.id = student.course_id;
''')
df_right = pd.DataFrame(cursor.fetchall(), columns=['id', 'course_name', 'student_id', 'name', 'class', 'course_id', 'score'])
print(df_right)
print("\n")
```

* FULL OUTER JOIN

```python
query = '''
SELECT * FROM student LEFT JOIN course
ON student.course_id = course.id
UNION ALL
SELECT * FROM course LEFT JOIN student
ON course.id = student.course_id
WHERE student.course_id IS NULL;
'''
df = pd.read_sql_query(query, conn)
print("FULL OUTER JOIN: ", df)
```

FULL OUTER JOIN giữ toàn bộ dữ liệu của cả hai bảng.
Các bản ghi khớp nhau sẽ được kết hợp lại.
Các bản ghi không khớp vẫn xuất hiện nhưng các cột còn lại sẽ có giá trị NULL/NaN.
Phép JOIN này giúp xem đầy đủ tất cả sinh viên và tất cả môn học, kể cả những dữ liệu chưa liên kết với nhau.

 bài 2. Hãy cập nhật những giá trị course_id còn thiếu trong bảng student bằng câu lệnh SQL,
trong đó các giá trị được điền là những giá trị nằm trong bảng course và loại bỏ những
bản ghi tham gia những môn học không tồn tại bảng course.

```python
# Cập nhật course_id còn thiếu bằng giá trị ngẫu nhiên từ bảng course
cursor.execute('''
UPDATE student
SET course_id = (
    SELECT id FROM course ORDER BY RANDOM() LIMIT 1
)
WHERE course_id IS NULL;
''')
conn.commit()

# Loại bỏ các bản ghi có course_id không hợp lệ
cursor.execute('''
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course);
''')
conn.commit()

# Kiểm tra lại dữ liệu
cursor.execute('SELECT * FROM student;')
print(pd.DataFrame(cursor.fetchall(), columns=['student_id', 'name', 'class', 'course_id', 'score']))
```

Bài toán yêu cầu cập nhật các giá trị course_id còn thiếu trong bảng student bằng một giá trị ngẫu nhiên từ bảng course, sau đó loại bỏ các bản ghi có course_id không tồn tại trong bảng course.

Kết quả thực hiện cho thấy các bản ghi thiếu course_id đã được cập nhật thành công, đồng thời các sinh viên đăng ký những môn học không hợp lệ đã được loại bỏ. Sau khi xử lý, bảng student chỉ còn các bản ghi có course_id hợp lệ, đảm bảo tính nhất quán và toàn vẹn dữ liệu giữa hai bảng.

```python
query = '''
SELECT class, COUNT(student_id) AS "tổng sinh viên", ROUND(AVG(score), 2) AS "điểm trung bình"
FROM student
GROUP BY class;
'''

df = pd.read_sql_query(query, conn)
print(df)
```

Câu truy vấn đã nhóm dữ liệu theo từng lớp để tính tổng số sinh viên và điểm trung bình của mỗi lớp. Kết quả cho thấy lớp Kinh Tế có số sinh viên nhiều nhất và cũng có điểm trung bình cao nhất. Lớp Máy Tính có điểm trung bình thấp hơn, trong khi lớp Toán Tin chỉ có một sinh viên nên điểm trung bình chính là điểm của sinh viên đó.

b, tổng số sinh viên, điểm trung bình từng môn học

```python
query='''
SELECT course.course_name,
COUNT(student.student_id) AS "tổng sinh viên",
ROUND(AVG(student.score),2) AS "điểm trung bình môn học"
FROM course
LEFT JOIN student ON student.course_id = course.id
GROUP BY course.course_name;
'''
df = pd.read_sql_query(query, conn)
print(df)
```

-Có 1 sinh viên tham gia.

Thống kê

-Có 2 sinh viên tham gia.

Tin học

-Có 3 sinh viên tham gia
điểm trung bình 7.37

c, phân loại thi đua theo số điểm
- i. Điểm TB ≥  9.0: Xuất sắc.
- ii. 6.0 ≤ Điểm TB ≤ 8.9: Tốt.
- iii. Điểm TB < 6.0: Kém.

```python
query = '''
SELECT course.course_name,
ROUND(AVG(student.score),2) AS "điểm trung bình",
CASE
    WHEN ROUND(AVG(student.score),2) >= 9.0 THEN "Xuất sắc"
    WHEN ROUND(AVG(student.score),2) BETWEEN 6.0 AND 8.9 THEN "Tốt"
    ELSE "Kém"
END AS "điểm"
FROM course
LEFT JOIN student ON student.course_id = course.id
GROUP BY course.course_name;
'''

df = pd.read_sql_query(query, conn)
print(df)
```

-Kết quả cho thấy môn Giải tích có điểm trung bình 6.7 và được xếp loại Tốt, môn Tin học có điểm trung bình 7.37 đạt loại Tốt . Điều này cho thấy sinh viên có kết quả học tập khá tốt ở hai môn này.
- đặc biệt, môn Thống kê đạt điêm trung bình 9.2 đạt loại Xuất sắc cho thấy sinh viên học môn này ở mức rất tốt.

cập nhật graduation_date dựa trên thứ hạng

```python
cursor.execute('''
WITH ranked_students AS (
    SELECT student_id, RANK() OVER (ORDER BY score DESC) AS rank
    FROM student
)
UPDATE student
SET graduation_date = DATETIME('now', '+' || (
    SELECT rank FROM ranked_students WHERE ranked_students.student_id = student.student_id
) || ' months');
''')
conn.commit()
```
