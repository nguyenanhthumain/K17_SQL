##### Kết nối database

```python
import sqlite3
import pandas as pd
from datetime import datetime, timedelta
```

```python
# kết nối
conn = sqlite3.connect("tanh.db")
```

##### Tạo bảng

```python
students = [
    (1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
    (2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
    (3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
    (4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
    (5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
    (6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
    (7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
    (8, 'Ho Ngọc Mai', 'Toan Tin', 20, 8.8),
    (9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
    (10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
]

courses = [
    (12, 'Giai tich'),
    (34, 'Thong ke'),
    (26, 'Tin hoc')
]

# 3. Tạo bảng và chèn dữ liệu
cursor.execute('CREATE TABLE course (id INTEGER PRIMARY KEY, course_name TEXT)')
cursor.execute('CREATE TABLE student (student_id INTEGER, name TEXT, class TEXT, course_id INTEGER, score REAL)')

cursor.executemany('INSERT INTO course VALUES (?, ?)', courses)
cursor.executemany('INSERT INTO student VALUES (?, ?, ?, ?, ?)', students)

print("Đã khởi tạo database và chèn dữ liệu thành công")
```

##### Câu 1


```python
# Tích Descartes
query_cartesian = "SELECT * FROM student CROSS JOIN course"
df_cartesian = pd.read_sql_query(query_cartesian, conn)
print("--- Tích Descartes ---")
display(df_cartesian)
```

```python
query_inner = """
SELECT s.*, c.course_name
FROM student s
INNER JOIN course c ON s.course_id = c.id
"""
df_inner = pd.read_sql_query(query_inner, conn)
print("--- INNER JOIN ---")
display(df_inner)
```

```python
query_left = """
SELECT s.*, c.course_name
FROM student s
LEFT JOIN course c ON s.course_id = c.id
"""
df_left = pd.read_sql_query(query_left, conn)
print("--- LEFT JOIN ---")
display(df_left)
```

```python
query_right = """
SELECT c.course_name, s.*
FROM course c
LEFT JOIN student s ON c.id = s.course_id
"""
df_right = pd.read_sql_query(query_right, conn)
print("--- RIGHT JOIN ---")
display(df_right)
```

```python
query_full = """
SELECT s.*, c.course_name
FROM student s
LEFT JOIN course c ON s.course_id = c.id
UNION
SELECT s.*, c.course_name
FROM course c
LEFT JOIN student s ON c.id = s.course_id
"""
df_full = pd.read_sql_query(query_full, conn)
print("--- FULL OUTER JOIN ---")
display(df_full)
```

> Nhận xét:
- Tích Descartes: Phù hợp để tạo tổ hợp nhưng dữ liệu thường bị nhiễu 

- INNER JOIN: Giúp dữ liệu sạch nhưng làm mất đi các sinh viên chưa có mã môn hợp lệ.

- LEFT/RIGHT JOIN: Dùng khi muốn giữ lại các đối tượng chưa có dữ liệu

##### Câu 3

```python
query_3a = """
SELECT name, score, 
       RANK() OVER (ORDER BY score DESC) as rank
FROM student
"""
df_3a = pd.read_sql_query(query_3a, conn)

print("--- Xếp hạng theo điểm số toàn trường ---")
display(df_3a)

# Top 3 cao nhất và thấp nhất
print("Top 3 điểm cao nhất:")
display(df_3a.head(3))
print("Top 3 điểm thấp nhất:")
display(df_3a.tail(3))
```

```python
query_3b = """
SELECT name, class, score,
       RANK() OVER (PARTITION BY class ORDER BY score DESC) as rank_in_class
FROM student
"""
df_3b = pd.read_sql_query(query_3b, conn)

print("--- Xếp hạng theo lớp học ---")
display(df_3b)

# Lấy Top 3 mỗi lớp 
top3_class = pd.read_sql_query(f"SELECT * FROM ({query_3b}) WHERE rank_in_class <= 3", conn)
print("Top 3 sinh viên của mỗi lớp:")
display(top3_class)
```

```python
query_3c = """
SELECT s.name, c.course_name, s.score,
       RANK() OVER (PARTITION BY s.course_id ORDER BY s.score DESC) as rank_in_course
FROM student s
JOIN course c ON s.course_id = c.id
"""
df_3c = pd.read_sql_query(query_3c, conn)

print("--- Xếp hạng theo môn học ---")
display(df_3c)

# Lấy Top 3 mỗi môn học
top3_course = pd.read_sql_query(f"SELECT * FROM ({query_3c}) WHERE rank_in_course <= 3", conn)
print("Top 3 sinh viên của mỗi môn học:")
display(top3_course)
```

> Nhận xét:
- Toàn trường: Những bạn học lớp Kinh Te đang chiếm giữ các vị trí đầu bảng với mức điểm tuyệt đối trong bộ dữ liệu (9.2).

- Theo lớp: Xếp hạng theo lớp giúp nhận ra các sinh viên nổi bật trong từng chuyên ngành (như bạn Vu Quoc Anh đứng đầu lớp Máy Tính dù điểm tổng quát có thể thấp hơn các bạn lớp Kinh Tế).

- Theo môn học: Thấy sự cạnh tranh trong từng học phần. Môn Giai tich có sự cạnh tranh cao hơn do số lượng sinh viên sau khi cập nhật dữ liệu tập trung khá đông ở đây.

##### Câu 4

```python
# Thêm cột graduation_date vào bảng student
cursor.execute("ALTER TABLE student ADD COLUMN graduation_date DATETIME")
print("Đã thêm cột graduation_date thành công!")
```

```python
# Sử dụng subquery để tính rank và cập nhật vào bảng chính
update_query = """
WITH student_ranks AS (
    SELECT student_id, 
           RANK() OVER (ORDER BY score DESC) as st_rank
    FROM student
)
UPDATE student
SET graduation_date = (
    SELECT date('now', '+' || st_rank || ' years')
    FROM student_ranks
    WHERE student_ranks.student_id = student.student_id
)
"""

cursor.execute(update_query)
conn.commit()

# Hiển thị kết quả kiểm tra
df_final = pd.read_sql_query("SELECT name, score, graduation_date FROM student ORDER BY score DESC", conn)
print("--- Kết quả cập nhật ngày tốt nghiệp ---")
display(df_final)
```

> Nhận xét:
- Bằng cách này, những sinh viên có kết quả học tập tốt hơn (thứ hạng thấp hơn như 1, 2) sẽ có ngày tốt nghiệp sớm hơn so với các bạn có thứ hạng cao (3, 4, 5...)
