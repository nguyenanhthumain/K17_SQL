### PHẦN 1

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
''')

students_data = [
    (1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
    (2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
    (3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
    (4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
    (5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
    (6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
    (7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
    (8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
    (9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
    (10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
]
cursor.executemany('INSERT INTO student VALUES (?,?,?,?,?)', students_data)

cursor.execute('''
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
''')

courses_data = [
    (12, 'Giai tich'),
    (34, 'Thong ke'),
    (26, 'Tin hoc')
]
cursor.executemany('INSERT INTO course VALUES (?,?)', courses_data)
def query(sql):
    df = pd.read_sql_query(sql, conn)
    print(f"Tổng số dòng: {len(df)}") 
    return df

df_student = pd.read_sql_query("SELECT * FROM student", conn)
print(f"Tổng số dòng student: {len(df_student)}")
display(df_student)

df_course = pd.read_sql_query("SELECT * FROM course", conn)
print(f"Tổng số dòng course: {len(df_course)}")
display(df_course)

print("Đã tạo xong 2 bảng")
```

```python
# Tích Descartes
sql_descartes = """
SELECT 
    student.*, 
    course.id AS course_table_id, 
    course.course_name 
FROM student 
CROSS JOIN course
"""

display(query(sql_descartes))
```

```python
# INNER JOIN
sql_inner = """
SELECT s.*, c.course_name
FROM student s
INNER JOIN course c ON s.course_id = c.id
"""
display(query(sql_inner))
```

```python
# LEFT JOIN 
sql_left = """
SELECT student.*, course.id AS course_table_id, course.course_name 
FROM student 
LEFT JOIN course ON student.course_id = course.id
"""
display(query(sql_left))
```

```python
# RIGHT JOIN 
sql_right = """
SELECT student.*, course.id AS course_table_id, course.course_name 
FROM course 
LEFT JOIN student ON student.course_id = course.id
"""
display(query(sql_right))
```

```python
# FULL OUTER JOIN 
sql_full = """
SELECT student.*, course.id AS course_table_id, course.course_name 
FROM student 
LEFT JOIN course ON student.course_id = course.id
UNION
SELECT student.*, course.id AS course_table_id, course.course_name 
FROM course 
LEFT JOIN student ON student.course_id = course.id
"""
display(query(sql_full))
```

```python
# a. Tổng số sinh viên và điểm trung bình từng lớp
sql_2a = """
SELECT class,
       COUNT(student_id) AS total_students,
       ROUND(AVG(score), 2) AS average_score
FROM student
GROUP BY class
"""
print(" Thống kê theo Lớp ")
display(query(sql_2a))
```

```python
# b. Tổng số sinh viên và điểm trung bình từng môn học
sql_2b = """
SELECT c.course_name, 
       COUNT(s.student_id) AS total_students, 
       ROUND(AVG(s.score), 2) AS average_score
FROM course c                       -- Đổi course lên trước
LEFT JOIN student s ON c.id = s.course_id  -- Dùng LEFT JOIN để lấy hết các môn
GROUP BY c.course_name
"""

print(" Thống kê theo Môn học (Đầy đủ các môn) ")
display(query(sql_2b))
```

```python
# c. Phân loại thi đua
sql_2c = """
SELECT c.course_name,
       ROUND(AVG(s.score), 2) AS avg_score,
       CASE
            WHEN AVG(s.score) >= 9.0 THEN 'Xuất sắc'
            WHEN AVG(s.score) >= 6.0 THEN 'Tốt'
            ELSE 'Kém'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
print(" Phân loại thi đua theo môn học ")
display(query(sql_2c))
```

### PHẦN 3


A. Xếp hạng sinh viên theo điểm số toàn trường

```python
# Xếp hạng sinh viên điểm cao nhất toàn trường
sql_3a = """
WITH RankedStudents AS (
    SELECT 
        student_id, 
        name, 
        class, 
        course_id, 
        score,
        RANK() OVER (ORDER BY score DESC) as rank_high
    FROM student
)
SELECT 
    student_id, 
    name, 
    class, 
    course_id, 
    score, 
    rank_high 
FROM RankedStudents
WHERE rank_high <= 6
ORDER BY score DESC
"""

print(" Top sinh viên điểm cao nhất toàn trường ")
display(query(sql_3a))
```

```python
# Xếp hạng sinh viên điểm thấp nhất toàn trường
sql_3a_low = """
WITH RankedLowStudents AS (
    SELECT 
        student_id, 
        name, 
        class, 
        course_id, 
        score,
        RANK() OVER (ORDER BY score ASC) as rank_low
    FROM student
)
SELECT 
    student_id, 
    name, 
    class, 
    course_id, 
    score, 
    rank_low 
FROM RankedLowStudents
WHERE rank_low <= 6
ORDER BY score ASC
"""

print(" Top sinh viên có điểm thấp nhất toàn trường ")
display(query(sql_3a_low))
```

B. Xếp hạng sinh viên theo điểm số lớp học

```python
# Xếp hạng sinh viên điểm cao nhất theo lớp học
sql_3b = """
WITH ClassRank AS (
    SELECT 
        student_id, 
        name, 
        class, 
        course_id, 
        score,
        RANK() OVER (PARTITION BY class ORDER BY score DESC) as rank_high
    FROM student
)
SELECT 
    student_id, 
    name, 
    class, 
    course_id, 
    score, 
    rank_high 
FROM ClassRank
WHERE rank_high <= 5
ORDER BY class, rank_high
"""

print(" Top sinh viên điểm cao nhất theo Lớp ")
display(query(sql_3b))
```

```python
# Xếp hạng sinh viên điểm thấp nhất theo lớp học
sql_3b_low = """
WITH ClassLowRank AS (
    SELECT 
        student_id, 
        name, 
        class, 
        course_id, 
        score,
        RANK() OVER (PARTITION BY class ORDER BY score ASC) as rank_low
    FROM student
)
SELECT 
    student_id, 
    name, 
    class, 
    course_id, 
    score, 
    rank_low 
FROM ClassLowRank
WHERE rank_low <= 5
ORDER BY class, rank_low
"""

print(" Top sinh viên có điểm thấp nhất theo từng Lớp ")
display(query(sql_3b_low))
```

C. Xếp hạng sinh viên theo điểm số môn học

```python
# Xếp hạng sinh viên điểm cao nhất theo môn học 
sql_3c = """
WITH CourseRank AS (
    SELECT 
        s.student_id, 
        s.name, 
        s.class, 
        s.course_id, 
        s.score, 
        c.course_name,
        ROW_NUMBER() OVER (PARTITION BY s.course_id ORDER BY s.score DESC) as rank_high
    FROM student s
    JOIN course c ON s.course_id = c.id
)
SELECT 
    student_id, 
    name, 
    class, 
    course_id, 
    course_name, 
    score, 
    rank_high
FROM CourseRank
WHERE rank_high <= 5
ORDER BY course_name, rank_high
"""

print(" Top sinh viên điểm cao nhất theo Môn học ")
display(query(sql_3c))
```

```python
#  Xếp hạng sinh viên điểm thấp nhất theo môn học 
sql_3c_low = """
WITH CourseLowRank AS (
    SELECT 
        s.student_id, 
        s.name, 
        s.class, 
        s.course_id, 
        s.score, 
        c.course_name,
        -- Sử dụng ASC để điểm thấp nhất trong mỗi môn đứng hạng 1
        RANK() OVER (PARTITION BY s.course_id ORDER BY s.score ASC) as rank_low
    FROM student s
    JOIN course c ON s.course_id = c.id
)
SELECT 
    student_id, 
    name, 
    class, 
    course_id, 
    course_name, 
    score, 
    rank_low
FROM CourseLowRank
WHERE rank_low <= 5
ORDER BY course_name, rank_low
"""

print(" Top sinh viên có điểm thấp nhất theo Môn học ")
display(query(sql_3c_low))
```

```python
# 1. Thêm cột mới với kiểu dữ liệu DATETIME
try:
    cursor.execute("ALTER TABLE student ADD COLUMN graduation_date DATETIME")
    conn.commit()
    print(" Đã thêm cột graduation_date thành công ")
except sqlite3.OperationalError:
    print("Cột đã tồn tại.")
```

```python
# 2. Cập nhật ngày tốt nghiệp: Ngày hiện tại + (Xếp hạng theo điểm số) ngày
sql_update_date = """
WITH Ranked AS (
    SELECT student_id,
           RANK() OVER (ORDER BY score DESC) as rnk
    FROM student
)
UPDATE student
SET graduation_date = (
    SELECT date('now', '+' || rnk || ' days')
    FROM Ranked
    WHERE Ranked.student_id = student.student_id
)
"""

cursor.execute(sql_update_date)
conn.commit()

print("Kết quả bảng Student sau khi cập nhật ngày tốt nghiệp")
display(query("SELECT student_id, name, score, graduation_date FROM student ORDER BY score DESC"))
```
