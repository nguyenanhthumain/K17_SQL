```python
import sqlite3
import pandas as pd

# tạo database
conn = sqlite3.connect(":memory:")
cur = conn.cursor()

# tạo bảng course
cur.execute("""
CREATE TABLE course(
    id INTEGER PRIMARY KEY,
    course_name TEXT
)
""")

# tạo bảng student
cur.execute("""
CREATE TABLE student(
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

# dữ liệu course
courses = [
    (12, "Giai tich"),
    (34, "Thong ke"),
    (26, "Tin hoc"),
    (20, "Tin hoc"),
    (24, "Tin hoc")
]

cur.executemany("INSERT INTO course VALUES (?,?)", courses)

# dữ liệu student
students = [
    (1,"Nguyen Minh Hoang","May Tinh",12,6.7),
    (2,"Tran Thi Lan","Kinh Te",34,9.2),
    (3,"Pham Van Nam","Toan Tin",None,7.9),
    (4,"Le Thanh Huyen","Toan Tin",20,7.2),
    (5,"Vu Quoc Anh","May Tinh",24,8.0),
    (6,"Dang Thuy Linh","May Tinh",24,5.5),
    (7,"Bui Tien Dung","Kinh Te",34,9.2),
    (8,"Ho Ngoc Mai","Toan Tin",20,8.8),
    (9,"Duong Huu Phuc","Kinh Te",None,7.2),
    (10,"Cao Thi Hanh","May Tinh",None,7.0)
]

cur.executemany("INSERT INTO student VALUES (?,?,?,?,?)", students)

conn.commit()

print("Database created!")
```

```python
# Decartes product
query = "SELECT * FROM student, course"
print("DECARTES PRODUCT")
display(pd.read_sql(query, conn))


# INNER JOIN
query = """
SELECT *
FROM student
INNER JOIN course
ON student.course_id = course.id
"""
print("INNER JOIN")
display(pd.read_sql(query, conn))


# LEFT JOIN
query = """
SELECT *
FROM student
LEFT JOIN course
ON student.course_id = course.id
"""
print("LEFT JOIN")
display(pd.read_sql(query, conn))


# 4. RIGHT JOIN #do sqlite không hỗ trợ nên em mô phỏng bằng left join , bằng cách đảo vị trí bảng
query = """
SELECT *
FROM course
LEFT JOIN student
ON student.course_id = course.id
"""
print("RIGHT JOIN")
display(pd.read_sql(query, conn))


# 5. FULL OUTER JOIN cũng do sqlite không hỗ trợ nên em kết hợp hai bảng bằng union
query = """
SELECT *
FROM student
LEFT JOIN course
ON student.course_id = course.id

UNION

SELECT *
FROM course
LEFT JOIN student
ON student.course_id = course.id
"""
print("FULL OUTER JOIN")
display(pd.read_sql(query, conn))
```

kết quả thu được :
+ decartes thì nó sẽ ra cái bảng mà mỗi thành phần của bảng student sẽ kết hợp với tât cả các thành phần của bảng course tạo ra một bảng có số dòng bằng số dòng bảng student nhân với số dòng bảng course
+ left join thì nó sẽ lấy phần chung và phần còn lại của bảng bên trái
+ right join thì cũng tương tự nhưng nó lấy phần chung và phần còn lại của bảng bên phải
+ full outer join là lấy tất

```python
# loại bỏ bản ghi course không tồn tại
cur.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
AND course_id IS NOT NULL
""")

conn.commit()

# a. Tổng số sinh viên + điểm TB từng lớp
query = """
SELECT class,
COUNT(*) AS total_students,
AVG(score) AS avg_score
FROM student
GROUP BY class
"""
print("Thống kê theo lớp")
display(pd.read_sql(query, conn))


# b. Tổng số sinh viên + điểm TB từng môn
query = """
SELECT course_name,
COUNT(student_id) AS total_students,
AVG(score) AS avg_score
FROM student
JOIN course
ON student.course_id = course.id
GROUP BY course_name
"""
print("Thống kê theo môn")
display(pd.read_sql(query, conn))


# c. Phân loại thi đua
query = """
SELECT student.name,
course.course_name,
score,
CASE
WHEN score >= 9 THEN 'Xuat sac'
WHEN score >= 6 THEN 'Tot'
ELSE 'Kem'
END AS classification
FROM student
JOIN course
ON student.course_id = course.id
"""
print("Phân loại sinh viên")
display(pd.read_sql(query, conn))
```

+ phần này dùng hàm gộp , sẽ gộp nhóm và tính toàn bằng các hàm có sẵn để tính theo nhóm , có sửu dụng cả hàm điều kiện

```python
# a. xếp hạng theo điểm
query = """
SELECT name,
score,
RANK() OVER (ORDER BY score DESC) as rank_score
FROM student
"""
display(pd.read_sql(query, conn))


# b. xếp hạng theo lớp
query = """
SELECT name,
class,
score,
RANK() OVER (PARTITION BY class ORDER BY score DESC) as rank_class
FROM student
"""
display(pd.read_sql(query, conn))


# c. xếp hạng theo môn
query = """
SELECT name,
course_id,
score,
RANK() OVER (PARTITION BY course_id ORDER BY score DESC) as rank_course
FROM student
"""
display(pd.read_sql(query, conn))


# thêm trường graduation_date
cur.execute("""
ALTER TABLE student
ADD COLUMN graduation_date DATETIME
""")

# cập nhật ngày tốt nghiệp dựa trên thứ hạng
cur.execute("""
UPDATE student
SET graduation_date =
datetime('now','+' || score || ' days')
""")

conn.commit()

print("Graduation date updated!")

display(pd.read_sql("SELECT * FROM student", conn))
```
