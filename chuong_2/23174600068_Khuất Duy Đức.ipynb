```python

import sqlite3

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Tạo bảng student
cursor.execute("""
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

# Tạo bảng course
cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")
```

```python
cursor.execute("SELECT * FROM student, course")
print(cursor.fetchall())
```

```python
# INNER JOIN
cursor.execute("""
SELECT *
FROM student s
JOIN course c ON s.course_id = c.id
""")
print(cursor.fetchall())
```

```python
# LEFT JOIN
cursor.execute("""
SELECT *
FROM student s
LEFT JOIN course c ON s.course_id = c.id
""")
print(cursor.fetchall())
```

```python
# RIGHT JOIN
cursor.execute("""
SELECT *
FROM course c
LEFT JOIN student s ON s.course_id = c.id
""")
print(cursor.fetchall())
```

```python
# FULL JOIN
cursor.execute("""
SELECT *
FROM student s
LEFT JOIN course c ON s.course_id = c.id

UNION

SELECT *
FROM course c
LEFT JOIN student s ON s.course_id = c.id
""")
print(cursor.fetchall())
```

```python
# Gán NULL → 26
cursor.execute("""
UPDATE student
SET course_id = 26
WHERE course_id IS NULL
""")
```

```python
# Xóa course không tồn tại
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")

conn.commit()
```

```python
# Thống kê theo lớp
cursor.execute("""
SELECT class, COUNT(*), AVG(score)
FROM student
GROUP BY class
""")
print(cursor.fetchall())
```

```python
# Theo môn
cursor.execute("""
SELECT c.course_name, COUNT(*), AVG(s.score)
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""")
print(cursor.fetchall())
```

```python
# Phân loại
cursor.execute("""
SELECT c.course_name,
       AVG(s.score),
       CASE
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""")
print(cursor.fetchall())
```

```python
# Xếp hạng theo điểm
cursor.execute("""
SELECT name, score,
       RANK() OVER (ORDER BY score DESC)
FROM student
""")
print(cursor.fetchall())
```

```python
# Top 3 cao nhất
cursor.execute("""
SELECT *
FROM (
    SELECT name, score,
           RANK() OVER (ORDER BY score DESC) r
    FROM student
)
WHERE r <= 3
""")
print(cursor.fetchall())
```

```python
# Top 3 thấp nhất
cursor.execute("""
SELECT *
FROM (
    SELECT name, score,
           RANK() OVER (ORDER BY score ASC) r
    FROM student
)
WHERE r <= 3
""")
print(cursor.fetchall())
```

```python
# Theo lớp
cursor.execute("""
SELECT name, class, score,
       RANK() OVER (PARTITION BY class ORDER BY score DESC)
FROM student
""")
print(cursor.fetchall())
```

```python
# Theo môn
cursor.execute("""
SELECT name, course_id, score,
       RANK() OVER (PARTITION BY course_id ORDER BY score DESC)
FROM student
""")
print(cursor.fetchall())
```

```python
#  graduation_date

cursor.execute("PRAGMA table_info(student)")
cols = [c[1] for c in cursor.fetchall()]

if "graduation_date" not in cols:
    cursor.execute("""
    ALTER TABLE student
    ADD COLUMN graduation_date TEXT
    """)

cursor.execute("""
UPDATE student
SET graduation_date = DATETIME('now', '+' ||
    (
        SELECT COUNT(*)
        FROM student s2
        WHERE s2.score > student.score
    ) || ' days'
)
""")

conn.commit()
```

```python
# Cập nhật ngày tốt nghiệp
cursor.execute("""
UPDATE student
SET graduation_date = DATETIME('now', '+' || (
    SELECT r FROM (
        SELECT student_id,
               RANK() OVER (ORDER BY score DESC) r
        FROM student
    ) t
    WHERE t.student_id = student.student_id
) || ' days')
""")

conn.commit()
```
