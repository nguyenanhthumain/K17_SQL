## 1. Hãy kết nối hai bảng trên theo những cách sau: 


- Sử dụng tích Decartes. 

```python
query_cross = """
SELECT * FROM student 
CROSS JOIN course;
"""
df_cross = pd.read_sql_query(query_cross, conn)
display(df_cross)
print(f"Kết luận: Tích Descartes tạo ra {len(df_cross)} bản ghi (10 sinh viên x 3 môn học).")
```

- Sử dụng JOIN: INNER JOIN, LEFT JOIN, RIGHT JOIN, FULL OUTER JOIN. 

```python
# INNER JOIN: 
query_inner = """
SELECT * FROM student 
INNER JOIN course ON student.course_id = course.id;
"""
df_inner = pd.read_sql_query(query_inner, conn)
print("\n--- INNER JOIN ---")
display(df_inner)

# LEFT JOIN:
query_left = """
SELECT * FROM student 
LEFT JOIN course ON student.course_id = course.id;
"""
df_left = pd.read_sql_query(query_left, conn)
print("\n--- LEFT JOIN ---")
display(df_left)

# RIGHT JOIN: 
query_right = """
SELECT * FROM student 
RIGHT JOIN course ON student.course_id = course.id;
"""
df_right = pd.read_sql_query(query_right, conn)
print("\n--- RIGHT JOIN ---")
display(df_right)

# FULL OUTER JOIN:  
query_full = """
SELECT * FROM student 
FULL OUTER JOIN course ON student.course_id = course.id;
"""
df_full = pd.read_sql_query(query_full, conn)
print("\n--- FULL OUTER JOIN ---")
display(df_full)
```

#### 2. Hãy cập nhật những giá trị course_id còn thiếu trong bảng student bằng câu lệnh SQL,trong đó các giá trị được điền là những giá trị nằm trong bảng course  và loại bỏ những bản ghi tham gia những môn học không tồn tại bảng course. Sau đó hãy cho biết: 

```python
cursor = conn.cursor()
query_update = """
UPDATE student 
SET course_id = 26 
WHERE course_id IS NULL;
"""
cursor.execute(query_update)

# Loại bỏ những bản ghi có course_id không tồn tại trong bảng course
query_delete = """
DELETE FROM student 
WHERE course_id NOT IN (SELECT id FROM course);
"""
cursor.execute(query_delete)
conn.commit()

# Hiển thị lại bảng student để kiểm tra
df_student_updated = pd.read_sql_query("SELECT * FROM student", conn)
print("Bảng student sau khi được làm sạch:")
display(df_student_updated)
print("Kết luận: Các sinh viên thiếu course_id đã được gán mã 26. Các sinh viên có mã môn học 20 và 24 đã bị xóa.\n")
```

a. Tổng số sinh viên, điểm trung bình của từng lớp. 

```python
query_2a = """
SELECT class, COUNT(student_id) AS tong_sinh_vien, ROUND(AVG(score), 2) AS diem_trung_binh
FROM student
GROUP BY class;
"""
df_2a = pd.read_sql_query(query_2a, conn)
print("\n--- 2a. Thống kê theo Lớp ---")
display(df_2a)
```

b. Tổng số sinh viên, điểm trung bình của từng môn học.

```python
query_2b = """
SELECT c.course_name, s.course_id, COUNT(s.student_id) AS tong_sinh_vien, ROUND(AVG(s.score), 2) AS diem_trung_binh
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY s.course_id;
"""
df_2b = pd.read_sql_query(query_2b, conn)
print("\n--- 2b. Thống kê theo Môn học ---")
display(df_2b)
```

c. Phân loại thi đua theo số điểm của từng môn học biết: 
- i. Điểm TB ≥  9.0: Xuất sắc. 
- ii. 6.0 ≤ Điểm TB ≤ 8.9: Tốt. 
- iii. Điểm TB < 6.0: Kém. 

```python
query_2c = """
SELECT student_id, name, course_id, score,
    CASE 
        WHEN score >= 9.0 THEN 'Xuất sắc'
        WHEN score >= 6.0 AND score <= 8.9 THEN 'Tốt'
        ELSE 'Kém'
    END AS phan_loai
FROM student;
"""
df_2c = pd.read_sql_query(query_2c, conn)
print("\n--- 2c. Phân loại thi đua ---")
display(df_2c)
print("Kết luận: Sinh viên Trần Thị Lan và Bùi Tiến Dũng đạt loại Xuất sắc do điểm >= 9.0, các sinh viên còn lại đạt loại Tốt. Không có sinh viên nào loại Kém (điểm < 6.0) trong danh sách hiện tại.")
```

### 3. Hãy xếp hạng sinh viên thông qua: 
- a. Điểm số. 
- b. Điểm số theo lớp học. 
- c. Điểm số theo mã môn học. 

và cho biết top 3 sinh viện đạt thứ hạng cao nhất, top 3 sinh viên đạt thứ hạng thấp nhất theo 
từng trường hợp trên. 

```python
# --- 3a. Xếp hạng sinh viên qua ĐIỂM SỐ ---
print("3a. XẾP HẠNG THEO ĐIỂM SỐ")

# Bảng xếp hạng 
query_3a_rank = """
WITH RankedStudents AS (
    SELECT student_id, name, class, course_id, score,
           RANK() OVER (ORDER BY score DESC) as hang_cao_xuong_thap,
           RANK() OVER (ORDER BY score ASC) as hang_thap_len_cao
    FROM student
)
SELECT * FROM RankedStudents;
"""
df_3a_rank = pd.read_sql_query(query_3a_rank, conn)
display(df_3a_rank[['student_id', 'name', 'score', 'hang_cao_xuong_thap']])

# Top 3 cao nhất 
df_3a_top3 = df_3a_rank[df_3a_rank['hang_cao_xuong_thap'] <= 3]
print("\nTop 3 sinh viên điểm CAO NHẤT:")
display(df_3a_top3[['student_id', 'name', 'score', 'hang_cao_xuong_thap']])

# Top 3 thấp nhất 
df_3a_bot3 = df_3a_rank[df_3a_rank['hang_thap_len_cao'] <= 3]
print("\nTop 3 sinh viên điểm THẤP NHẤT:")
display(df_3a_bot3[['student_id', 'name', 'score', 'hang_thap_len_cao']])
```

```python
# --- 3b. Xếp hạng sinh viên theo LỚP HỌC ---
print("3b. XẾP HẠNG THEO LỚP HỌC")

query_3b_rank = """
WITH RankedByClass AS (
    SELECT student_id, name, class, course_id, score,
           RANK() OVER (PARTITION BY class ORDER BY score DESC) as hang_cao_xuong_thap,
           RANK() OVER (PARTITION BY class ORDER BY score ASC) as hang_thap_len_cao
    FROM student
)
SELECT * FROM RankedByClass;
"""
df_3b_rank = pd.read_sql_query(query_3b_rank, conn)
display(df_3b_rank[['student_id', 'name', 'class', 'score', 'hang_cao_xuong_thap']])

# Top 3 cao nhất theo lớp
df_3b_top3 = df_3b_rank[df_3b_rank['hang_cao_xuong_thap'] <= 3]
print("\nTop 3 sinh viên điểm CAO NHẤT trong từng lớp:")
display(df_3b_top3[['student_id', 'name', 'class', 'score', 'hang_cao_xuong_thap']])

# Top 3 thấp nhất theo lớp
df_3b_bot3 = df_3b_rank[df_3b_rank['hang_thap_len_cao'] <= 3]
print("\nTop 3 sinh viên điểm THẤP NHẤT trong từng lớp:")
display(df_3b_bot3[['student_id', 'name', 'class', 'score', 'hang_thap_len_cao']])
```

```python
# --- 3c. Xếp hạng sinh viên theo MÃ MÔN HỌC ---
print("3c. XẾP HẠNG THEO MÃ MÔN HỌC")

query_3c_rank = """
WITH RankedByCourse AS (
    SELECT student_id, name, class, course_id, score,
           RANK() OVER (PARTITION BY course_id ORDER BY score DESC) as hang_cao_xuong_thap,
           RANK() OVER (PARTITION BY course_id ORDER BY score ASC) as hang_thap_len_cao
    FROM student
)
SELECT * FROM RankedByCourse;
"""
df_3c_rank = pd.read_sql_query(query_3c_rank, conn)
display(df_3c_rank[['student_id', 'name', 'course_id', 'score', 'hang_cao_xuong_thap']])

# Top 3 cao nhất theo môn
df_3c_top3 = df_3c_rank[df_3c_rank['hang_cao_xuong_thap'] <= 3]
print("\nTop 3 sinh viên điểm CAO NHẤT trong từng môn học:")
display(df_3c_top3[['student_id', 'name', 'course_id', 'score', 'hang_cao_xuong_thap']])

# Top 3 thấp nhất theo môn
df_3c_bot3 = df_3c_rank[df_3c_rank['hang_thap_len_cao'] <= 3]
print("\nTop 3 sinh viên điểm THẤP NHẤT trong từng môn học:")
display(df_3c_bot3[['student_id', 'name', 'course_id', 'score', 'hang_thap_len_cao']])
```

#### 4. Hãy bổ sung thêm một trường graduation_date có kiểu dữ liệu là DATETIME vào bảng student để xác định thời gian tốt nghiệp của từng bạn, trong đó thời gian tốt nghiệp được xác định bởi thời gian hiện tại cộng với số hạng tương ứng của bạn đó tính theo điểm số. 

```python
# --- 4. Bổ sung trường graduation_date và cập nhật thời gian tốt nghiệp ---
print("4. BỔ SUNG TRƯỜNG GRADUATION_DATE")

cursor = conn.cursor()
try:
    cursor.execute("ALTER TABLE student ADD COLUMN graduation_date DATETIME;")
except sqlite3.OperationalError:
    pass 
query_update_date = """
WITH Ranked AS (
    SELECT student_id, DENSE_RANK() OVER (ORDER BY score DESC) as thu_hang
    FROM student
)
UPDATE student
SET graduation_date = datetime('now', 'localtime', '+' || (SELECT thu_hang FROM Ranked WHERE Ranked.student_id = student.student_id) || ' days');
"""
cursor.execute(query_update_date)
conn.commit()
query_final = """
SELECT student_id, name, score, 
       DENSE_RANK() OVER (ORDER BY score DESC) as thu_hang,
       graduation_date 
FROM student 
ORDER BY score DESC;
"""
df_final = pd.read_sql_query(query_final, conn)
display(df_final)
```
