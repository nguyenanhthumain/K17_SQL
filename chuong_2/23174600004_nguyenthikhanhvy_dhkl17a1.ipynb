```python
import sqlite3

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Tạo bảng
cursor.execute("""
CREATE TABLE student (
    student_id INT,
    name TEXT,
    class TEXT,
    course_id INT,
    score FLOAT
);
""")

cursor.execute("""
INSERT INTO student VALUES
(1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
(2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
(3, 'Pham Van Nam', 'Toan Tin', NULL, 7.9),
(4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
(5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
(6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
(7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
(8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
(9, 'Duong Huu Phuc', 'Kinh Te', NULL, 7.2),
(10, 'Cao Thi Hanh', 'May Tinh', NULL, 7.0);
""")

cursor.execute("""
CREATE TABLE course (
    id INT,
    course_name TEXT
);
""")

cursor.execute("""
INSERT INTO course VALUES
(12, 'Giai tich'),
(34, 'Thong ke'),
(26, 'Tin hoc');
""")

# JOIN
print("\n--- JOIN ---")
for row in cursor.execute("""
SELECT s.student_id, s.name, s.class, c.course_name, s.score
FROM student s
LEFT JOIN course c ON s.course_id = c.id
"""):
    print(row)

# Thống kê
print("\n--- TB theo lớp ---")
for row in cursor.execute("""
SELECT class, COUNT(*), AVG(score)
FROM student
GROUP BY class
"""):
    print(row)

# Phân loại
print("\n--- Phân loại ---")
for row in cursor.execute("""
SELECT name, score,
CASE 
    WHEN score >= 9 THEN 'Xuat sac'
    WHEN score >= 6 THEN 'Tot'
    ELSE 'Kem'
END
FROM student
"""):
    print(row)

# Top 3
print("\n--- Top 3 cao nhất ---")
for row in cursor.execute("""
SELECT * FROM student ORDER BY score DESC LIMIT 3
"""):
    print(row)
```
