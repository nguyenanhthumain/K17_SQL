## **Câu 1. Hãy kết nối hai bảng trên theo những cách sau:**

### 1.1. Sử dụng tích Descartes

```python
print("TÍCH DESCARTES:")

query_descartes = """
SELECT
    s.student_id,
    s.name,
    s.class,
    s.course_id,
    s.score,
    c.id,
    c.course_name
FROM student s, course c;
"""

df_descartes = pd.read_sql_query(query_descartes, conn)
display(df_descartes)

print(f"Số dòng kết quả: {len(df_descartes)}")
```

Kết luận:
* Tích Descartes ghép mỗi sinh viên với tất cả các môn học trong bảng course.
* Vì bảng student có 10 dòng và bảng course có 3 dòng nên kết quả có 10 x 3 = 30 dòng.
* Kết quả này không phản ánh đúng quan hệ đăng ký môn học thực tế, mà chỉ minh họa phép kết hợp mọi khả năng giữa hai bảng.
* Các sinh viên có course_id bị thiếu vẫn xuất hiện bình thường vì phép tích Descartes không cần điều kiện nối.

### 1.2. INNER JOIN

```python
print("INNER JOIN:")

query_inner = """
SELECT
    s.student_id,
    s.name,
    s.class,
    s.course_id,
    s.score,
    c.id,
    c.course_name
FROM student s
INNER JOIN course c
    ON s.course_id = c.id;
"""

df_inner = pd.read_sql_query(query_inner, conn)
display(df_inner)

print(f"Số dòng kết quả: {len(df_inner)}")
```

Kết luận:
* INNER JOIN chỉ lấy các dòng có course_id trong bảng student khớp với id trong bảng course.
* Với dữ liệu hiện tại chỉ có các mã môn hợp lệ là 12 và 34 xuất hiện trong bảng student.
* Kết quả chỉ còn 3 sinh viên có mã môn hợp lệ:
  * Nguyễn Minh Hoàng (12)
  * Trần Thị Lan (34)
  * Bùi Tiến Dũng (34)
* Các dòng có course_id bị thiếu hoặc có giá trị không tồn tại trong course như 20, 24 đều không xuất hiện.

### 1.3. LEFT JOIN

```python
print("LEFT JOIN:")

query_left = """
SELECT
    s.student_id,
    s.name,
    s.class,
    s.course_id,
    s.score,
    c.id,
    c.course_name
FROM student s
LEFT JOIN course c
    ON s.course_id = c.id;
"""

df_left = pd.read_sql_query(query_left, conn)
display(df_left)

print(f"Số dòng kết quả: {len(df_left)}")
```

Kết luận:
* LEFT JOIN giữ lại toàn bộ dữ liệu của bảng student.
* Nếu course_id khớp với id trong bảng course thì sẽ lấy thêm course_name.
* Nếu không khớp thì các cột từ bảng course sẽ là NULL/NaN.
* Kết quả cho thấy:
  * Sinh viên có course_id = 12, 34 được ghép đúng môn học
  * Sinh viên có course_id = 20, 24 hoặc NULL không có thông tin môn học tương ứng

### 1.4. RIGHT JOIN trong SQLite

SQLite không hỗ trợ RIGHT JOIN trực tiếp trong nhiều cách làm cơ bản của môn học, nên em làm bằng cách đổi vị trí hai bảng và dùng LEFT JOIN.

```python
print("RIGHT JOIN:")

query_right = """
SELECT
    c.id,
    c.course_name,
    s.student_id,
    s.name,
    s.class,
    s.course_id,
    s.score
FROM course c
LEFT JOIN student s
    ON c.id = s.course_id;
"""

df_right = pd.read_sql_query(query_right, conn)
display(df_right)

print(f"Số dòng kết quả: {len(df_right)}")
```

Kết luận:
* RIGHT JOIN có ý nghĩa là giữ lại toàn bộ dữ liệu của bảng bên phải trong câu student RIGHT JOIN course.
* Vì SQLite không làm trực tiếp, ta đổi thành: course LEFT JOIN student
* Kết quả giữ lại toàn bộ các môn học trong bảng course.
* Môn Tin hoc có mã 26 vẫn xuất hiện dù chưa có sinh viên nào đăng ký.
* Những dòng không có sinh viên khớp sẽ có các cột phía student bằng NULL/NaN.

### 1.5. FULL OUTER JOIN trong SQLite

SQLite cũng không hỗ trợ FULL OUTER JOIN trực tiếp, nên em làm bằng cách lấy hợp của LEFT JOIN và phần còn lại từ phía bảng course.

```python
print("FULL OUTER JOIN:")

query_full = """
SELECT
    s.student_id,
    s.name,
    s.class,
    s.course_id,
    s.score,
    c.id,
    c.course_name
FROM student s
LEFT JOIN course c
    ON s.course_id = c.id

UNION

SELECT
    s.student_id,
    s.name,
    s.class,
    s.course_id,
    s.score,
    c.id,
    c.course_name
FROM course c
LEFT JOIN student s
    ON c.id = s.course_id
WHERE s.student_id IS NULL;
"""

df_full = pd.read_sql_query(query_full, conn)
display(df_full)

print(f"Số dòng kết quả: {len(df_full)}")
```

Kết luận:
* FULL OUTER JOIN giữ lại toàn bộ dữ liệu của cả hai bảng.
* Các dòng khớp nhau sẽ ghép lại thành một dòng.
* Các dòng không khớp từ bảng student hoặc course vẫn được giữ lại.
* Trong kết quả:
  * Các sinh viên có course_id không hợp lệ hoặc bị thiếu vẫn xuất hiện
  * Môn Tin hoc (id = 26) cũng xuất hiện dù chưa có sinh viên đăng ký

## **2. Hãy cập nhật những giá trị course_id còn thiếu trong bảng student bằng câu lệnh SQL, trong đó các giá trị được điền là những giá trị nằm trong bảng course và loại bỏ những bản ghi tham gia những môn học không tồn tại bảng course. Sau đó hãy cho biết:**

**a. Tổng số sinh viên, điểm trung bình của từng lớp.**

**b. Tổng số sinh viên, điểm trung bình của từng môn học.**

**c. Phân loại thi đua theo số điểm của từng môn học biết:**

**-   ii. 6.0 ≤ Điểm TB ≤ 8.9: Tốt.**

### 2.1. Cập nhật course_id còn thiếu và xóa các bản ghi không hợp lệ

```python
print("Cập nhật course_id còn thiếu và xóa dữ liệu không hợp lệ")

# Cập nhật các giá trị course_id bị thiếu thành một mã môn hợp lệ trong bảng course
cursor.execute("""
UPDATE student
SET course_id = 26
WHERE course_id IS NULL;
""")

# Xóa các bản ghi có course_id không tồn tại trong bảng course
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course);
""")

conn.commit()

# Kiểm tra lại dữ liệu sau cập nhật
query_check = """
SELECT *
FROM student;
"""

student_updated_df = pd.read_sql_query(query_check, conn)
display(student_updated_df)
```

### 2.2. Tổng số sinh viên và điểm trung bình của từng lớp

```python
print("Tổng số sinh viên và điểm trung bình của từng lớp:")

query_class = """
SELECT
    class,
    COUNT(student_id) AS total_students,
    ROUND(AVG(score), 2) AS avg_score
FROM student
GROUP BY class;
"""

df_class = pd.read_sql_query(query_class, conn)
display(df_class)
```

Kết luận:
* Kinh Te có điểm trung bình cao nhất.
* May Tinh có điểm trung bình thấp hơn, cần cải thiện thêm.
* Toan Tin có ít sinh viên sau khi làm sạch dữ liệu nên kết quả trung bình có thể ít đại diện hơn.

### 2.3. Tổng số sinh viên và điểm trung bình của từng môn học

```python
print("Tổng số sinh viên và điểm trung bình của từng môn học:")

query_course = """
SELECT
    course.course_name,
    COUNT(student.student_id) AS total_students,
    ROUND(AVG(student.score), 2) AS avg_score
FROM student
LEFT JOIN course
    ON student.course_id = course.id
GROUP BY course.course_name;
"""

df_course = pd.read_sql_query(query_course, conn)
display(df_course)
```

Kết luận:
* Thong ke có điểm trung bình cao nhất.
* Giai tich có số lượng sinh viên ít và điểm trung bình chưa cao.
* Tin hoc có số lượng sinh viên nhiều hơn nhưng điểm trung bình ở mức khá.

### 2.4. Phân loại thi đua theo điểm trung bình của từng môn học

```python
print("Phân loại thi đua theo điểm trung bình của từng môn học:")

query_rating = """
SELECT
    course.course_name,
    ROUND(AVG(student.score), 2) AS avg_score,
    CASE
        WHEN AVG(student.score) >= 9.0 THEN 'Xuất sắc'
        WHEN AVG(student.score) BETWEEN 6.0 AND 8.9 THEN 'Tốt'
        ELSE 'Kém'
    END AS rating
FROM student
LEFT JOIN course
    ON student.course_id = course.id
GROUP BY course.course_name;
"""

df_rating = pd.read_sql_query(query_rating, conn)
display(df_rating)
```

Kết luận:
* Môn có điểm trung bình từ 9.0 trở lên được xếp loại Xuất sắc.
* Môn có điểm trung bình từ 6.0 đến 8.9 được xếp loại Tốt.
* Môn có điểm trung bình dưới 6.0 được xếp loại Kém.
* Với dữ liệu hiện tại, Thong ke thuộc nhóm Xuất sắc, còn Giai tich và Tin hoc thuộc nhóm Tốt.

## 4. Hãy bổ sung thêm một trường graduation_date có kiểu dữ liệu là DATETIME vào bảng student để xác định thời gian tốt nghiệp của từng bạn, trong đó thời gian tốt nghiệp được xác định bởi thời gian hiện tại cộng với số hạng tương ứng của bạn đó tính theo điểm số

### 4.1. Thêm cột graduation_date vào bảng student

```python
print("Thêm cột graduation_date vào bảng student")

# Kiểm tra các cột hiện có trong bảng student
columns_info = pd.read_sql_query("PRAGMA table_info(student);", conn)

if "graduation_date" not in columns_info["name"].values:
    cursor.execute("""
    ALTER TABLE student ADD COLUMN graduation_date DATETIME;
    """)
    conn.commit()
    print("Đã thêm cột graduation_date.")
else:
    print("Cột graduation_date đã tồn tại.")
```

### 4.2. Cập nhật graduation_date dựa trên thứ hạng điểm số

```python
print("Cập nhật graduation_date dựa trên thứ hạng điểm số")

cursor.execute("""
WITH ranked_students AS (
    SELECT
        student_id,
        RANK() OVER (ORDER BY score DESC) AS score_rank
    FROM student
)
UPDATE student
SET graduation_date = DATETIME(
    'now',
    '+' || (
        SELECT score_rank
        FROM ranked_students
        WHERE ranked_students.student_id = student.student_id
    ) || ' months'
);
""")

conn.commit()
print("Đã cập nhật graduation_date.")
```

### 4.3. Kiểm tra kết quả sau khi cập nhật
