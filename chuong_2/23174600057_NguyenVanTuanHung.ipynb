## **Chuẩn bị dữ liệu**

```python
import pandas as pd
import sqlite3
from datetime import datetime, timedelta

# Tạo bảng student
student_data = [
    (1, "Nguyen Minh Hoang", "May Tinh", 12, 6.7),
    (2, "Tran Thi Lan", "Kinh Te", 34, 9.2),
    (3, "Pham Van Nam", "Toan Tin", None, 7.9),
    (4, "Le Thanh Huyen", "Toan Tin", 20, 7.2),
    (5, "Vu Quoc Anh", "May Tinh", 24, 8.0),
    (6, "Dang Thuy Linh", "May Tinh", 24, 5.5),
    (7, "Bui Tien Dung", "Kinh Te", 34, 9.2),
    (8, "Ho Ngoc Mai", "Toan Tin", 20, 8.8),
    (9, "Duong Huu Phuc", "Kinh Te", None, 7.2),
    (10, "Cao Thi Hanh", "May Tinh", None, 7.0),
]

student = pd.DataFrame(student_data, columns=["student_id","name","class","course_id","score"])

# Tạo bảng course
course_data = [
    (12, "Giai tich"),
    (34, "Thong ke"),
    (26, "Tin hoc")
]

course = pd.DataFrame(course_data, columns=["id","course_name"])
```

## **Tích Descartes**

```python
cartesian = student.merge(course, how='cross')
print(cartesian)
```

## **INNER JOIN**

```python
inner_join = student.merge(course, left_on="course_id", right_on="id", how="inner")
print(inner_join)
```

## **LEFT JOIN**

```python
left_join = student.merge(course, left_on="course_id", right_on="id", how="left")
print(left_join)
```

## **RIGHT JOIN**

```python
right_join = student.merge(course, left_on="course_id", right_on="id", how="right")
print(right_join)
```

## **FULL JOIN**

```python
full_join = student.merge(course, left_on="course_id", right_on="id", how="outer")
print(full_join)
```

# **Bài 2**

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

# Tạo bảng student
cursor.execute("""
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

# Tạo bảng course
cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")
```

## **Chèn dữ liệu**

```python
student_data = [
    (1, "Nguyen Minh Hoang", "May Tinh", 12, 6.7),
    (2, "Tran Thi Lan", "Kinh Te", 34, 9.2),
    (3, "Pham Van Nam", "Toan Tin", None, 7.9),
    (4, "Le Thanh Huyen", "Toan Tin", 20, 7.2),
    (5, "Vu Quoc Anh", "May Tinh", 24, 8.0),
    (6, "Dang Thuy Linh", "May Tinh", 24, 5.5),
    (7, "Bui Tien Dung", "Kinh Te", 34, 9.2),
    (8, "Ho Ngoc Mai", "Toan Tin", 20, 8.8),
    (9, "Duong Huu Phuc", "Kinh Te", None, 7.2),
    (10, "Cao Thi Hanh", "May Tinh", None, 7.0),
]

course_data = [
    (12, "Giai tich"),
    (34, "Thong ke"),
    (26, "Tin hoc")
]

cursor.executemany("INSERT INTO student VALUES (?,?,?,?,?)", student_data)
cursor.executemany("INSERT INTO course VALUES (?,?)", course_data)

conn.commit()
```

## **Cập nhật course_id còn thiếu**

```python
cursor.execute("""
UPDATE student
SET course_id = 26
WHERE course_id IS NULL
""")
conn.commit()
```

## **Xóa bacác bản ghi course không tồn tại**

```python
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")
conn.commit()
```

## **a.Tổng số SV & điểm TB theo lớp**

```python
query_a = """
SELECT class,
       COUNT(*) AS total_students,
       AVG(score) AS avg_score
FROM student
GROUP BY class
"""
df_a = pd.read_sql_query(query_a, conn)
print(df_a)
```

## **b. Tổng số SV & điểm TB theo môn học**

```python
query_b = """
SELECT c.course_name,
       COUNT(*) AS total_students,
       AVG(s.score) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
df_b = pd.read_sql_query(query_b, conn)
print(df_b)
```

## **c. Phân loại thi đua**

```python
query_c = """
SELECT c.course_name,
       AVG(s.score) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
df_c = pd.read_sql_query(query_c, conn)
print(df_c)
```

# **Bài 3**

## **a. Xếp hạng theo điểm toàn bộ**

```python
query_rank_all = """
SELECT *,
       RANK() OVER (ORDER BY score DESC) AS rank_all
FROM student
"""
df_rank_all = pd.read_sql(query_rank_all, conn)
print(df_rank_all)
```

## **Top 3 cao nhất & thấp nhất**

```python
# Top 3 cao nhất
top3_all = pd.read_sql("""
SELECT *
FROM student
ORDER BY score DESC
LIMIT 3
""", conn)

# Top 3 thấp nhất
bottom3_all = pd.read_sql("""
SELECT *
FROM student
ORDER BY score ASC
LIMIT 3
""", conn)

print("Top 3 cao nhất:\n", top3_all)
print("\nTop 3 thấp nhất:\n", bottom3_all)
```

## **b. Xếp hạng theo lớp**

```python
query_rank_class = """
SELECT *,
       RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_in_class
FROM student
"""
df_rank_class = pd.read_sql(query_rank_class, conn)
print(df_rank_class)
```

## **Top 3 mỗi lớp**

```python
top3_class = pd.read_sql("""
SELECT *
FROM (
    SELECT *,
           RANK() OVER (PARTITION BY class ORDER BY score DESC) AS r
    FROM student
)
WHERE r <= 3
""", conn)

print(top3_class)
```

## **Bottom 3 mỗi lớp**

```python
bottom3_class = pd.read_sql("""
SELECT *
FROM (
    SELECT *,
           RANK() OVER (PARTITION BY class ORDER BY score ASC) AS r
    FROM student
)
WHERE r <= 3
""", conn)

print(bottom3_class)
```

## **c. Xếp hạng theo môn học**

```python
query_rank_course = """
SELECT *,
       RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_in_course
FROM student
"""
df_rank_course = pd.read_sql(query_rank_course, conn)
print(df_rank_course)
```

## **Top 3 mỗi môn**

```python
top3_course = pd.read_sql("""
SELECT *
FROM (
    SELECT *,
           RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS r
    FROM student
)
WHERE r <= 3
""", conn)

print(top3_course)
```

## **Bottom 3 mỗi môn**

```python
bottom3_course = pd.read_sql("""
SELECT *
FROM (
    SELECT *,
           RANK() OVER (PARTITION BY course_id ORDER BY score ASC) AS r
    FROM student
)
WHERE r <= 3
""", conn)

print(bottom3_course)
```
