## Tạo bảng

### Tạo bảng student

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('student.db')
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")
```

### Tạo bảng course

```python
cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")
```

### Insert dữ liệu

```python
cursor.executemany("INSERT INTO student VALUES (?,?,?,?,?)", [
(1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
(2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
(3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
(4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
(5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
(6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
(7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
(8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
(9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
(10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
])

cursor.executemany("INSERT INTO course VALUES (?,?)", [
(12, 'Giai tich'),
(34, 'Thong ke'),
(26, 'Tin hoc')
])

conn.commit()
```

## 1. JOIN & DECARTES

### Tích Descartes

```python
pd.read_sql_query("SELECT * FROM student, course", conn)
```

### INNER JOIN

```python
pd.read_sql_query("""
SELECT * FROM student
INNER JOIN course
ON student.course_id = course.id
""", conn)
```

### LEFT JOIN

```python
pd.read_sql_query("""
SELECT * FROM student
LEFT JOIN course
ON student.course_id = course.id
""", conn)
```

### RIGHT JOIN (giả lập)

```python
pd.read_sql_query("""
SELECT * FROM course
LEFT JOIN student
ON student.course_id = course.id
""", conn)
```

### FULL OUTER JOIN (SQLite giả lập)

```python
pd.read_sql_query("""
SELECT * FROM student
LEFT JOIN course ON student.course_id = course.id
UNION
SELECT * FROM course
LEFT JOIN student ON student.course_id = course.id
""", conn)
```

**Nhận xét JOIN & Tích Descartes**

Tích Descartes (CROSS JOIN) tạo ra tất cả các tổ hợp có thể giữa hai bảng, dẫn đến số lượng bản ghi rất lớn và thường không có ý nghĩa thực tế.

INNER JOIN chỉ lấy các bản ghi có khóa khớp giữa hai bảng, đảm bảo tính chính xác của dữ liệu liên kết.

LEFT JOIN giữ lại toàn bộ dữ liệu từ bảng student, kể cả khi không có thông tin tương ứng trong bảng course.

RIGHT JOIN (giả lập trong SQLite) giữ lại toàn bộ dữ liệu từ bảng course.

FULL OUTER JOIN kết hợp cả hai bảng, đảm bảo không mất dữ liệu ở bất kỳ phía nào.

## 2. Xử lý dữ liệu

### Xóa các course_id không tồn tại (20, 24)

```python
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")
```

### Cập nhật NULL

```python
cursor.execute("""
UPDATE student
SET course_id = 12
WHERE course_id IS NULL
""")

conn.commit()
```

### Theo lớp

```python
pd.read_sql_query("""
SELECT class,
COUNT(*) AS total_students,
AVG(score) AS avg_score
FROM student
GROUP BY class
""", conn)
```

### Theo môn học

```python
pd.read_sql_query("""
SELECT course_name,
COUNT(*) AS total_students,
AVG(score) AS avg_score
FROM student
JOIN course ON student.course_id = course.id
GROUP BY course_name
""", conn)
```

### Phân loại

```python
pd.read_sql_query("""
SELECT name, score,
CASE
    WHEN score >= 9 THEN 'Xuất sắc'
    WHEN score >= 6 THEN 'Tốt'
    ELSE 'Kém'
END AS classification
FROM student
""", conn)
```

Nhận xét xử lý dữ liệu

Dữ liệu ban đầu tồn tại các giá trị course_id không hợp lệ (không có trong bảng course) và giá trị NULL.

Việc loại bỏ các bản ghi không hợp lệ giúp đảm bảo tính toàn vẹn dữ liệu.
Việc cập nhật các giá trị NULL giúp dữ liệu đầy đủ hơn để phục vụ cho phân tích.
Sau khi xử lý, dữ liệu đã sạch và sẵn sàng cho các phép tính thống kê.

## 3. Xếp hạng

### Toàn bộ

```python
pd.read_sql_query("""
SELECT name, score,
RANK() OVER (ORDER BY score DESC) AS rank
FROM student
""", conn)
```

### Theo lớp

```python
pd.read_sql_query("""
SELECT name, class, score,
RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank
FROM student
""", conn)
```

### Theo môn

```python
pd.read_sql_query("""
SELECT name, course_id, score,
RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank
FROM student
""", conn)
```

### Top 3 cao nhất

```python
pd.read_sql_query("""
SELECT * FROM (
    SELECT name, score,
    RANK() OVER (ORDER BY score DESC) as rank
    FROM student
)
WHERE rank <= 3
""", conn)
```

### Top 3 thấp nhất

```python
pd.read_sql_query("""
SELECT * FROM (
    SELECT name, score,
    RANK() OVER (ORDER BY score ASC) as rank
    FROM student
)
WHERE rank <= 3
""", conn)
```

## 4. graduation_date

### Thêm cột

```python
columns = pd.read_sql_query("PRAGMA table_info(student)", conn)

if 'graduation_date' not in columns['name'].values:
    cursor.execute("""
    ALTER TABLE student ADD COLUMN graduation_date DATETIME
    """)
```

### Update theo rank

```python
pd.read_sql_query("""
SELECT name, score,
RANK() OVER (ORDER BY score DESC) as rank,
graduation_date
FROM student
""", conn)
```
