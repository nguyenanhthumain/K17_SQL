```python
#CELL 1: Import & tạo database
import sqlite3
import pandas as pd

conn = sqlite3.connect("du_lieu_dantri.db")
cursor = conn.cursor()

# Xóa bảng cũ nếu có
cursor.execute("DROP TABLE IF EXISTS student")
cursor.execute("DROP TABLE IF EXISTS course")

# Tạo bảng
cursor.execute("""
CREATE TABLE student(
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

cursor.execute("""
CREATE TABLE course(
    id INTEGER,
    course_name TEXT
)
""")

# Dữ liệu
student_data = [
    (1, "Nguyen Minh Hoang", "May Tinh", 12, 6.7),
    (2, "Tran Thi Lan", "Kinh Te", 34, 9.2),
    (3, "Pham Van Nam", "Toan Tin", None, 7.9),
    (4, "Le Thanh Huyen", "Toan Tin", 20, 7.2),
    (5, "Vu Quoc Anh", "May Tinh", 24, 8.0),
    (6, "Dang Thuy Linh", "May Tinh", 24, 5.5),
    (7, "Bui Tien Dung", "Kinh Te", 34, 9.2),
    (8, "Ho Ngoc Mai", "Toan Tin", 20, 8.8),
    (9, "Duong Huu Phuc", "Kinh Te", None, 7.2),
    (10, "Cao Thi Hanh", "May Tinh", None, 7.0),
]

course_data = [
    (12, "Giai tich"),
    (34, "Thong ke"),
    (26, "Tin hoc"),
]

cursor.executemany("INSERT INTO student VALUES (?,?,?,?,?)", student_data)
cursor.executemany("INSERT INTO course VALUES (?,?)", course_data)

conn.commit()
```

```python
#CELL 2: Tích Descartes
df = pd.read_sql("SELECT * FROM student, course;", conn)
df
```

```python
#CELL 3: JOIN
pd.read_sql("""
SELECT *
FROM student s
INNER JOIN course c
ON s.course_id = c.id
""", conn)
```

```python
#LEFT JOIN
pd.read_sql("""
SELECT *
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
""", conn)
```

```python
#RIGHT JOIN (giả lập)
pd.read_sql("""
SELECT *
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
""", conn)
```

```python
#FULL JOIN (giả lập)
pd.read_sql("""
SELECT * FROM student s
LEFT JOIN course c ON s.course_id = c.id

UNION

SELECT * FROM course c
LEFT JOIN student s ON s.course_id = c.id
""", conn)
```

```python
#CELL 4: Cập nhật course_id NULL
cursor.execute("""
UPDATE student
SET course_id = (
    SELECT id FROM course ORDER BY RANDOM() LIMIT 1
)
WHERE course_id IS NULL
""")
conn.commit()
```

```python
#CELL 5: Thống kê theo lớp
pd.read_sql("""
SELECT class,
       COUNT(*) AS total_students,
       ROUND(AVG(score),2) AS avg_score
FROM student
GROUP BY class
""", conn)
```

```python
#CELL 6: Thống kê theo môn
pd.read_sql("""
SELECT c.course_name,
       COUNT(*) AS total_students,
       ROUND(AVG(s.score),2) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)
```

```python
#CELL 7: Phân loại môn
pd.read_sql("""
SELECT c.course_name,
       ROUND(AVG(s.score),2) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END AS rank
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)
```

```python
#CELL 8: Xếp hạng
#Theo điểm
pd.read_sql("""
SELECT *,
RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student
""", conn)
```

```python
#Theo lớp
pd.read_sql("""
SELECT *,
RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_class
FROM student
""", conn)
```

```python
#Theo môn
pd.read_sql("""
SELECT s.*, c.course_name,
RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_course
FROM student s
JOIN course c ON s.course_id = c.id
""", conn)
```

```python
#CELL 9: Top 3 cao nhất & thấp nhất
#Top cao
pd.read_sql("""
SELECT * FROM (
    SELECT *, RANK() OVER (ORDER BY score DESC) r
    FROM student
) WHERE r <= 3
""", conn)
```

```python
#Top thấp
pd.read_sql("""
SELECT * FROM (
    SELECT *, RANK() OVER (ORDER BY score ASC) r
    FROM student
) WHERE r <= 3
""", conn)
```

```python
#CELL 10: Thêm graduation_date
cursor.execute("ALTER TABLE student ADD COLUMN graduation_date TEXT")
conn.commit()
```

```python
#CELL 11: Cập nhật ngày tốt nghiệp
cursor.execute("""
UPDATE student
SET graduation_date = DATE('now', '+' || (
    SELECT COUNT(*)
    FROM student s2
    WHERE s2.score > student.score
) || ' days')
""")
conn.commit()
```
