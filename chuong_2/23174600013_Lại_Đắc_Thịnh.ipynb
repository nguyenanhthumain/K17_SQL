```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
```

```python
# bảng student
cursor.execute("""
CREATE TABLE student (
    student_id INT,
    name TEXT,
    class TEXT,
    course_id INT,
    score FLOAT
)
""")
```

```python
# Bảng course
cursor.execute("""
CREATE TABLE course (
    id INT,
    course_name TEXT
)
""")
```

Cau 1: JOIN

```python
# Tích decartes
pd.read_sql("""
SELECT * FROM student, course
""", conn)
```

```python
# Inner Join
pd.read_sql("""
SELECT * 
FROM student s
INNER JOIN course c 
ON s.course_id = c.id
""", conn)
```

```python
# Left Join
pd.read_sql("""
SELECT * 
FROM student s
LEFT JOIN course c 
ON s.course_id = c.id
""", conn)
```

```python
# Right Join
pd.read_sql("""
SELECT * 
FROM course c
LEFT JOIN student s 
ON s.course_id = c.id
""", conn)
```

```python
# Full Join
pd.read_sql("""
SELECT * 
FROM student s
LEFT JOIN course c ON s.course_id = c.id

UNION

SELECT * 
FROM course c
LEFT JOIN student s ON s.course_id = c.id
""", conn)
```

```python
# Cập nhật course_id NULL
cursor.execute("""
UPDATE student
SET course_id = 20
WHERE course_id IS NULL
""")
```

```python
# Xóa bản ghi course không tồn tại
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")
conn.commit()
```

```python
# a. Theo lớp
pd.read_sql("""
SELECT class,
COUNT(*) AS total_student,
ROUND(AVG(score),2) AS avg_score
FROM student
GROUP BY class
""", conn)
```

```python
# b. Theo môn học
pd.read_sql("""
SELECT c.course_name,
COUNT(*) AS total_student,
ROUND(AVG(score),2) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)
```

```python
# c. Phân loại thi đua
pd.read_sql("""
SELECT c.course_name,
ROUND(AVG(score),2) AS avg_score,
CASE
    WHEN AVG(score) >= 9 THEN 'Xuat sac'
    WHEN AVG(score) >= 6 THEN 'Tot'
    ELSE 'Kem'
END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)
```

Cau 3:Xep Hang

```python
# a. Theo điểm
pd.read_sql("""
SELECT *,
RANK() OVER (ORDER BY score DESC) AS rank
FROM student
""", conn)
```

```python
# b. Theo lớp
pd.read_sql("""
SELECT *,
RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank
FROM student
""", conn)
```

```python
# c. Theo môn học
pd.read_sql("""
SELECT *,
RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank
FROM student
""", conn)
```

```python
# Top 3 cao nhất
pd.read_sql("""
SELECT * FROM (
    SELECT *, RANK() OVER (ORDER BY score DESC) AS rank
    FROM student
)
WHERE rank <= 3
""", conn)
```

```python
# Top 3 thấp nhất
pd.read_sql("""
SELECT * FROM (
    SELECT *, RANK() OVER (ORDER BY score ASC) AS rank
    FROM student
)
WHERE rank <= 3
""", conn)
```
