```python
# Câu 1
#  1. Tạo dữ liệu 2 bảng
import pandas as pd
import sqlite3

# Kết nối SQLite trong bộ nhớ
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

# Tạo bảng student
cursor.execute("""
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

# Tạo bảng course
cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")

# Chèn dữ liệu vào bảng student
students = [
    (1, "Nguyen Minh Hoang", "May Tinh", 12, 6.7),
    (2, "Tran Thi Lan", "Kinh Te", 34, 9.2),
    (3, "Pham Van Nam", "Toan Tin", None, 7.9),
    (4, "Le Thanh Huyen", "Toan Tin", 20, 7.2),
    (5, "Vu Quoc Anh", "May Tinh", 24, 8.0),
    (6, "Dang Thuy Linh", "May Tinh", 24, 5.5),
    (7, "Bui Tien Dung", "Kinh Te", 34, 9.2),
    (8, "Ho Ngoc Mai", "Toan Tin", 20, 8.8),
    (9, "Duong Huu Phuc", "Kinh Te", None, 7.2),
    (10, "Cao Thi Hanh", "May Tinh", None, 7.0)
]

cursor.executemany("INSERT INTO student VALUES (?, ?, ?, ?, ?)", students)

# Chèn dữ liệu vào bảng course
courses = [
    (12, "Giai tich"),
    (34, "Thong ke"),
    (26, "Tin hoc")
]

cursor.executemany("INSERT INTO course VALUES (?, ?)", courses)

conn.commit()
```

```python
# Hiển thị 2 bảng gốc
print("Bảng student:")
display(pd.read_sql("SELECT * FROM student", conn))

print("Bảng course:")
display(pd.read_sql("SELECT * FROM course", conn))
```

```python
# Tích Descartes (CROSS JOIN)
query_cross = """
SELECT *
FROM student
CROSS JOIN course
"""
df_cross = pd.read_sql(query_cross, conn)
display(df_cross)
print("Số dòng tích Descartes:", len(df_cross))
```

```python
# INNER JOIN
query_inner = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM student s
INNER JOIN course c
ON s.course_id = c.id
"""
df_inner = pd.read_sql(query_inner, conn)
display(df_inner)
```

```python
# LEFT JOIN
query_left = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
"""
df_left = pd.read_sql(query_left, conn)
display(df_left)
```

```python
# RIGHT JOIN
query_right = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""
df_right = pd.read_sql(query_right, conn)
display(df_right)
```

```python
# FULL OUTER JOIN
query_full = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.id, c.course_name
FROM student s
LEFT JOIN course c
ON s.course_id = c.id

UNION

SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.id, c.course_name
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""
df_full = pd.read_sql(query_full, conn)
display(df_full)
```

Kết luận câu 1:

* Đã thực hiện kết nối hai bảng student và course bằng các cách: tích Descartes, INNER JOIN, LEFT JOIN, RIGHT JOIN mô phỏng và FULL OUTER JOIN mô phỏng.
* Kết quả CROSS JOIN cho ra tất cả các tổ hợp giữa dữ liệu của hai bảng.
* INNER JOIN chỉ giữ lại các bản ghi có course_id của sinh viên trùng với id trong bảng course.
* LEFT JOIN giữ toàn bộ dữ liệu của bảng student, kể cả những sinh viên chưa có course_id hoặc có course_id không tồn tại trong bảng course.
* RIGHT JOIN mô phỏng giữ toàn bộ dữ liệu của bảng course, nên có thể thấy môn Tin hoc vẫn xuất hiện dù chưa có sinh viên nào khớp.
* FULL OUTER JOIN mô phỏng cho thấy đầy đủ dữ liệu từ cả hai bảng; các giá trị NaN hoặc None xuất hiện tại những vị trí không có bản ghi tương ứng giữa hai bảng.
* Như vậy, câu 1 đã cho thấy sự khác nhau giữa từng kiểu nối bảng và giúp xác định các bản ghi khớp hoặc không khớp giữa student và course

```python
# 2. Loại bỏ các bản ghi có course_id không tồn tại trong bảng course
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")
conn.commit()

print("Sau khi xóa các bản ghi có course_id không tồn tại trong course:")
display(pd.read_sql("SELECT * FROM student", conn))
```

```python
# 3. Tổng số sinh viên và điểm trung bình theo từng lớp
query_class = """
SELECT 
    class,
    COUNT(*) AS tong_so_sinh_vien,
    ROUND(AVG(score), 2) AS diem_trung_binh
FROM student
GROUP BY class
ORDER BY class
"""
df_class = pd.read_sql(query_class, conn)
print("Tổng số sinh viên và điểm trung bình theo từng lớp:")
display(df_class)
```

```python
# 4. Tổng số sinh viên và điểm trung bình theo từng môn học
query_course = """
SELECT 
    c.id AS course_id,
    c.course_name,
    COUNT(s.student_id) AS tong_so_sinh_vien,
    ROUND(AVG(s.score), 2) AS diem_trung_binh
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
GROUP BY c.id, c.course_name
ORDER BY c.id
"""
df_course = pd.read_sql(query_course, conn)
print("Tổng số sinh viên và điểm trung bình theo từng môn học:")
display(df_course)
```

```python
# 5. Phân loại từng môn học theo điểm trung bình
query_rank_course = """
SELECT 
    c.id AS course_id,
    c.course_name,
    COUNT(s.student_id) AS tong_so_sinh_vien,
    ROUND(AVG(s.score), 2) AS diem_trung_binh,
    CASE
        WHEN AVG(s.score) >= 9.0 THEN 'Xuat sac'
        WHEN AVG(s.score) >= 6.0 AND AVG(s.score) <= 8.9 THEN 'Tot'
        ELSE 'Kem'
    END AS xep_loai
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
GROUP BY c.id, c.course_name
ORDER BY c.id
"""
df_rank_course = pd.read_sql(query_rank_course, conn)
print("Phân loại môn học theo điểm trung bình:")
display(df_rank_course)
```

```python
# Câu 3
# 1. Xếp hạng sinh viên theo điểm số toàn bộ
query_rank_all = """
SELECT
    student_id,
    name,
    class,
    course_id,
    score,
    RANK() OVER (ORDER BY score DESC) AS ranking
FROM student
"""
df_rank_all = pd.read_sql(query_rank_all, conn)
print("Xếp hạng sinh viên theo điểm số toàn bộ:")
display(df_rank_all)
```

```python
# 2. Top 3 sinh viên điểm cao nhất toàn bộ
query_top3_all = """
SELECT *
FROM (
    SELECT
        student_id,
        name,
        class,
        course_id,
        score,
        RANK() OVER (ORDER BY score DESC) AS ranking
    FROM student
)
WHERE ranking <= 3
"""
df_top3_all = pd.read_sql(query_top3_all, conn)
print("Top 3 sinh viên điểm cao nhất toàn bộ:")
display(df_top3_all)
```

```python
# 3. Top 3 sinh viên điểm thấp nhất toàn bộ
query_bottom3_all = """
SELECT *
FROM (
    SELECT
        student_id,
        name,
        class,
        course_id,
        score,
        RANK() OVER (ORDER BY score ASC) AS ranking
    FROM student
)
WHERE ranking <= 3
"""
df_bottom3_all = pd.read_sql(query_bottom3_all, conn)
print("Top 3 sinh viên điểm thấp nhất toàn bộ:")
display(df_bottom3_all)
```

```python
# 4. Xếp hạng sinh viên theo điểm số trong từng lớp
query_rank_class = """
SELECT
    student_id,
    name,
    class,
    course_id,
    score,
    RANK() OVER (PARTITION BY class ORDER BY score DESC) AS ranking_in_class
FROM student
ORDER BY class, ranking_in_class
"""
df_rank_class = pd.read_sql(query_rank_class, conn)
print("Xếp hạng sinh viên theo từng lớp:")
display(df_rank_class)
```

```python
# 5. Top 3 cao nhất theo từng lớp
query_top3_class = """
SELECT *
FROM (
    SELECT
        student_id,
        name,
        class,
        course_id,
        score,
        RANK() OVER (PARTITION BY class ORDER BY score DESC) AS ranking_in_class
    FROM student
)
WHERE ranking_in_class <= 3
ORDER BY class, ranking_in_class
"""
df_top3_class = pd.read_sql(query_top3_class, conn)
print("Top 3 cao nhất theo từng lớp:")
display(df_top3_class)
```

```python
# Câu 4
# 1. Thêm cột graduation_date
cursor.execute("""
ALTER TABLE student
ADD COLUMN graduation_date DATETIME
""")
conn.commit()

print("Sau khi thêm cột graduation_date:")
display(pd.read_sql("SELECT * FROM student", conn))
```

```python
# 2. Xếp hạng sinh viên theo điểm số
query_rank = """
SELECT
    student_id,
    name,
    score,
    RANK() OVER (ORDER BY score DESC) AS ranking
FROM student
"""
df_rank = pd.read_sql(query_rank, conn)
print("Bảng xếp hạng sinh viên theo điểm số:")
display(df_rank)
```

```python
# 3. Cập nhật graduation_date theo hạng
cursor.execute("""
UPDATE student
SET graduation_date = (
    SELECT datetime('now', '+' || ranking || ' days')
    FROM (
        SELECT
            student_id,
            RANK() OVER (ORDER BY score DESC) AS ranking
        FROM student
    ) AS ranked
    WHERE ranked.student_id = student.student_id
)
""")
conn.commit()
```

```python
# 4. Hiển thị kết quả cuối cùng
query_result = """
SELECT
    student_id,
    name,
    class,
    course_id,
    score,
    graduation_date
FROM student
ORDER BY score DESC
"""
df_result = pd.read_sql(query_result, conn)
print("Bảng student sau khi cập nhật graduation_date:")
display(df_result)
```

```python
# 5. Có thể xem kèm thứ hạng để dễ giải thích
query_final = """
SELECT
    student_id,
    name,
    class,
    course_id,
    score,
    RANK() OVER (ORDER BY score DESC) AS ranking,
    graduation_date
FROM student
ORDER BY ranking, student_id
"""
df_final = pd.read_sql(query_final, conn)
print("Kết quả cuối cùng kèm thứ hạng:")
display(df_final)
```

Kết luận câu 4
* Bảng student đã được bổ sung thêm trường graduation_date với kiểu dữ liệu DATETIME.
* Thời gian tốt nghiệp của mỗi sinh viên được xác định bằng thời điểm hiện tại cộng với số hạng tương ứng của sinh viên đó trong bảng xếp hạng theo điểm số.
* Hai sinh viên có điểm cao nhất là Tran Thi Lan và Bui Tien Dung cùng hạng 1, nên có cùng graduation_date sớm nhất.
* Sinh viên có thứ hạng thấp hơn sẽ có graduation_date muộn hơn theo đúng quy tắc đề bài đã yêu cầu.
