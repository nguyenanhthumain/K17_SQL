#Câu 1

```python
import sqlite3
import pandas as pd

# 1. Tạo kết nối tới cơ sở dữ liệu SQLite trong bộ nhớ
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# 2. Tạo bảng course
cursor.execute('''
CREATE TABLE course (
    id INTEGER PRIMARY KEY,
    course_name TEXT
)
''')

# 3. Tạo bảng student
cursor.execute('''
CREATE TABLE student (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
''')

# 4. Chèn dữ liệu vào bảng course
course_data = [
    (12, 'Giai tich'),
    (34, 'Thong ke'),
    (26, 'Tin hoc')
]
cursor.executemany('INSERT INTO course VALUES (?, ?)', course_data)

# 5. Chèn dữ liệu vào bảng student
student_data = [
    (1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
    (2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
    (3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
    (4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
    (5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
    (6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
    (7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
    (8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
    (9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
    (10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
]
cursor.executemany('INSERT INTO student VALUES (?, ?, ?, ?, ?)', student_data)
conn.commit()

# Hàm hỗ trợ hiển thị kết quả truy vấn SQL bằng Pandas
def run_query(query):
    return pd.read_sql_query(query, conn)

print("Đã khởi tạo dữ liệu thành công!")
```

```python
# Tích Decartes (Cross Join) - Kết nối mọi dòng của bảng này với mọi dòng của bảng kia
query_cross = """
SELECT s.*, c.* FROM student s
CROSS JOIN course c;
"""
display(run_query(query_cross))

# INNER JOIN - Chỉ lấy những sinh viên có course_id tồn tại trong bảng course
query_inner = """
SELECT s.*, c.course_name
FROM student s
INNER JOIN course c ON s.course_id = c.id;
"""
display(run_query(query_inner))

# LEFT JOIN - Lấy tất cả sinh viên, nếu không có môn học tương ứng thì để NULL
query_left = """
SELECT s.*, c.course_name
FROM student s
LEFT JOIN course c ON s.course_id = c.id;
"""
display(run_query(query_left))

# RIGHT JOIN - Lấy tất cả môn học, môn nào chưa có sinh viên sẽ để NULL
query_right = """
SELECT s.*, c.course_name
FROM student s
RIGHT JOIN course c ON s.course_id = c.id;
"""
display(run_query(query_right))

# FULL OUTER JOIN - Lấy tất cả dữ liệu từ cả 2 bảng
query_full = """
SELECT s.*, c.course_name
FROM student s
FULL OUTER JOIN course c ON s.course_id = c.id;
"""
display(run_query(query_full))
```

```python
cursor.execute("""
UPDATE student
SET course_id = 26
WHERE course_id IS NULL;
""")

# Xóa những bản ghi tham gia môn học không tồn tại trong bảng course
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course);
""")
conn.commit()

print("Dữ liệu bảng student sau khi Cập nhật và Xóa:")
display(run_query("SELECT * FROM student"))
```

```python
# a. Tổng số sinh viên, điểm trung bình của từng lớp
query_2a = """
SELECT class, COUNT(student_id) as total_students, ROUND(AVG(score), 2) as avg_score
FROM student
GROUP BY class;
"""
display(run_query(query_2a))

# b. Tổng số sinh viên, điểm trung bình của từng môn học
query_2b = """
SELECT c.course_name, COUNT(s.student_id) as total_students, ROUND(AVG(s.score), 2) as avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name;
"""
display(run_query(query_2b))

# c. Phân loại thi đua theo điểm số
query_2c = """
SELECT *,
    CASE
        WHEN score >= 9.0 THEN 'Xuất sắc'
        WHEN score >= 6.0 AND score <= 8.9 THEN 'Tốt'
        ELSE 'Kém'
    END as classification
FROM student;
"""
display(run_query(query_2c))
```

#Câu 3

```python
# Xếp hạng tổng quát theo điểm số
query_rank_score = """
SELECT student_id, name, score,
       RANK() OVER(ORDER BY score DESC) as rank_by_score,
       RANK() OVER(PARTITION BY class ORDER BY score DESC) as rank_by_class,
       RANK() OVER(PARTITION BY course_id ORDER BY score DESC) as rank_by_course
FROM student;
"""
display(run_query(query_rank_score))

# Tìm TOP 3 cao nhất và TOP 3 thấp nhất (Ví dụ theo điểm số tổng quát)
query_top_bottom = """
WITH RankedStudents AS (
    SELECT student_id, name, score,
           RANK() OVER(ORDER BY score DESC) as rank_desc,
           RANK() OVER(ORDER BY score ASC) as rank_asc
    FROM student
)
SELECT name, score, 'Top 3 Cao Nhất' as category
FROM RankedStudents WHERE rank_desc <= 3
UNION ALL
SELECT name, score, 'Top 3 Thấp Nhất' as category
FROM RankedStudents WHERE rank_asc <= 3;
"""
display(run_query(query_top_bottom))
```

```python
# 1. Thêm cột graduation_date
try:
    cursor.execute("ALTER TABLE student ADD COLUMN graduation_date DATETIME;")
except sqlite3.OperationalError:
    pass # Bỏ qua nếu cột đã tồn tại (chạy lại cell nhiều lần)

# 2. Cập nhật graduation_date = Thời gian hiện tại + X ngày (với X là thứ hạng điểm số)
cursor.execute("""
WITH Ranked AS (
    SELECT student_id, RANK() OVER(ORDER BY score DESC) as rank_val
    FROM student
)
UPDATE student
SET graduation_date = (
    SELECT datetime('now', '+' || r.rank_val || ' days')
    FROM Ranked r
    WHERE r.student_id = student.student_id
);
""")
conn.commit()

# Xem kết quả cuối cùng
display(run_query("SELECT student_id, name, score, graduation_date FROM student ORDER BY score DESC;"))
```
