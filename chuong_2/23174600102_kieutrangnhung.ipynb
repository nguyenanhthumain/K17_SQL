## **1. Kết nối hai bảng**

### **Sử dụng tích Decartes**

```python
query1 = """
SELECT *
FROM students
CROSS JOIN course
"""
cursor.execute(query1)
rows = cursor.fetchall()

for row in rows:
    print(row)
```

### **Sử dụng JOIN**

#### **INNER JOIN**

```python
query2 = """
SELECT *
FROM students s
INNER JOIN course c
ON s.course_id = c.id
"""

cursor.execute(query2)
rows = cursor.fetchall()

for row in rows:
    print(row)
```

#### **LEFT JOIN**

```python
query3 = """
SELECT *
FROM students s
LEFT JOIN course c
ON s.course_id = c.id
"""

cursor.execute(query3)
rows = cursor.fetchall()

for row in rows:
    print(row)
```

#### **RIGHT JOIN**

```python
query4 = """
SELECT *
FROM course c
LEFT JOIN students s
ON s.course_id = c.id
"""

cursor.execute(query4)
rows = cursor.fetchall()

for row in rows:
    print(row)
```

#### **FULL OUTER JOIN**

```python
query5 = """
SELECT s.name, s.class, s.score, c.course_name
FROM students s
LEFT JOIN course c
ON s.course_id = c.id

UNION

SELECT s.name, s.class, s.score, c.course_name
FROM course c
LEFT JOIN students s
ON s.course_id = c.id
"""

cursor.execute(query5)

rows = cursor.fetchall()

for row in rows:
    print(row)
```

#### **a. Tổng số sinh viên và điểm trung bình từng lớp**

```python
query6 = """
SELECT class,
       COUNT(student_id) AS total_students,
       ROUND(AVG(score),2) AS avg_score
FROM students
GROUP BY class
"""

cursor.execute(query6)
rows = cursor.fetchall()

for row in rows:
    print(row)
```

#### **b. Tổng số sinh viên, điểm trung bình của từng môn học**

```python
query7 = """
SELECT c.course_name,
       COUNT(s.student_id) AS total_students,
       ROUND(AVG(s.score), 2) AS avg_score
FROM course c
LEFT JOIN students s
ON s.course_id = c.id
GROUP BY c.course_name
"""

cursor.execute(query7)
rows = cursor.fetchall()

for row in rows:
    print(row)
```

#### **c. Phân loại thi đua theo số điểm của từng môn học biết:**
- Điểm TB >= 9.0: Xuất sắc.
- 6.0 <= Điểm TB <= 8.9: Tốt
- Điểm TB < 6.0: Kém

```python
query8 = """
SELECT c.course_name,
       ROUND(AVG(s.score), 2) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9.0 THEN 'Xuất sắc'
           WHEN AVG(s.score) >= 6.0 THEN 'Tốt'
           ELSE 'Kém'
       END AS classification
FROM course c
LEFT JOIN students s
ON s.course_id = c.id
GROUP BY c.course_name
"""

cursor.execute(query8)
rows = cursor.fetchall()

for row in rows:
    print(row)
```

## **4. Thêm thời gian tốt nghiệp**

```python
# 1) Thêm cột mới
cursor.execute("""
ALTER TABLE students
ADD COLUMN graduation_date DATETIME
""")

# 2) Cập nhật ngày tốt nghiệp theo thứ hạng điểm
cursor.execute("""
WITH ranked_students AS (
    SELECT student_id,
           ROW_NUMBER() OVER (ORDER BY score DESC) AS ranking
    FROM students
)
UPDATE students
SET graduation_date = DATETIME(
    'now',
    '+' || (
        SELECT ranking
        FROM ranked_students
        WHERE ranked_students.student_id = students.student_id
    ) || ' days'
)
""")
conn.commit()
```
