# Tạo database + bảng

```python
import sqlite3

conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

# Tạo bảng student
cursor.execute("""
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

# Tạo bảng course
cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")
```

```python
import pandas as pd
from IPython.display import display

def show(cursor):
    """Hien thi ket qua query dang bang dep voi pandas."""
    rows = cursor.fetchall()
    if not rows:
        print("(Khong co du lieu)")
        return
    cols = [desc[0] for desc in cursor.description]
    df = pd.DataFrame(rows, columns=cols)
    display(df)
```

# Câu 1

nối 2 bảng theo tích decarts

```python
# nối 2 bảng theo tích decarts
cursor.execute("""
SELECT * FROM student, course
""")
show(cursor)
```

INNER JOIN

```python
# INNER JOIN
cursor.execute("""
SELECT * FROM student s
INNER JOIN course c
ON s.course_id = c.id
""")
show(cursor)
```

LEFT JOIN

```python
# LEFT JOIN
cursor.execute("""
SELECT * FROM student s
LEFT JOIN course c
ON s.course_id = c.id
""")
show(cursor)
```

Right JOIN

```python
# Right JOIN
cursor.execute("""
SELECT * FROM course c
RIGHT JOIN student s
ON s.course_id = c.id
""")
show(cursor)
```

FULL OUTER JOIN

```python
# FULL OUTER JOIN
cursor.execute("""
SELECT * FROM student s
LEFT JOIN course c ON s.course_id = c.id

UNION

SELECT * FROM student s
RIGHT JOIN course c ON s.course_id = c.id
""")

show(cursor)
```

# Câu 2

Cập nhật giá trị thiếu và loại bỏ bản ghi không tồn tại

```python
# 1. UPDATE trước
cursor.execute("""
UPDATE student
SET course_id = 26
WHERE course_id IS NULL
""")

# 2. DELETE sau
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")

conn.commit()
```

a. Tổng số sinh viên, điểm trung bình của từng lớp

```python
cursor.execute("""
SELECT class,
       COUNT(*) AS total_students,
       AVG(score) AS avg_score
FROM student
GROUP BY class
""")
show(cursor)
```

b. Tổng số sinh viên, điểm trung bình của từng môn học.

```python
cursor.execute("""
SELECT c.course_name,
       COUNT(*) AS total_students,
       AVG(s.score) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""")
show(cursor)
```

c. Phân loại thi đua theo số điểm của từng môn học biết:

```python
cursor.execute("""
SELECT c.course_name,
       AVG(s.score) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""")
show(cursor)
```

# BÀI 3: Xếp hạng

# xếp hạng qua điểm số

```python
# xếp hạng qua điểm số
cursor.execute("""
SELECT *,
       RANK() OVER (ORDER BY score DESC) AS rank_overall
FROM student
""")
```

```python
cursor.execute("""
SELECT * FROM (
    SELECT *,
           RANK() OVER (ORDER BY score DESC) AS r
    FROM student
)
WHERE r <= 3
""")
print("top 3 sinh viên đạt thứ hạng cao nhất theo điểm số")
show(cursor)
```

```python
cursor.execute("""
SELECT * FROM (
    SELECT *,
           RANK() OVER (ORDER BY score ASC) AS r
    FROM student
)
WHERE r <= 3
""")
print("top 3 sinh viên đạt thứ hạng thấp nhất theo điểm số")
show(cursor)
```
