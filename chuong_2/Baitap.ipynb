b. INNER JOIN

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('university.db')

# Execute the SQL query
query = """SELECT *
FROM student s
INNER JOIN course c
ON s.course_id = c.id;"""
df_inner_join = pd.read_sql_query(query, conn)

# Display the results
display(df_inner_join)

# Close the connection
conn.close()
```

c. LEFT JOIN

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('university.db')

# Execute the SQL query with LEFT JOIN
query = """SELECT *
FROM student s
LEFT JOIN course c
ON s.course_id = c.id;"""
df_left_join = pd.read_sql_query(query, conn)

# Display the results
display(df_left_join)

# Close the connection
conn.close()
```

d. RIGHT JOIN

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('university.db')

# Execute the SQL query with RIGHT JOIN
query = """SELECT *
FROM student s
RIGHT JOIN course c
ON s.course_id = c.id;"""

# Rewriting the RIGHT JOIN as a LEFT JOIN by swapping tables
query_rewritten = """SELECT 
    s.student_id, s.name, s.class, s.course_id, s.score,
    c.id, c.course_name
FROM course c
LEFT JOIN student s
ON s.course_id = c.id;"""

df_right_join = pd.read_sql_query(query_rewritten, conn)

# Display the results
display(df_right_join)

# Close the connection
conn.close()
```

e. FULL OUTER JOIN

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('university.db')

# Query for the LEFT JOIN part
# Explicitly select and alias columns for consistency when using UNION
query_left_join = """
SELECT
    s.student_id,
    s.name,
    s.class,
    s.course_id AS student_course_id, -- Alias to distinguish from course.id
    s.score,
    c.id AS course_id,
    c.course_name
FROM student s
LEFT JOIN course c ON s.course_id = c.id;
"""
df_left = pd.read_sql_query(query_left_join, conn)

# Query for the RIGHT JOIN part (rewritten as a LEFT JOIN from the 'right' table to the 'left' table)
# Ensure columns are selected and aliased in the same order as the LEFT JOIN query for UNION compatibility
query_right_join_rewritten = """
SELECT
    s.student_id,
    s.name,
    s.class,
    s.course_id AS student_course_id,
    s.score,
    c.id AS course_id,
    c.course_name
FROM course c
LEFT JOIN student s ON c.id = s.course_id;
"""
df_right = pd.read_sql_query(query_right_join_rewritten, conn)

# Combine the results of both joins using pd.concat and drop_duplicates to simulate UNION
df_full_outer_join = pd.concat([df_left, df_right]).drop_duplicates().reset_index(drop=True)

# Display the results
display(df_full_outer_join)

# Close the connection
conn.close()
```

2. Cập nhật course_id & loại bỏ dữ liệu sai

b. Xóa bản ghi course_id không tồn tại

a. Tổng SV + điểm TB theo lớp

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('university.db')

# Execute the SQL query to get class-wise statistics
query = """SELECT class,
       COUNT(*) AS total_students,
       AVG(score) AS avg_score
FROM student
GROUP BY class;"""
df_class_stats = pd.read_sql_query(query, conn)

# Display the results
display(df_class_stats)

# Close the connection
conn.close()
```

b. Tổng SV + điểm TB theo môn

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('university.db')

# Execute the SQL query to get course-wise statistics
query = """SELECT c.course_name,
       COUNT(s.student_id) AS total_students,
       AVG(s.score) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name;"""
df_course_stats = pd.read_sql_query(query, conn)

# Display the results
display(df_course_stats)

# Close the connection
conn.close()
```

c. Phân loại theo điểm TB môn học

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('university.db')

# Execute the SQL query to get course-wise classification
query = """SELECT c.course_name,
       AVG(s.score) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name;"""
df_course_classification = pd.read_sql_query(query, conn)

# Display the results
display(df_course_classification)

# Close the connection
conn.close()
```

3. Xếp hạng sinh viên

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('university.db')

# Execute the SQL query to rank students by score
query = """SELECT *,
       RANK() OVER (ORDER BY score DESC) AS ranking
FROM student;"""
df_student_ranking = pd.read_sql_query(query, conn)

# Display the results
display(df_student_ranking)

# Close the connection
conn.close()
```

```python
import sqlite3
import pandas as pd

# Connect to the database
conn = sqlite3.connect('university.db')

# Execute the SQL query to rank students by score within each class
query = """SELECT *,
       RANK() OVER (PARTITION BY class ORDER BY score DESC) AS ranking
FROM student;"""
df_class_ranking = pd.read_sql_query(query, conn)

# Display the results
display(df_class_ranking)

# Close the connection
conn.close()
```

Cập nhật theo thứ hạng

```python
import sqlite3
import pandas as pd
from datetime import datetime, timedelta

# Connect to the database
conn = sqlite3.connect('university.db')
cursor = conn.cursor()

# Get students with their ranks
query_ranked_students = """
SELECT student_id, score, RANK() OVER (ORDER BY score DESC) AS rnk
FROM student;
"""
ranked_students_df = pd.read_sql_query(query_ranked_students, conn)

# Calculate graduation dates
current_date = datetime.now()
updates = []
for index, row in ranked_students_df.iterrows():
    student_id = row['student_id']
    rank = row['rnk']
    # Add rank days to the current date
    # SQLite datetime format is 'YYYY-MM-DD HH:MM:SS'
    graduation_date = current_date + timedelta(days=int(rank))
    updates.append((graduation_date.strftime('%Y-%m-%d %H:%M:%S'), student_id))

# Update the graduation_date column for each student
update_query = "UPDATE student SET graduation_date = ? WHERE student_id = ?;"
cursor.executemany(update_query, updates)

# Commit the changes and close the connection
conn.commit()
conn.close()

print("Graduation dates updated successfully based on student rankings.")
```
