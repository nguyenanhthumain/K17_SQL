```python
import sqlite3
import pandas as pd
```

```python
connection = sqlite3.connect('oanh.db')
cursor = connection.cursor()
```

```python
cursor.execute("""
CREATE TABLE student (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

# Tạo bảng course
cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")
```

```python
# Kết hợp 
query_inner_join = """
SELECT student.student_id,
       student.name,
       student.class,
       student.course_id,
       student.score,
       course.course_name
FROM student
INNER JOIN course
ON student.course_id = course.id
"""

df_inner = pd.read_sql_query(query_inner_join, connection)
print("INNER JOIN:")
display(df_inner)
```

```python
query_left_join = """
SELECT student.student_id,
       student.name,
       student.class,
       student.course_id,
       student.score,
       course.course_name
FROM student
LEFT JOIN course
ON student.course_id = course.id
"""

df_left = pd.read_sql_query(query_left_join, connection)
print("LEFT JOIN:")
display(df_left)
```

```python
query_right_join = """
SELECT student.student_id,
       student.name,
       student.class,
       student.course_id,
       student.score,
       course.course_name
FROM course
LEFT JOIN student
ON student.course_id = course.id
"""

df_right = pd.read_sql_query(query_right_join, connection)
print("RIGHT JOIN (giả lập):")
display(df_right)
```

```python
query_full_join = """
SELECT student.student_id,
       student.name,
       student.class,
       student.course_id,
       student.score,
       course.course_name
FROM student
LEFT JOIN course
ON student.course_id = course.id

UNION

SELECT student.student_id,
       student.name,
       student.class,
       student.course_id,
       student.score,
       course.course_name
FROM course
LEFT JOIN student
ON student.course_id = course.id
"""

df_full = pd.read_sql_query(query_full_join, connection)
print("FULL OUTER JOIN (giả lập):")
display(df_full)
```

```python
query_a = """
SELECT class,
       COUNT(student_id) AS tong_sinh_vien,
       AVG(score) AS diem_trung_binh
FROM student
GROUP BY class
"""

df_a = pd.read_sql_query(query_a, connection)
display(df_a)
```

```python
query_b = """
SELECT course.course_name,
       COUNT(student.student_id) AS tong_sinh_vien,
       AVG(student.score) AS diem_trung_binh
FROM student
INNER JOIN course
ON student.course_id = course.id
GROUP BY course.course_name
"""

df_b = pd.read_sql_query(query_b, connection)
display(df_b)
```

```python
query_c = """
SELECT course.course_name,
       AVG(student.score) AS diem_trung_binh,
       CASE
           WHEN AVG(student.score) >= 9.0 THEN 'Xuat sac'
           WHEN AVG(student.score) >= 6.0 THEN 'Tot'
           ELSE 'Kem'
       END AS xep_loai
FROM student
INNER JOIN course
ON student.course_id = course.id
GROUP BY course.course_name
"""

df_c = pd.read_sql_query(query_c, connection)
display(df_c)
```

```python
# xếp theo điểm
query_rank_all = """
SELECT student_id,
       name,
       class,
       score,
       RANK() OVER (ORDER BY score DESC) AS thu_hang
FROM student
"""

df_rank_all = pd.read_sql_query(query_rank_all, connection)
display(df_rank_all)
```

```python
query_top3 = """
SELECT *
FROM (
    SELECT student_id,
           name,
           class,
           score,
           RANK() OVER (ORDER BY score DESC) AS thu_hang
    FROM student
)
WHERE thu_hang <= 3
"""

df_top3 = pd.read_sql_query(query_top3, connection)
display(df_top3)
```

```python
query_bottom3 = """
SELECT *
FROM (
    SELECT student_id,
           name,
           class,
           score,
           RANK() OVER (ORDER BY score ASC) AS thu_hang
    FROM student
)
WHERE thu_hang <= 3
"""

df_bottom3 = pd.read_sql_query(query_bottom3, connection)
display(df_bottom3)
```

```python
# xếp theo lớp
query_rank_class = """
SELECT student_id,
       name,
       class,
       score,
       RANK() OVER (PARTITION BY class ORDER BY score DESC) AS thu_hang
FROM student
"""

df_rank_class = pd.read_sql_query(query_rank_class, connection)
display(df_rank_class)
```
