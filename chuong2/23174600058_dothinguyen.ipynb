1. Hãy kết nối hai bảng trên theo những cách sau: 

Sử dụng tích Decartes. 

Sử dụng JOIN: INNER JOIN, LEFT JOIN, RIGHT JOIN, FULL OUTER JOIN.

```python
import sqlite3

conn = sqlite3.connect(":memory:")
cursor = conn.cursor()
cursor.execute("""
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")


cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")
```

```python
import pandas as pd
from IPython.display import display

def show(cursor):
    """Hien thi ket qua query dang bang dep voi pandas."""
    rows = cursor.fetchall()
    if not rows:
        print("(Khong co du lieu)")
        return
    cols = [desc[0] for desc in cursor.description]
    df = pd.DataFrame(rows, columns=cols)
    display(df)
```

```python
# Insert student
students = [
(1,"Nguyen Minh Hoang","May Tinh",12,6.7),
(2,"Tran Thi Lan","Kinh Te",34,9.2),
(3,"Pham Van Nam","Toan Tin",None,7.9),
(4,"Le Thanh Huyen","Toan Tin",20,7.2),
(5,"Vu Quoc Anh","May Tinh",24,8.0),
(6,"Dang Thuy Linh","May Tinh",24,5.5),
(7,"Bui Tien Dung","Kinh Te",34,9.2),
(8,"Ho Ngoc Mai","Toan Tin",20,8.8),
(9,"Duong Huu Phuc","Kinh Te",None,7.2),
(10,"Cao Thi Hanh","May Tinh",None,7.0)
]

cursor.executemany("INSERT INTO student VALUES (?,?,?,?,?)", students)

# Insert course
courses = [
(12,"Giai tich"),
(34,"Thong ke"),
(26,"Tin hoc")
]

cursor.executemany("INSERT INTO course VALUES (?,?)", courses)

conn.commit()
```

```python
# nối 2 bảng theo tích decarts
cursor.execute("""
SELECT * 
FROM student, course
""")
show(cursor)
```

```python
# INNER JOIN
cursor.execute("""
SELECT *
FROM student s
INNER JOIN course c
ON s.course_id = c.id
""")
show(cursor)
```

```python
# LEFT JOIN
cursor.execute("""
SELECT *
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
""")
show(cursor)
```

```python
# Right JOIN
cursor.execute("""
SELECT * 
FROM course c
RIGHT JOIN student s
ON s.course_id = c.id
""")
show(cursor)
```

```python
# FULL OUTER JOIN
cursor.execute("""
SELECT *
FROM student s
LEFT JOIN course c ON s.course_id = c.id

UNION

SELECT *
FROM course c
LEFT JOIN student s ON s.course_id = c.id
""")

show(cursor)
```

2. Hãy cập nhật những giá trị course_id còn thiếu trong bảng student bằng câu lệnh SQL, 
trong đó các giá trị được điền là những giá trị nằm trong bảng course  và loại bỏ những 
bản ghi tham gia những môn học không tồn tại bảng course. Sau đó hãy cho biết: 

a. Tổng số sinh viên, điểm trung bình của từng lớp. 

b. Tổng số sinh viên, điểm trung bình của từng môn học. 

c. Phân loại thi đua theo số điểm của từng môn học biết: 

   -i. Điểm TB ≥  9.0: Xuất sắc. 

   -ii. 6.0 ≤ Điểm TB ≤ 8.9: Tốt. 

   -iii. Điểm TB < 6.0: Kém.

```python
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
   OR course_id IS NULL
""")
```

```python
cursor.execute("""
SELECT class,
       COUNT(*) AS total_students,
       AVG(score) AS avg_score
FROM student
GROUP BY class
""")
show(cursor)
```

```python
cursor.execute("""
SELECT c.course_name,
       COUNT(*) AS total_students,
       AVG(s.score) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""")
show(cursor)
```

```python
cursor.execute("""
SELECT c.course_name,
       AVG(s.score) AS avg_score,
       CASE 
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""")
show(cursor)
```

3. Hãy xếp hạng sinh viên thông qua: 

a. Điểm số. 

b. Điểm số theo lớp học. 

c. Điểm số theo mã môn học. 

và cho biết top 3 sinh viện đạt thứ hạng cao nhất, top 3 sinh viên đạt thứ hạng thấp nhất theo 
từng trường hợp trên.

```python
# xếp hạng qua điểm số
cursor.execute("""
SELECT *,
       RANK() OVER (ORDER BY score DESC) AS rank_overall
FROM student
""")
```

```python
cursor.execute("""
SELECT *,
       RANK() OVER (ORDER BY score DESC) AS rank_all
FROM student
""")
show(cursor)
```

```python
cursor.execute("""
SELECT *,
       RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_class
FROM student
""")
show(cursor)
```

```python
cursor.execute("""
SELECT * FROM (
    SELECT *,
           RANK() OVER (ORDER BY score ASC) AS r
    FROM student
)
WHERE r <= 3
""")
print("top 3 sinh viên đạt thứ hạng thấp nhất theo điểm số")
show(cursor)
```

4. Hãy bổ sung thêm một trường graduation_date có kiểu dữ liệu là DATETIME vào bảng 
student để xác định thời gian tốt nghiệp của từng bạn, trong đó thời gian tốt nghiệp được 
xác định bởi thời gian hiện tại cộng với số hạng tương ứng của bạn đó tính theo điểm số.

```python
cursor.execute("""
ALTER TABLE student ADD COLUMN graduation_date TEXT
""")
```

```python
cursor.execute("""
UPDATE student
SET graduation_date = datetime('now', '+' || (
    SELECT r FROM (
        SELECT student_id,
               RANK() OVER (ORDER BY score DESC) AS r
        FROM student
    ) t
    WHERE t.student_id = student.student_id
) || ' days')
""")
conn.commit()
cursor.execute("""
SELECT * FROM student
""")
show(cursor)
```
