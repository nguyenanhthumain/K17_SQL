```python
import sqlite3
import pandas as pd

# Tạo kết nối đến cơ sở dữ liệu SQLite trong bộ nhớ
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Tạo bảng course
cursor.execute('''
CREATE TABLE course (
    id INTEGER PRIMARY KEY,
    course_name TEXT
)''')

# Tạo bảng student
cursor.execute('''
CREATE TABLE student (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)''')

# Chèn dữ liệu vào bảng course
courses = [(12, 'Giai tich'), (34, 'Thong ke'), (26, 'Tin hoc')]
cursor.executemany("INSERT INTO course VALUES (?, ?)", courses)

# Chèn dữ liệu vào bảng student 
students = [
    (1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
    (2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
    (3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
    (4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
    (5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
    (6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
    (7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
    (8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
    (9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
    (10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
]
cursor.executemany("INSERT INTO student VALUES (?, ?, ?, ?, ?)", students)
conn.commit()

print("Khởi tạo cơ sở dữ liệu thành công!")
```

```python
print("1. Tích Decartes:")
df_cross = pd.read_sql_query("SELECT * FROM student, course;", conn)
display(df_cross)
```

```python
print("\n2. INNER JOIN:")
df_inner = pd.read_sql_query("SELECT * FROM student INNER JOIN course ON student.course_id = course.id", conn)
display(df_inner)
```

```python
print("\n3. LEFT JOIN:")
df_left = pd.read_sql_query("SELECT * FROM student LEFT JOIN course ON student.course_id = course.id", conn)
display(df_left)
```

```python
print("\n4. RIGHT JOIN:")
df_right = pd.read_sql_query("SELECT * FROM student RIGHT JOIN course ON student.course_id = course.id", conn)
display(df_right)
```

```python
print("\n5. FULL OUTER JOIN:")
df_full = pd.read_sql_query("SELECT * FROM student FULL OUTER JOIN course ON student.course_id = course.id", conn)
display(df_full)
```

```python
# 2a. Tổng số sinh viên, điểm trung bình của từng lớp
print("\n2a. Thống kê theo lớp:")
df_2a = pd.read_sql_query('''
    SELECT class, COUNT(student_id) as so_hoc_sinh, AVG(score) as Diem_TB
    FROM student GROUP BY class
''', conn)
df_2a
```

```python
# 2b. Tổng số sinh viên, điểm trung bình của từng môn học
print("\n2b. Thống kê theo môn học:")
df_2b = pd.read_sql_query('''
    SELECT course_name as Mon_Hoc, COUNT(student_id) as Tong_SV, ROUND(AVG(s.score), 2) as Diem_TB
    FROM student s JOIN course c ON course_id = id
    GROUP BY course_name
''', conn)
display(df_2b)
```

```python
print("\n2c. Phân loại thi đua:")
df_2c = pd.read_sql_query('''
    SELECT Mon_Hoc, 
           Diem_TB,
           CASE 
               WHEN Diem_TB >= 9.0 THEN 'Xuất sắc'
               WHEN Diem_TB >= 6.0 THEN 'Tốt'
               ELSE 'Kém'
           END as Phan_Loai
    FROM (
        SELECT course.course_name as Mon_Hoc, 
               ROUND(AVG(student.score), 2) as Diem_TB
        FROM student, course 
        WHERE student.course_id = course.id
        GROUP BY course.course_name
    )
''', conn)
df_2c
```

```python
# 3a. Xếp hạng toàn bộ
df_3a=pd.read_sql_query( "SELECT  student_id,name, class, score, RANK() OVER (ORDER BY score DESC) as Ranked FROM student;",conn)
df_3a

```

```python
# 3b. Xếp hạng theo lớp
df_3b=pd.read_sql_query("SELECT  student_id,name, class, score, RANK() OVER (PARTITION BY class ORDER BY score DESC) as Ranked FROM student;",conn)
df_3b
```

```python
# 3c. Xếp hạng theo mã môn học
df_3c=pd.read_sql_query('''
    SELECT s.student_id, s.name, c.course_name, s.score,
           RANK() OVER (PARTITION BY s.course_id ORDER BY s.score DESC) as Ranked
    FROM student s
    JOIN course c ON s.course_id = c.id
''', conn)
df_3c
```

```python
print("-> TOP 3 CAO NHẤT TOÀN TRƯỜNG:")
display(df_3a[df_3a['Ranked'] <= 3])

print("-> TOP 3 THẤP NHẤT TOÀN TRƯỜNG:")
display(df_3a.sort_values(by='score', ascending=True).head(3))
```

```python
print("-> TOP 3 CAO NHẤT TỪNG LỚP:")
display(df_3b[df_3b['Ranked'] <= 3])

print("-> TOP 3 THẤP NHẤT TỪNG LỚP:")
display(df_3b.sort_values(by='score', ascending=True).groupby('class').head(3))
```

```python
print("-> TOP 3 CAO NHẤT TỪNG MÔN:")
display(df_3c[df_3c['Ranked'] <= 3])

print("-> TOP 3 THẤP NHẤT TỪNG MÔN:")
display(df_3c.sort_values(by='score', ascending=True).groupby('course_name').head(3))
```

```python
try:
    cursor.execute("ALTER TABLE student ADD COLUMN graduation_date TEXT")
    print("Đã thêm cột graduation_date thành công!")
except sqlite3.OperationalError:
    print("Cột graduation_date đã tồn tại, bỏ qua bước thêm cột.")

conn.execute('''
    UPDATE student 
    SET graduation_date = date('now', 'localtime', '+' || (
        SELECT COUNT(*) + 1 
        FROM student s2 
        WHERE s2.score > student.score
    ) || ' months');
''')
conn.commit()
print("\n=== BẢNG SINH VIÊN SAU KHI CẬP NHẬT NGÀY TỐT NGHIỆP ===")
df_result = pd.read_sql_query("SELECT * FROM student ORDER BY score DESC", conn)
display(df_result)
```
