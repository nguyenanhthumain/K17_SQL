```python
import sqlite3
import pandas as pd
from datetime import datetime

conn = sqlite3.connect("student_course.db")
cursor = conn.cursor()
```

```python
# Xóa bảng cũ nếu có
cursor.execute("DROP TABLE IF EXISTS student")
cursor.execute("DROP TABLE IF EXISTS course")

# Tạo bảng course
cursor.execute("""
CREATE TABLE course (
    id INTEGER PRIMARY KEY,
    course_name TEXT
)
""")

# Tạo bảng student
cursor.execute("""
CREATE TABLE student (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

# Thêm dữ liệu course
courses = [
    (12, 'Giai tich'),
    (34, 'Thong ke'),
    (26, 'Tin hoc')
]

cursor.executemany("INSERT INTO course VALUES (?, ?)", courses)

# Thêm dữ liệu student
students = [
    (1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
    (2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
    (3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
    (4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
    (5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
    (6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
    (7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
    (8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
    (9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
    (10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
]

cursor.executemany("INSERT INTO student VALUES (?, ?, ?, ?, ?)", students)

conn.commit()
print("Đã tạo xong dữ liệu")
```

```python
print("Bảng student:")
display(pd.read_sql("SELECT * FROM student", conn))

print("Bảng course:")
display(pd.read_sql("SELECT * FROM course", conn))
```

```python
query = """
SELECT *
FROM student s
CROSS JOIN course c
"""
display(pd.read_sql(query, conn))
```

```python
query = """
SELECT s.*, c.course_name
FROM student s
INNER JOIN course c
ON s.course_id = c.id
"""
display(pd.read_sql(query, conn))
```

```python
query = """
SELECT s.*, c.course_name
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
"""
display(pd.read_sql(query, conn))
```

```python
query = """
SELECT s.*, c.course_name
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""
display(pd.read_sql(query, conn))
```

```python
query = """
SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM student s
LEFT JOIN course c ON s.course_id = c.id

UNION

SELECT s.student_id, s.name, s.class, s.course_id, s.score, c.course_name
FROM course c
LEFT JOIN student s ON s.course_id = c.id
"""
display(pd.read_sql(query, conn))
```

```python
# Cập nhật các course_id bị thiếu
cursor.execute("""
UPDATE student
SET course_id = 26
WHERE course_id IS NULL
""")

# Xóa các bản ghi có course_id không tồn tại trong bảng course
cursor.execute("""
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
""")

conn.commit()

display(pd.read_sql("SELECT * FROM student", conn))
```

```python
query = """
SELECT class,
       COUNT(*) AS total_students,
       ROUND(AVG(score), 2) AS avg_score
FROM student
GROUP BY class
"""
display(pd.read_sql(query, conn))
```

```python
query = """
SELECT c.course_name,
       COUNT(*) AS total_students,
       ROUND(AVG(s.score), 2) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
display(pd.read_sql(query, conn))
```

```python
query = """
SELECT c.course_name,
       ROUND(AVG(s.score), 2) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9.0 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6.0 AND AVG(s.score) <= 8.9 THEN 'Tot'
           ELSE 'Kem'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
"""
display(pd.read_sql(query, conn))
```

```python
query = """
SELECT *,
       RANK() OVER (ORDER BY score DESC) AS ranking
FROM student
"""
display(pd.read_sql(query, conn))
```

```python
query = """
SELECT *
FROM (
    SELECT *,
           RANK() OVER (ORDER BY score DESC) AS ranking
    FROM student
)
WHERE ranking <= 3
"""
display(pd.read_sql(query, conn))
```
