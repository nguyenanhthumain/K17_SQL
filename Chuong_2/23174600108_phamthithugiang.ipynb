```python
import sqlite3
import pandas as pd

from IPython.display import display
```

```python
# Dữ liệu
# Bảng student
student_data = [
    (1, "Nguyen Minh Hoang", "May Tinh", 12, 6.7),
    (2, "Tran Thi Lan", "Kinh Te", 34, 9.2),
    (3, "Pham Van Nam", "Toan Tin", None, 7.9),
    (4, "Le Thanh Huyen", "Toan Tin", 20, 7.2),
    (5, "Vu Quoc Anh", "May Tinh", 24, 8.0),
    (6, "Dang Thuy Linh", "May Tinh", 24, 5.5),
    (7, "Bui Tien Dung", "Kinh Te", 34, 9.2),
    (8, "Ho Ngoc Mai", "Toan Tin", 20, 8.8),
    (9, "Duong Huu Phuc", "Kinh Te", None, 7.2),
    (10, "Cao Thi Hanh", "May Tinh", None, 7.0),
]

df_student = pd.DataFrame(student_data, columns=["student_id","name","class","course_id","score"])


# Bảng course
course_data = [
    (12, "Giai tich"),
    (34, "Thong ke"),
    (26, "Tin hoc"),
]

df_course = pd.DataFrame(course_data, columns=["id","course_name"])


# Lưu vào SQLite
conn = sqlite3.connect("school.db")

df_student.to_sql("student", conn, if_exists="replace", index=False)
df_course.to_sql("course", conn, if_exists="replace", index=False)
```

```python
# Tích Descartes
query = """
SELECT *
FROM student
CROSS JOIN course
"""
df = pd.read_sql(query, conn)
display(df)
```

```python
# INNER JOIN
query = """
SELECT s.*, c.course_name
FROM student s
INNER JOIN course c
ON s.course_id = c.id
"""
display(pd.read_sql(query, conn))
```

```python
# LEFT JOIN
query = """
SELECT s.*, c.course_name
FROM student s
LEFT JOIN course c
ON s.course_id = c.id
"""
display(pd.read_sql(query, conn))
```

```python
# RIGHT JOIN
query = """
SELECT s.*, c.course_name
FROM course c
LEFT JOIN student s
ON s.course_id = c.id
"""
display(pd.read_sql(query, conn))
```

```python
# FULL OUTER JOIN
query = """
SELECT s.*, c.course_name
FROM student s
LEFT JOIN course c ON s.course_id = c.id

UNION

SELECT s.*, c.course_name
FROM course c
LEFT JOIN student s ON s.course_id = c.id
"""
display(pd.read_sql(query, conn))
```

```python
query = """
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course);
"""
conn.execute(query)
conn.commit()
```

```python
# Tổng số sinh viên, điểm TB theo lớp
query = """
SELECT class,
       COUNT(*) AS total_students,
       ROUND(AVG(score), 2) AS avg_score
FROM student
GROUP BY class;
"""
display(pd.read_sql(query, conn))
```

```python
# Tổng số sinh viên, điểm TB theo môn học
query = """
SELECT c.course_name,
       COUNT(s.student_id) AS total_students,
       ROUND(AVG(s.score), 2) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name;
"""
display(pd.read_sql(query, conn))
```

```python
# Phân loại thi đua theo môn học
query = """
SELECT c.course_name,
       ROUND(AVG(s.score), 2) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name;
"""
display(pd.read_sql(query, conn))
```

```python
# Xếp hạng theo điểm số
query = """
SELECT student_id, name, class, course_id, score,
       RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student;
"""
display(pd.read_sql(query, conn))
```

```python
# Top 3 cao nhất
query = """
SELECT *
FROM (
    SELECT student_id, name, class, score,
           RANK() OVER (ORDER BY score DESC) AS r
    FROM student
) t
WHERE r <= 3;
"""
display(pd.read_sql(query, conn))
```

```python
# Thêm cột graduation_date
query = """
ALTER TABLE student
ADD COLUMN graduation_date DATETIME;
"""
conn.execute(query)
conn.commit()
```

```python
query = """
UPDATE student
SET graduation_date = (
    SELECT DATE('now', '+' || ranked.r || ' day')
    FROM (
        SELECT student_id,
               RANK() OVER (ORDER BY score DESC) AS r
        FROM student
    ) AS ranked
    WHERE student.student_id = ranked.student_id
);
"""
conn.execute(query)
conn.commit()
```
