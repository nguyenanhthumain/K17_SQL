```
import sqlite3
import pandas as pd

# Tạo kết nối tới database student.db
conn = sqlite3.connect("student.db")

# Tạo con trỏ để thực thi các câu lệnh SQL
cur = conn.cursor()
```

```
# Tạo bảng student
sql = """
CREATE TABLE student(
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
"""
cur.execute(sql)
```

```
# Tạo bảng course
sql = """
CREATE TABLE course(
    id INTEGER,
    course_name TEXT
)
"""
cur.execute(sql)
```

```
# Thêm dữ liệu cho bảng student
students = [
(1,"Nguyen Minh Hoang","May Tinh",12,6.7),
(2,"Tran Thi Lan","Kinh Te",34,9.2),
(3,"Pham Van Nam","Toan Tin",None,7.9),
(4,"Le Thanh Huyen","Toan Tin",20,7.2),
(5,"Vu Quoc Anh","May Tinh",24,8.0),
(6,"Dang Thuy Linh","May Tinh",24,5.5),
(7,"Bui Tien Dung","Kinh Te",34,9.2),
(8,"Ho Ngoc Mai","Toan Tin",20,8.8),
(9,"Duong Huu Phuc","Kinh Te",None,7.2),
(10,"Cao Thi Hanh","May Tinh",None,7.0)
]

for s in students:
    cur.execute("INSERT INTO student VALUES (?,?,?,?,?)",s)

# Thêm dữ liệu cho bảng course
courses = [
(12,"Giai tich"),
(34,"Thong ke"),
(26,"Tin hoc")
]

for c in courses:
    cur.execute("INSERT INTO course VALUES (?,?)",c)

conn.commit()
```

```
# Tích Decartes
query = """
SELECT *
FROM student, course
"""

df = pd.read_sql_query(query, conn)
df
```

```
# INNER JOIN
query = """
SELECT *
FROM student
INNER JOIN course
ON student.course_id = course.id
"""

df = pd.read_sql_query(query, conn)
df
```

```
# LEFT JOIN
query = """
SELECT *
FROM student
LEFT JOIN course
ON student.course_id = course.id
"""

df = pd.read_sql_query(query, conn)
df
```

```
# SQLite không hỗ trợ RIGHT JOIN nên đổi vị trí hai bảng và sử dụng LEFT JOIN
query = """
SELECT *
FROM course
LEFT JOIN student
ON student.course_id = course.id
"""

df = pd.read_sql_query(query, conn)
df
```

```
# SQLite không hỗ trợ FULL OUTER JOIN nên sử dụng UNION của LEFT JOIN hai chiều
query = """
SELECT *
FROM student
LEFT JOIN course
ON student.course_id = course.id

UNION

SELECT *
FROM course
LEFT JOIN student
ON student.course_id = course.id
"""

df = pd.read_sql_query(query, conn)
df
```

```
# Xóa các sinh viên có course_id không tồn tại trong bảng course
sql = """
DELETE FROM student
WHERE course_id NOT IN (SELECT id FROM course)
"""
cur.execute(sql)
conn.commit()
```

```
# Cập nhật các sinh viên chưa có course_id (NULL) thành 26
sql = """
UPDATE student
SET course_id = 26
WHERE course_id IS NULL
"""

cur.execute(sql)
conn.commit()
```

```
# Thống kê số sinh viên và điểm trung bình theo từng lớp
query = """
SELECT class,
COUNT(*) as tong_sv,
AVG(score) as diem_tb
FROM student
GROUP BY class
"""

df = pd.read_sql_query(query, conn)
df
```

```
# Thống kê số sinh viên và điểm trung bình theo từng môn học
query = """
SELECT course_name,
COUNT(student_id) as tong_sv,
AVG(score) as diem_tb
FROM student
JOIN course
ON student.course_id = course.id
GROUP BY course_name
"""

df = pd.read_sql_query(query, conn)
df
```

```
# Phân loại thi đua dựa trên điểm trung bình của từng môn học
query = """
SELECT course_name,
AVG(score) as diem_tb,

CASE
WHEN AVG(score) >= 9 THEN 'Xuat sac'
WHEN AVG(score) >= 6 THEN 'Tot'
ELSE 'Kem'
END as xep_loai

FROM student
JOIN course
ON student.course_id = course.id

GROUP BY course_name
"""

df = pd.read_sql_query(query, conn)
df
```

```
# Xếp hạng sinh viên theo điểm số từ cao xuống thấp
query = """
SELECT *,
RANK() OVER (ORDER BY score DESC) as rank_score
FROM student
"""

df = pd.read_sql_query(query, conn)
df
```

```
# Xếp hạng sinh viên trong từng lớp
query = """
SELECT *,
RANK() OVER (PARTITION BY class ORDER BY score DESC) as rank_class
FROM student
"""

df = pd.read_sql_query(query, conn)
df
```

```
# Xếp hạng sinh viên theo mã môn học
query = """
SELECT *,
RANK() OVER (PARTITION BY course_id ORDER BY score DESC) as rank_course
FROM student
"""

df = pd.read_sql_query(query, conn)
df
```
