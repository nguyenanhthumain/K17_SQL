```python
import sqlite3

conn = sqlite3.connect('Student.db')
cursor = conn.cursor()

cursor.execute("DROP TABLE IF EXISTS Student")

cursor.execute('''
CREATE TABLE Student (
    student_id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    class TEXT,
    course_id INTEGER,
    score REAL
)
''')

students_data = [
    (1, 'Nguyen Minh Hoang', 'May Tinh', 12, 6.7),
    (2, 'Tran Thi Lan', 'Kinh Te', 34, 9.2),
    (3, 'Pham Van Nam', 'Toan Tin', None, 7.9),
    (4, 'Le Thanh Huyen', 'Toan Tin', 20, 7.2),
    (5, 'Vu Quoc Anh', 'May Tinh', 24, 8.0),
    (6, 'Dang Thuy Linh', 'May Tinh', 24, 5.5),
    (7, 'Bui Tien Dung', 'Kinh Te', 34, 9.2),
    (8, 'Ho Ngoc Mai', 'Toan Tin', 20, 8.8),
    (9, 'Duong Huu Phuc', 'Kinh Te', None, 7.2),
    (10, 'Cao Thi Hanh', 'May Tinh', None, 7.0)
]

cursor.executemany('''
INSERT INTO Student (student_id, name, class, course_id, score)
VALUES (?, ?, ?, ?, ?)
''', students_data)

conn.commit()

cursor.execute("SELECT * FROM Student")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

```python
cursor.execute("DROP TABLE IF EXISTS course")

cursor.execute('''
CREATE TABLE course (
    id INTEGER PRIMARY KEY,
    course_name TEXT NOT NULL
)
''')

courses_data = [
    (12, 'Giai tich'),
    (34, 'Thong ke'),
    (26, 'Tin hoc')
]

cursor.executemany('''
INSERT INTO course (id, course_name)
VALUES (?, ?)
''', courses_data)

conn.commit()

cursor.execute("SELECT * FROM course")
rows = cursor.fetchall()

for row in rows:
    print(row)
```

# Câu 1:

```python
# INNER JOIN
print("\n2. INNER JOIN")
cursor.execute("SELECT * FROM Student INNER JOIN course ON Student.course_id = course.id")
for row in cursor.fetchall():
    print(row)

# LEFT JOIN
print("\n3. LEFT JOIN")
cursor.execute("SELECT * FROM Student LEFT JOIN course ON Student.course_id = course.id")
for row in cursor.fetchall():
    print(row)

# RIGHT JOIN
print("\n4. RIGHT JOIN")
try:
    cursor.execute("SELECT * FROM Student RIGHT JOIN course ON Student.course_id = course.id")
    for row in cursor.fetchall():
        print(row)
except sqlite3.OperationalError as e:
    print("Lỗi (SQLite bản cũ chưa hỗ trợ):", e)

# FULL OUTER JOIN
print("\n5. FULL OUTER JOIN")
try:
    cursor.execute("SELECT * FROM Student FULL OUTER JOIN course ON Student.course_id = course.id")
    for row in cursor.fetchall():
        print(row)
except sqlite3.OperationalError as e:
    print("Lỗi (SQLite bản cũ chưa hỗ trợ):", e)
    
# Tích Descartes
print("\n1. Tích Descartes")
cursor.execute("SELECT * FROM Student CROSS JOIN course")
for row in cursor.fetchall():
    print(row)
```

# Câu 2:

```python
cursor.execute("UPDATE Student SET course_id = 26 WHERE course_id IS NULL")

cursor.execute("DELETE FROM Student WHERE course_id NOT IN (SELECT id FROM course)")

conn.commit()

#a. Tính tổng số sinh viên và điểm trung bình của từng lớp
print("Thống kê: Lớp | Tổng SV | Điểm TB")


cursor.execute('''
SELECT class, COUNT(*), AVG(score) 
FROM Student 
GROUP BY class
''')

for row in cursor.fetchall():
    lop = row[0]
    tong_sv = row[1]
    diem_tb = round(row[2], 2)
    print(f"{lop:<10} | {tong_sv:<7} | {diem_tb}")
    
#b. Tính tổng số sinh viên và điểm trung bình của môn
print("Thống kê: Môn | Tổng SV | Điểm TB")

cursor.execute('''
SELECT course_id, COUNT(*), AVG(score) 
FROM Student 
GROUP BY course_id
''')

for row in cursor.fetchall():
    mon = row[0]
    tong_sv = row[1]
    diem_tb = round(row[2], 2)
    print(f"{mon:<10} | {tong_sv:<7} | {diem_tb}")

#c. phân loại
print("Phân loại thi đua từng môn học:")

cursor.execute('''
SELECT 
    c.course_name, 
    ROUND(AVG(s.score), 2) as diem_tb,
    CASE 
        WHEN AVG(s.score) >= 9.0 THEN 'Xuất sắc'
        WHEN AVG(s.score) >= 6.0 AND AVG(s.score) <= 8.9 THEN 'Tốt'
        ELSE 'Kém'
    END as xep_loai
FROM Student s
JOIN course c ON s.course_id = c.id
GROUP BY c.id, c.course_name
''')

for row in cursor.fetchall():
    ten_mon = row[0]
    diem_tb = row[1]
    xep_loai = row[2]
    print(f"Môn: {ten_mon:<10} | Điểm TB: {diem_tb:<5} | Phân loại: {xep_loai}")
```

```python
# a. Xếp hạng theo Điểm số
print("\na. Xếp hạng theo Điểm số:")
print("Top 3 cao nhất:")
cursor.execute("SELECT name, score FROM Student ORDER BY score DESC LIMIT 3")
for row in cursor.fetchall():
    print(row)

print("Top 3 thấp nhất:")
cursor.execute("SELECT name, score FROM Student ORDER BY score ASC LIMIT 3")
for row in cursor.fetchall():
    print(row)


# b. Xếp hạng theo Lớp học
print("\nb. Xếp hạng theo Lớp học")
print("Top 3 cao nhất mỗi lớp:")
cursor.execute('''
SELECT class, name, score 
FROM (
    SELECT class, name, score, RANK() OVER(PARTITION BY class ORDER BY score DESC) as rank 
    FROM Student
) WHERE rank <= 3
''')
for row in cursor.fetchall():
    print(row)

print("Top 3 thấp nhất mỗi lớp:")
cursor.execute('''
SELECT class, name, score 
FROM (
    SELECT class, name, score, RANK() OVER(PARTITION BY class ORDER BY score ASC) as rank 
    FROM Student
) WHERE rank <= 3
''')
for row in cursor.fetchall():
    print(row)


# c. Xếp hạng theo Mã môn học
print("\nc. Xếp hạng theo Mã môn học")
print("Top 3 cao nhất mỗi môn:")
cursor.execute('''
SELECT course_id, name, score 
FROM (
    SELECT course_id, name, score, RANK() OVER(PARTITION BY course_id ORDER BY score DESC) as rank 
    FROM Student
) WHERE rank <= 3
''')
for row in cursor.fetchall():
    print(row)

print("Top 3 thấp nhất mỗi môn:")
cursor.execute('''
SELECT course_id, name, score 
FROM (
    SELECT course_id, name, score, RANK() OVER(PARTITION BY course_id ORDER BY score ASC) as rank 
    FROM Student
) WHERE rank <= 3
''')
for row in cursor.fetchall():
    print(row)
```
