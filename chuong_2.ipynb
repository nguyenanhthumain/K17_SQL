##1. Hãy kết nối hai bảng trên theo những cách sau:
- Sử dụng tích Decartes.
- Sử dụng JOIN: INNER JOIN, LEFT JOIN, RIGHT JOIN, FULL OUTER JOIN.

```python
df = pd.read_sql_query("""
SELECT * FROM student CROSS JOIN course
""", conn)
display(df)
```

###INNER JOIN

```python
df = pd.read_sql_query("""
SELECT * FROM student s
INNER JOIN course c ON s.course_id = c.id
""", conn)
display(df)
```

###LEFT JOIN

```python
df = pd.read_sql_query("""
SELECT * FROM student s
LEFT JOIN course c ON s.course_id = c.id
""", conn)
display(df)
```

###RIGHT JOIN

```python
df = pd.read_sql_query("""
SELECT * FROM course c
LEFT JOIN student s ON s.course_id = c.id
""", conn)
display(df)
```

###FULL OUTER JOIN

```python
df = pd.read_sql_query("""
SELECT * FROM student s
LEFT JOIN course c ON s.course_id = c.id

UNION

SELECT * FROM course c
LEFT JOIN student s ON s.course_id = c.id
""", conn)
display(df)
```

###a. Tổng số sinh viên, điểm trung bình của từng lớp.

```python
df = pd.read_sql_query("""
SELECT class,
       COUNT(*) AS total_students,
       ROUND(AVG(score),2) AS avg_score
FROM student
GROUP BY class
""", conn)
display(df)
```

###b. Tổng số sinh viên, điểm trung bình của từng môn học.

```python
df = pd.read_sql_query("""
SELECT c.course_name,
       COUNT(*) AS total_students,
       ROUND(AVG(score),2) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)
display(df)
```

###c. Phân loại thi đua theo số điểm của từng môn học

```python
df = pd.read_sql_query("""
SELECT c.course_name,
       ROUND(AVG(score),2) AS avg_score,
       CASE
           WHEN AVG(score) >= 9 THEN 'Xuat sac'
           WHEN AVG(score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END AS classification
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)
display(df)
```

##3. Hãy xếp hạng sinh viên

###a. Điểm số

```python
df = pd.read_sql_query("""
SELECT *,
       RANK() OVER (ORDER BY score DESC) AS rank_score
FROM student
""", conn)
display(df)
```

###b. Điểm số theo lớp học.

```python
df = pd.read_sql_query("""
SELECT *,
       RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_class
FROM student
""", conn)
display(df)
```

###c. Điểm số theo mã môn học.

```python
df = pd.read_sql_query("""
SELECT *,
       RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_course
FROM student
""", conn)
display(df)
```

###Top 3 sinh viện đạt thứ hạng cao nhất

```python
df = pd.read_sql_query("""
SELECT * FROM (
    SELECT *,
           RANK() OVER (ORDER BY score DESC) AS r
    FROM student
)
WHERE r <= 3
""", conn)
display(df)
```

### Top 3 sinh viên đạt thứ hạng thấp nhất

```python
df = pd.read_sql_query("""
SELECT * FROM (
    SELECT *,
           RANK() OVER (ORDER BY score ASC) AS r
    FROM student
)
WHERE r <= 3
""", conn)
display(df)
```

##4. Hãy bổ sung thêm một trường graduation_date có kiểu dữ liệu là DATETIME vào bảng student để xác định thời gian tốt nghiệp của từng bạn, trong đó thời gian tốt nghiệp được xác định bởi thời gian hiện tại cộng với số hạng tương ứng của bạn đó tính theo điểm số.

```python
cursor.execute("""
ALTER TABLE student
ADD COLUMN graduation_date TEXT
""")
cursor.execute("""
UPDATE student
SET graduation_date = DATETIME('now', '+' || (
    SELECT rnk FROM (
        SELECT student_id,
               RANK() OVER (ORDER BY score DESC) AS rnk
        FROM student
    ) tmp
    WHERE tmp.student_id = student.student_id
) || ' days')
""")
df = pd.read_sql_query("SELECT * FROM student", conn)
display(df)
```
