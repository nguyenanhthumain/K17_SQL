# Tạo database và bảng

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Tạo bảng student
cursor.execute("""
CREATE TABLE student (
    student_id INTEGER,
    name TEXT,
    class TEXT,
    course_id INTEGER,
    score REAL
)
""")

# Tạo bảng course
cursor.execute("""
CREATE TABLE course (
    id INTEGER,
    course_name TEXT
)
""")
```

# Insert dữ liệu

```python
students = [
(1,"Nguyen Minh Hoang","May Tinh",12,6.7),
(2,"Tran Thi Lai","Kinh Te",34,9.2),
(3,"Pham Van Nam","Toan Tin",None,7.9),
(4,"Le Thanh Huyen","Toan Tin",20,7.2),
(5,"Vu Quoc Anh","May Tinh",24,8.0),
(6,"Dang Thuy Linh","May Tinh",24,5.5),
(7,"Bui Tien Dung","Kinh Te",34,9.2),
(8,"Ho Ngoc Mai","Toan Tin",20,8.8),
(9,"Duong Huu Phuc","Kinh Te",34,7.2),
(10,"Cao Thi Hanh","May Tinh",None,7.0)
]

courses = [
(12,"Giai tich"),
(34,"Thong ke"),
(20,"Tin hoc")
]

cursor.executemany("INSERT INTO student VALUES (?,?,?,?,?)", students)
cursor.executemany("INSERT INTO course VALUES (?,?)", courses)

conn.commit()
```

# JOINS

## CROSS JOIN

```python
pd.read_sql_query("""
SELECT * 
FROM student CROSS JOIN course
""", conn)
```

## INNER JOIN

```python
pd.read_sql_query("""
SELECT * 
FROM student s
INNER JOIN course c ON s.course_id = c.id
""", conn)
```

## LEFT JOIN

```python
pd.read_sql_query("""
SELECT * 
FROM student s
LEFT JOIN course c ON s.course_id = c.id
""", conn)
```

## RIGHT JOIN

```python
pd.read_sql_query("""
SELECT * 
FROM course c
LEFT JOIN student s ON s.course_id = c.id
""", conn)
```

## FULL OUTER JOIN

```python
pd.read_sql_query("""
SELECT * FROM student s
LEFT JOIN course c ON s.course_id = c.id

UNION

SELECT * FROM course c
LEFT JOIN student s ON s.course_id = c.id
""", conn)
```

# Thống kê theo lớp 

```python
pd.read_sql_query("""
SELECT class, 
       COUNT(*) AS total_students,
       ROUND(AVG(score),2) AS avg_score
FROM student
WHERE course_id IN (SELECT id FROM course)
GROUP BY class
""", conn)
```

# Thống kê theo môn 

```python
pd.read_sql_query("""
SELECT c.course_name,
       COUNT(*) AS total_students,
       ROUND(AVG(s.score),2) AS avg_score
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)
```

# Phân loại

```python
pd.read_sql_query("""
SELECT c.course_name,
       ROUND(AVG(s.score),2) AS avg_score,
       CASE
           WHEN AVG(s.score) >= 9 THEN 'Xuat sac'
           WHEN AVG(s.score) >= 6 THEN 'Tot'
           ELSE 'Kem'
       END AS ranking
FROM student s
JOIN course c ON s.course_id = c.id
GROUP BY c.course_name
""", conn)
```

# Xếp hạng toàn bộ

```python
pd.read_sql_query("""
SELECT *,
RANK() OVER (ORDER BY score DESC) AS rank_all
FROM student
WHERE course_id IN (SELECT id FROM course)
""", conn)
```

# Xếp hạng theo lớp

```python
pd.read_sql_query("""
SELECT *,
RANK() OVER (PARTITION BY class ORDER BY score DESC) AS rank_class
FROM student
WHERE course_id IN (SELECT id FROM course)
""", conn)
```

# Xếp hạng theo môn

```python
pd.read_sql_query("""
SELECT *,
RANK() OVER (PARTITION BY course_id ORDER BY score DESC) AS rank_course
FROM student
WHERE course_id IN (SELECT id FROM course)
""", conn)
```

# Top 3 cao & thấp

```python
# Top 3 sinh viên có điểm cao nhất
pd.read_sql_query("""
SELECT *
FROM student
WHERE course_id IN (SELECT id FROM course)
ORDER BY score DESC
LIMIT 3
""", conn)
```

```python
# Top3 sinh viên có điểm thấp nhất
pd.read_sql_query("""
SELECT *
FROM student
WHERE course_id IN (SELECT id FROM course)
ORDER BY score ASC
LIMIT 3
""", conn)
```

# Cập nhật graduation_date theo rank

```python
pd.read_sql_query("""
SELECT student_id, name, score,
DATETIME('now', '+' || 
        RANK() OVER (ORDER BY score DESC) || ' days') AS graduation_date
FROM student
WHERE course_id IN (SELECT id FROM course)
""", conn)
```
